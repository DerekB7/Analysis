**ex 0.4.2**

using Theorem 0.2.4

$L\big(f,P_\delta\big) = 1 \cdot(2+\delta) + 2\cdot(1-\delta) = 4-\delta $   
$U\big(f,P_\delta\big) = 1 \cdot(2-\delta) + 2\cdot(1+\delta) = 4+\delta $   

$\implies U\big(f, P_\delta\big) -L\big(f, P_\delta\big)\leq 2 \delta  \lt \epsilon$  
for any $\epsilon\gt 0$ by setting e.g. $\delta:= \frac{\epsilon}{3}$

$4-\delta\leq \sup_P L(f,P) \leq 4+\delta$ and $4-\delta\leq \inf_P U(f,P) \leq 4+\delta\implies \sup_P L(f,P) = 4=\inf_P U(f,P) $

**ex 0.4.6** 
let $\{f_n\}$ be a sequence of functions with $f_n \in R[a,b]$ for each $n$.  Suppose the sequence $\{f_n\}$ converges uniformly to $f$ on $[a,b]$.  Show that $f\in R[a,b]$  

Applying Theorem 0.2.4, we need to show, for any $\epsilon \gt 0$ that $U(f,P)-L(f,P)\lt \epsilon$   

$U(f,P)-L(f,P)= \big \vert U(f,P)-L(f,P)\big \vert = \big \vert \big(U(f,P)- U(f_n,P)\big) + \big(U(f_n,P) -L(f_n,P)\big) + \big(L(f_n,P) -L(f,P)\big)\big \vert$  
$\leq \big \vert U(f,P)- U(f_n,P)\big \vert  +\big \vert U(f_n,P) -L(f_n,P)\big \vert + \big \vert U(f_n,P) -L(f,P)\big \vert $  
$\lt \frac{\epsilon}{3}+\frac{\epsilon}{3}+\frac{\epsilon}{3}$  
$=\epsilon$  

where $\big \vert f_n(x)-f(x)\big \vert \lt \frac{\epsilon}{(b-a)\cdot 3}$ for all $n\geq N$  by uniform convergence. So fix some $n$ and  
$\big \vert U(f_n,P) -L(f_n,P)\big \vert \lt \frac{\epsilon}{3}$ by choosing a suitable partition $P$ since $f_n$ is Riemann integrable (Theorem 0.2.4) and  


$\big \vert U(f,P)- U(f_n,P)\big \vert  = \big \vert \sum_{i=1}^n (M_i-M_i^{(f_n)})\cdot (x_i-x_{i-1})\big \vert\leq  \sum_{i=1}^n \big \vert M_i-M_i^{(f_n)}\big \vert \cdot (x_i-x_{i-1})\leq  \frac{\epsilon}{(b-a)\cdot 3} \cdot \sum_{i=1}^n  (x_i-x_{i-1})=  \frac{\epsilon}{3}$  
$\big \vert L(f,P)- L(f_n,P)\big \vert  = \big \vert \sum_{i=1}^n (m_i-m_i^{(f_n)})\cdot (x_i-x_{i-1})\big \vert\leq  \sum_{i=1}^n \big \vert m_i-m_i^{(f_n)}\big \vert \cdot (x_i-x_{i-1})\leq  \frac{\epsilon}{(b-a)\cdot 3} \cdot \sum_{i=1}^n  (x_i-x_{i-1})=  \frac{\epsilon}{3}$  

**technical commentary:**  
the way to interpret e.g. $\big \vert M_2 - M_2^{(f_n)}\big \vert \leq \frac{\epsilon}{(b-a)\cdot 3}$ is that while 
the supremum may not exist in the set, it is the limit of a sequence of points in that set so  
$M_2 = \lim_{k\to \infty} f(x^{(k)})$ and $M_2^{(f_n)} = \lim_{j\to \infty} f_n(x^{(k)})$

so if $M_2 \geq  M_2^{(f_n)}$ then  
$M_2 -  M_2^{(f_n)} = \big \vert M_2 -  M_2^{(f_n)}\big \vert = \lim_{k\to \infty}\big \vert f(x^{(k)}) - M_2^{(f_n)}\big \vert \leq \lim_{k\to \infty}\big \vert f(x^{(k)}) - f_n(x^{(k)})\big \vert \leq \frac{\epsilon}{(b-a)\cdot 3}$   
where the first inequality holds by definition of the supremum and the second since $\big \vert f(x_k) - f_n(x_k)\big \vert \lt \frac{\epsilon}{(b-a)\cdot 3}$ for all $k$; the case of $M_2 \leq  M_2^{(f_n)}$ follows similarly          

and for avoidance of doubt, when dealing with bounding $\big \vert m_2 - m_2^{(f_n)}\big \vert$  
consider if $m_2 \geq  m_2^{(f_n)}$ then  
$m_2 -  m_2^{(f_n)} = \big \vert m_2 -  m_2^{(f_n)}\big \vert = \lim_{j\to \infty}\big \vert m_2 - f_n(x^{(j)})\big \vert \leq \lim_{j\to \infty}\big \vert f(x^{(j)}) - f_n(x^{(j)})\big \vert \leq \frac{\epsilon}{(b-a)\cdot 3}$   


**remark:**  
$\int_a^b f_n(x)dx\to \int_a^b f(x)dx$  
write $g_n: [a,b]\longrightarrow \mathbb R$ given by $g_n(x):= f_n(x) -f(x)$  
and for any $\epsilon \gt 0$, then for all $n$ large enough 
$\vert g_n(x) \vert \lt \epsilon \iff -\epsilon \lt g_n(x)\lt \epsilon$  

Applying Theorem 0.3.1 (exercises 14 and 15) we have   
$-\epsilon\cdot (b-a)\lt \int_{a}^b g_n(x) dx = \int_{a}^b f_n(x) dx - \int_{a}^b f(x) dx\lt \epsilon\cdot (b-a)$  
where the upper and lower bounds are from Theorem 0.3.2 (ex 3, compare $g_n$ to the constant function $=\epsilon$) with $\int_{a}^b \epsilon dx = \epsilon\cdot (b-a)$ since $\epsilon \cdot (b-a) \leq L(f,P)\leq U(f,P)\leq \epsilon \cdot (b-a)$ for any partition $P$   
$\implies \big \vert \int_{a}^b f_n(x) dx - \int_{a}^b f(x) dx \big \vert \lt \epsilon$ for all $n$ large enough

**ex 0.4.9**  
Prove or give a counterexample: Suppose $f\in R[a,b]$ and there exists $k\gt 0$ such that $f(x)\geq k$ for all $x \in[a,b]$.  Then $\frac{1}{f}\in R[a,b]$.  

The statement is true; using Theorem 0.3.1 (ex 15) it suffices to show the Riemann Integrability of $\frac{1}{g}$ with $g(x):= \frac{f(x)}{k}$ (equivalently WLOG $f(x) \geq 1$).  

Since $g\in R[a,b]$ we apply Theorem 0.2.4: for any $\epsilon  \gt 0$ we have some partition $P$ such that  (using Def 0.1.5)  
$\sum_{i=1}^n (M_i-m_i)\cdot (x_i - x_{i-1})= U(f,P)-L(f,P)\lt \epsilon$  
$\implies  \sum_{i=1}^n \big(\frac{1}{m_i}-\frac{1}{M_i}\big)\cdot (x_i - x_{i-1}) \leq  \sum_{i=1}^n (M_i-m_i)\cdot (x_i - x_{i-1})\lt \epsilon$  

since for $x\geq y\geq 1$ we have $xy\cdot\big(\frac{1}{y}-\frac{1}{x}\big)= x - y\leq xy\cdot \big(x-y\big)$  which holds since $1\leq xy$  

$\implies U(g,P)-L(f,p)= \sum_{i=1}^n \big(\frac{1}{m_i}-\frac{1}{M_i}\big)\cdot (x_i - x_{i-1}) \lt\epsilon$  
i.e. with $J:\mathbb R_{\gt 0}\rightarrow \mathbb R_{\gt 0}$ given by $J(x) =\frac{1}{x}$ we observe that e.g. for $M_i$ we have $f(x)\leq M_i$ for all $x\in [x_{i-1},x_i]$ and $f\big(x^{(k)}\big)\to M_i$ for some sequence.  Now $J\big(f(x)\big)\geq \frac{1}{M_i}$ for all $x\in [x_{i-1},x]$ by algebraic properties of the (positive) inversion map and $g\big(x^{(k)}\big)=J\Big(f\big(x^{(k)}\big)\Big)\to J\big(M_i\big)=\frac{1}{M_i}$ by continuity of $J$.  I.e. $\frac{1}{M_i}=m_i^{(g)}$ and by nearly identical reasoning $\frac{1}{m_i}=M_i^{(g)}$.  

*remark:*  
The fact that $f$ is bounded away from zero is vital.  if it was not, then $\frac{1}{f}$ would not be a bounded function hence per Def 0.1.9 $\frac{1}{f}$ would not be Riemann integrable.
