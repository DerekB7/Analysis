Technical detail on Monotone Convergence Theorem (Theorem 4.3.9 on page 181)   

the proof argues that since $0\leq f_n(x)\leq f_{n+1}(x) \leq f(x)$ a.e. wrt $(\mu)$ for $n\in \mathbb N$  then by Prop 4.3.6 (stated on page 176 but left as exercise 15)  
$\int f_n d\mu \leq \int f d\mu$  
*but Prop 4.3.6 requires both of these functions to be integrable with respect to* $\mu$.  

*(i.)*  $\int f_n = \infty$ case  
Checking Def 4.3.2 (page 173), if $\int f_n d\mu =\infty$ then $f_n$ is not integrable however that would mean we have sequence of $\phi_k(x) \leq f_n(x)\leq f_{n+1}(x)\leq f(x) \text{ a.e. }(\mu)$ where $\int \phi_k d\mu \to \infty\implies $  that same sequence can be used (perhaps with adjustments in each case on a set of measure zero which does not affect the integral of a simple function) to show $\int f_{n+1} d\mu=\infty = \int f d\mu$   

*(ii.)* assume $\int f_n d \mu \lt\infty $ for all $n$.  
the $f_n$ are known to be measurable but $f$ is not.  If $f$ is measurable and $\int f d\mu =\infty$ then Theorem 4.3.7 (preliminary Fatou's Lemma) tells us $\infty = \int f d\mu \leq \liminf_{n\to  \infty}\int f_n d\mu \leq \lim_{n\to  \infty}\int f_n d\mu$ so Monotone Convergence holds in a somewhat trivial way.  The main case of  $\int f d\mu\lt\infty$ is shown in the text but that **still requires we show** $f$ **is measurable**.  First construct $g_n\leq g_{n+1}$ where $g_n \to g$ everywhere by letting $S_0$ be the set of $x$'s where $f_n(x)\not\to f(x)$   and $S_n$ contain the set of $x$'s where $f_n(x)\not \leq f_{n+1}(x)$.  
$S:= \bigcup_{k=0}^\infty S_k\implies \mu\big(S\big)\leq \sum_{k=0}^\infty \mu\big(S_k\big) = 0$ (Theorem 4.1.18)  
set $g(x)=g_1(x)=g_2(x)=g_3(x) \dots = 0$ for $x\in S$  

So select $s\in \mathbb R$ and  
$A:= \big\{x \big \vert g(x)\gt s\big\}$  and $A_n:= \big\{x \big \vert g_n(x)\gt s\big\}$  
We since $\big\{g_n\big\}$ is a monotone sequence we have the filtration $A_1 \subseteq A_2 \subseteq A_3 \subseteq \dots$.  

$p \in A_k\implies p\in A$ since $g$ is monotone and $p \in A\implies p\in A_m$ for (almost all $m$) by the definition of $g_n \to g$ pointwise.  
Thus $A= \bigcup_{k=0}^\infty A_k$ is measurable. Finally  
$\big\{x \big \vert f(x)\gt s\big\} = \big\{x \big \vert g(x)=f(x)\gt s\big\} \cup \big\{x \big \vert g(x)\neq f(x)\gt s\big\} = A \cup \Big(\big\{x \big \vert  f(x)\gt s\big\}\cap S_0\Big) $

$\implies \int_x G d\mu =\lim_{k\to \infty}\int_X \psi_k d\mu \leq \int_X g^{+} d\mu\lt \infty$   
$\implies G$ is integrable since this supremum is finite  

applying def 4.3.2  
$\int_x F d\mu =\sup\Big\{\int_X \psi d\mu \big \vert \psi\text{ is a simple function with }0\leq \psi \leq F\Big\}$
