Technical detail on Monotone Convergence Theorem (Theorem 4.3.9 on page 181)   

the proof argues that since $0\leq f_n(x)\leq f_{n+1}(x) \leq f(x)$ a.e. wrt $(\mu)$ for $n\in \mathbb N$  then by Prop 4.3.6 (stated on page 176 but left as exercise 15)  
$\begin{align}\int f_n d\mu \leq \int f d\mu\end{align}$  
*but Prop 4.3.6 requires both of these functions to be integrable with respect to* $\mu$.  

*(i.)*  $\begin{align}\int f_n = \infty\end{align}$ case  
Checking Def 4.3.2 (page 173), if $\begin{align}\int f_n d\mu =\infty\end{align}$ then $f_n$ is not integrable however that would mean we have sequence of $\phi_k(x) \leq f_n(x)\leq f_{n+1}(x)\leq f(x) \text{ a.e. }(\mu)$ where $\begin{align}\int \phi_k d\mu \to \infty\implies \end{align}$  that same sequence can be used (perhaps with adjustments in each case on a set of measure zero which does not affect the integral of a simple function) to show $\begin{align} \int f_{n+1} d\mu=\infty = \end{align}\int f d\mu$   

*(ii.)* assume $\begin{align}\int f_n d \mu \lt\infty\end{align} $ for all $n$.  
the $f_n$ are known to be measurable but $f$ is not.  If $f$ is measurable and $\begin{align}\int f d\mu =\infty\end{align}$ then Theorem 4.3.7 (preliminary Fatou's Lemma) tells us $\begin{align}\infty = \int f d\mu \leq \liminf_{n\to  \infty}\int f_n d\mu \leq \lim_{n\to  \infty}\int f_n d\mu\end{align}$ so Monotone Convergence holds in a somewhat trivial way.  The main case of  $\begin{align}\int f d\mu\lt\infty\end{align}$ is shown in the text but that **still requires we show** $f$ **is measurable**.  Since $f_n\leq f_{n+1}$ a.e. ($\mu$) and $f_n\to f$ a.e. ($\mu$) we want to 'fix' this by constructing $g$ that is non-decreasing everywhere and $g_n\to g$ everywhere.  So construct $g_n\leq g_{n+1}$ where $g_n \to g$ everywhere by letting $S_0$ be the set of $x$'s where $f_n(x)\not\to f(x)$   and $S_n$ contain the set of $x$'s where $f_n(x)\not \leq f_{n+1}(x)$.  
$S:= \bigcup_{k=0}^\infty S_k\implies \mu\big(S\big)\leq \sum_{k=0}^\infty \mu\big(S_k\big) = 0$ (Theorem 4.1.18)  
set $g(x)=g_1(x)=g_2(x)=g_3(x) \dots = 0$ for $x\in S$  

So select $s\in \mathbb R$ and  
$A:= \big\{x \big \vert g(x)\gt s\big\}$  and $A_n:= \big\{x \big \vert g_n(x)\gt s\big\}$  
Since $\big\{g_n\big\}$ is a monotone sequence we have the filtration $A_1 \subseteq A_2 \subseteq A_3 \subseteq \dots$.  

$p \in A_k\implies p\in A$ since $g$ is monotone and $p \in A\implies p\in A_m$ for (almost all $m$) by the definition of $g_n \to g$ pointwise.  
Thus $A= \bigcup_{k=0}^\infty A_k$ is measurable. Which shows the nice case, i.e. that $g$ is measurable.  

It follows that $f$ is measurable since $f=g$ a.e. ($\mu$).  I.e.  
$\big\{x \big \vert f(x)\gt s\big\} = \big\{x \big \vert g(x)=f(x)\gt s\big\} \cup \big\{x \big \vert g(x)\neq f(x)\gt s\big\} = A \cup \Big(\big\{x \big \vert  f(x)\gt s\big\}\cap S\Big) $    
since $A$ is measurable and $\Big(\big\{x \big \vert  f(x)\gt s\big\}\cap S\Big)\subset S$ has measure zero and thus is a measurable set since we are in a complete measure space -- recall that is assumed throughout section 4.3 per top of page 173)  

**effect of a linear change of variables for Lebesgue Integrals**  
let $T \in GL_n\big(\mathbb R^n)\big)$, ref 'Jones_general_notes.ipynb' we know for Lebesgue measurable $A\subseteq X$ that $m\big(T(A)\big) = \big \vert \det T\big \vert\cdot m\big(A\big)$.  

For Lebesgue integrable $f$ and since we consider $f^+$ and $f^-$ separately we may as well focus on the integral $\begin{align} \int f^+(x) \end{align}$   

Applying Theorem 4.2.12 we have a monotone increasing sequence $\big\{\phi_n\big\}$ of simple functions such that $\phi_n(x)\to f^+(x)$ for all $x \in X$ 
(where $X=\mathbb R^n$ here or we must otherwise impose $T(X) \subseteq X$ for the question to be sensible)

For single characteristic function, using results from chapter 2 (e.g. ex 2.6.15) or chapter 4 (Def 4.3.1), we have 

For measurable $A$,  
$\mathcal X_A\circ T =  \mathcal X_{T^{-1}(A)}$   
$\begin{align}\big\vert\det T\big\vert^{-1} \cdot \int_{\mathbb R^n} \mathcal X_A(x) dx=\big\vert\det T\big\vert^{-1} \cdot m\big(A\big) =\big\vert\det\big(T^{-1}\big)\big\vert \cdot m\big(A\big) = m\big(T^{-1}(A)\big)=\int_{\mathbb R^n} \mathcal X_{T^{-1}(A)} dx=\int_{\mathbb R^n} \mathcal X_A\circ T(x) dx\end{align}$    
$\begin{align}\implies \big\vert\det T\big\vert^{-1} \cdot \int_{\mathbb R^n} \phi_n(x) dx=\int_{\mathbb R^n} \phi_n\circ T(x) dx\end{align}$    
$\begin{align}\implies \big\vert\det T\big\vert^{-1} \cdot \int_{\mathbb R^n} f^+(x) dx=\int_{\mathbb R^n} f^+\circ T(x) dx\end{align}$    
where the first implication is due to linearity of the integral and the second implication involves taking a limit of the simple sum (i.e. Monotone Convergence)  

**reconciling chapter 2 vs chapter 4 Lebesgue Integrals**  

In chapter 2 we take what amounts to the Darboux approach to integration, starting with bounded measurable $f$ using sequences of upper and lower bounds based around measurable partitions.  The below reconciles the two definitions; in each case it suffices to consider $f^+$ and $f^-$ separately so we may **assume WLOG that** $f\geq 0$.  Further the chapter spent almost the entire time discussing integration over $[a,b]$ and only offers glimpses of unbounded sets and higher dimensional $\mathbb R^n$, so the below focuses on $[a,b]$.  

**(i.)** $f$ is chapter 2 Lebesgue Integrable over $[a,b]\implies f$ is chapter 4 Lebesgue Integrable   
We have the same definition of Lebesgue measurability in both chapters and $f$ is Lebesgue measurable in each case-- this is directly stated in Def 4.3.2 and something developed e.g. under "fix 2" under ex 2.6.16 in nelson_chp2.ipynb).  The truncation $^N f$ is Lebesgue measurable-- e.g. since the set $A = \big\{x \vert f(x) \leq N\big\}$ is measurable hence $^N f = \mathcal X_A \cdot f + (1-\mathcal X_A)\cdot N$ is measurable since measurable functions form a ring (Theorem 4.2.5 or Theorem 2.1.6).  Def 2.3.1 tells us $^N f$ is integrable as well and  
$\begin{align}\int_a^b f dm = \lim_{N\to \infty}\int_a^b  \mkern0mu ^N f  dm\end{align}$  

Now this problem reduces to showing $^N f$ is Lebesgue Integrable in the chapter 4 sense, since Monotone Covergence (Theorem 4.3.9) then gives us the above limit.  For some fixed $N$ we have $\int_a^b  (^N f)  dm$ is given by $\inf_{P\in\mathcal P} U\big[^N f, P\big]=\sup_{P\in\mathcal P} L\big[^N f, P\big]$ per Def 2.2.5.  Now for $P\in \mathcal P$ we have $L\big[ ^N f, P\big] = \int_{a}^b \phi^{(P)} dm$ where the RHS is Def 4.3.1  and Def 4.3.2-- i.e. the measurable partition from chapter 2 has finitely many elements, each of which we can assign to a characteristic function with a scalar coefficient given by the infimum taken by $f$ on that measurable partition (Def 2.2.4) and adding these up yields a simple function.  Thus taking the supremum,  

$\begin{align} \sup_{P\in\mathcal P} L\big[^N f, P\big]  = \sup\left \{\int \phi d m \big\vert \phi\text{ is simple with }0\leq \phi \leq ^N f\right\}= \int_X f dm =\int_a^b f dm
\end{align}$  
where $X:=[a,b]$ and the RHS is the chapter 4 definition of the integral (Def 4.3.2)  

**(ii.)** $f$ is chapter 4 Lebesgue Integrable over $[a,b]\implies f$ is chapter 2 Lebesgue Integrable   
again proceed by truncation and consider $^N f$; once we've established that is Lebesgue Integrable in the chapter 2 sense then Monotone Convergence Theorem (Theorem 2.4.13) gives the result.  Since $f$ Lebesgue measurable then $^N f$ is Lebesgue measurable and bounded on $[a,b]$ hence it is Lebesgue Integrable per Theorem 2.2.12.

**ex 4.6.5**  
Let $f:[a,b]\longrightarrow \mathbb R$ be a (Lebesgue) measurable function on $[a,b]$.  Prove that the inverse image of every Borel set is a (Lebesgue) measurable set.  

This is a surprisingly tricky problem as we cannot attempt an argument directly on generators without using transfinite induction which would be quite involved and is outside the scope of this text.  There is a comparable argument on page 109 of Jones that has $f$ continuous and argues that $A\text{ Borel}\implies f^{-1}(A)\text{ Borel}$ and that gives the structure for this argument  

$S:= \Big\{A\big\vert A\subseteq \mathbb R \text{ & } f^{-1}\big(A\big) \text{ is Lebesgue measurable}\Big\}$  
by construction $A \in S \implies f^{-1}\big(A\big) \text{ is Lebesgue measurable}$ so it suffices to prove that $S$ contains all Borel sets $A$ (in the codomain of $\mathbb R$) i.e. that $\mathcal B\subseteq S$ so then the argument is for arbitrary $A\in \mathcal B\subseteq S$ we have $f^{-1}\big(A\big) \text{ is Lebesgue measurable}$.  Thus it remains to show that $S$ is a $\sigma$-algebra and it contains all open sets.  

$\sigma$-algebra  
(i.) the empty set is trivially included  
(ii.) if $A_k \in S$ then $f^{-1}\big(\bigcup_{k=1}^\infty A_k\big) = \bigcup_{k=1}^\infty f^{-1}\big(A_k\big) $  where the RHS is Lebesgue measurable so the LHS is hence $\bigcup_{k=1}^\infty A_k\in S$  
(iii.) if $A\in S$ then $f^{-1}\big(A^C\big) =f^{-1}\big(\mathbb R - A \big)= \big[a,b\big]- f^{-1}\big(A \big)= \big[a,b\big]\cap \Big( f^{-1}\big(A \big)\Big)^C$ which is the intersection of two Lebesgue measurable sets hence Lebesgue measurable thus $A^C \in S$     

*open sets*  
if $A\subseteq \mathbb R$ is open then per ex 2.6.2(b) we know $f^{-1}\big(A\big)$ is Lebesgue measurable since $f$ is, hence $S$ includes all open sets in $\mathbb R$.  

**generalizing**  
def 2.35 in Axler's MIRA (p. 31 as of writing this) *defines* $f$ to be ($\mathcal S$) measurable on $\big(X, \mathcal S\big)$ if 
$f^{-1}\big(B\big)\in \mathcal S$ for every Borel Set $B \subseteq \mathbb R$.  Being measurable per Axler trivially implies a function is measurable under Nelson's definition since $\big\{x \vert f(x)\lt s\big\} = f^{-1}\big(-\infty , s\big)$ where $\big(-\infty , s\big)$ is open hence a Borel set since the Borel sets contain all open sets.  

Conversely we can generalize the above exercise slightly to show $\text{a function being measurable per Nelson's  definition} \implies \text{a function being measurable per Axler's  definition}$   
we do this by essentially removing any reference to Lebesgue and instead using "measurable" to mean $\mathcal S$ measurable. I.e.  
$S^*:= \Big\{A\big\vert A\subseteq \mathbb R \text{ & } f^{-1}\big(A\big)\in \mathcal S \text{ (is measurable) per Nelson}\Big\}$  
as before, it is a $\sigma$ algebra, noting that with respect to the above (iii.) a sigma algebra contains the whole space (the complement of the empty set) and $f^{-1}\big(A\big)$ is measurable [per Nelson criterion] by assumption and the sigma algebra is closed under complements (and countable intersections).  And again $S^*$ contains all open sets by running ex 2.6.2 essentially verbatim, **except** we are checking membership of being in the set of measurable sets, i.e. $\in\mathcal S$, instead of being a Lebesgue measurable set.  Since $S^*$ is a sigma algebra containing all open sets it contains the Borel sets so $\mathcal B^1 \subseteq S^*$ hence $f^{-1}\big(B\big)\in\mathcal S$ (i.e. is measurable according to Nelson) for $B\in\mathcal B^1 \subseteq  S^*$-- i.e. the pre-image of each Borel set is measurable when a function is Nelson-measurable.  

*remark:*  
if $f,g:\mathbb R\longrightarrow \mathbb R$ are Borel measurable then $f\circ g$ is measurable since for Borel set $B$ we have $\big(f\circ g\big)^{-1}\big(B\big)= g^{-1}\circ f^{-1}\big(B\big)=g^{-1}\Big( f^{-1}\big(B\big)\Big)=g^{-1}\Big(B'\Big)$  i.e. the RHS is the inverse image of Borel set $B'$ is a Borel set since $g$ is Borel measurable.  With some care this can be generalized to $f:\mathbb R^n\longrightarrow \mathbb R^m$ and $g:\mathbb R^m\longrightarrow \mathbb R$.  Also recall that the Borel $\sigma$-algebra is a sub ($\sigma$) algebra of the Lebesgue sets so if a function $f$ is Borel measurable then it is Lebesgue measurable.

# Project 1: Egorov's Theorem  
If $f_n\to f$ point-wise on $[a,b]$  then for any $\epsilon \gt 0$ there is a closed set $F\subseteq [a,b]$ such that $m\big([a,b] -F\big)\lt \epsilon$ and $f_n\to f$ uniformly on $F$.  

*proof:*   
Fix some $\epsilon\gt 0$ and for $r\in\mathbb N$ set $A_r:=\Big\{x \in [a,b]\Big\vert \big\vert f_n(x)-f(x)\big\vert\gt \frac{1}{r}\text{  for some }n\geq J^{(r)}\Big\}$ with $J^{(r)}$ chosen large enough such that $m\big(A_r\big) \lt \epsilon \cdot 2^{-r-2}$ [justification for this is deferred to the end].   Since $A_r$ is measurable, select open set $G_r\supseteq A_r$ with $m\big(G_r- A_r\big)\lt \epsilon \cdot 2^{-r-2}\implies m\big(G_r\big)= m\big(A_r\big) + m\big(G_r-A_r\big) \leq \epsilon\cdot 2^{-r-1}$  

$G:=\bigcup_{r=1}^\infty G_r\implies m\big(G\big)\leq \sum_{r=1}^\infty m\big(G_r\big)\leq \sum_{r=1}^\infty \epsilon\cdot 2^{-r-1}=\frac{\epsilon}{2}\lt \epsilon $  
$G$ is a countable union of open sets, hence open.  It's complement in $[a,b]$ given by $F:=G^C = \big(\mathbb R- G\big)\cap \big[a,b\big]$ is a closed subset of $\big[a,b\big]$ as required.   

We deduce that when restricted to $F$ we have $f_n^{(F)}\to f^{(F)}$ *uniformly*.  If this wasn't true then there would be some $\epsilon'\gt 0$ such that $\big\vert f_{n_k}(x_{n_k})-f(x_{n_k})\big\vert \geq \epsilon '$ for *infinitely many* $n_k\in \mathbb N$ where each $x_{n_k}\in F$.  But $\epsilon' \gt \frac{1}{m}$ for large enough $m\in \mathbb N$  and since there are infinitely many, we may select some $n_k^* \geq J^{(m)}$
$\implies x_{n_k^*}\in \Big\{x \in [a,b]\Big\vert \big\vert f_n(x)-f(x)\big\vert\gt \frac{1}{m}\text{  for some }n\geq J^{(m)}\Big\}=A_m \subseteq G_m\subseteq G = F^C\implies x_{n_k^*}\in F \cap F^C =\big\{\big\}$ and that is a contradiction. 

*justification of* $m\big(A_r\big) \lt \epsilon \cdot 2^{-r-2}$  
$S_j^{(r)}:=\Big\{x \in [a,b]\Big\vert \big\vert f_n(x)-f(x)\big\vert\leq \frac{1}{r}\text{  all }n\geq j\Big\}=\bigcap_{n=j}^\infty \Big\{x \in [a,b]\Big\vert \big\vert f_n(x)-f(x)\big\vert\leq \frac{1}{r}\Big\}$  
a countable intersection of measurable sets, so $S_j^{(r)}$ is measurable  

This forms a filtration $S_j^{(r)}\subseteq S_{j+1}^{(r)}\subseteq S_{j+2}^{(r)}\subseteq \dots$  
so $m\big(S_j\big)\leq m\big(S_{j+1}\big)$ a monotone increasing sequence and since $f_n$ converges in measure to $f$ [Project 2] we know   
$\lim_{j\to \infty} m\big(S_j^{(r)}\big) = m\big([a,b]\big)=b-a$  
Thus $(b-a)- \epsilon \cdot 2^{-r-2} \leq m\big(S_j^{(r)}\big)\leq (b-a)$ for $j\geq J^{(r)}$  
Then the complement  
$A_r:= [a,b] - S_j^{(r)}= \Big\{x \in [a,b]\Big\vert \big\vert f_n(x)-f(x)\big\vert\gt \frac{1}{r}\text{  for some }n\geq J^{(r)}\Big\}$
where $m\big(A_r\big)\leq \epsilon \cdot 2^{-r-2}$

**generalizing and** $\sigma$**-finiteness**

$$\int_{I^{(r\times s)}} f dm=  \int_{I_1^{(r)}}d\mathbf x\left[ \int_{I_2^{(s)}} f\big(\mathbf x,\mathbf y\big)d\mathbf y\right]=\int_{I_1^{(r)}} g_s\big(\mathbf x\big) d\mathbf x$$

$$\int_{I^{(\mathbb R^n \times s)}} f dm= \lim_{r\to \infty}\int_{\mathbb R^{n+m}}\mathcal X_{I^{(r\times s)}}\cdot  f dm = \lim_{r\to \infty}\int_{I^{(r\times s)}} f dm=   \lim_{r\to \infty}\int_{I_1^{(r)}} g_s\big(\mathbf x\big) d\mathbf x=\lim_{r\to \infty}\int_{\mathbb R^n} \Big(\mathcal X_{I_1^{(r)}}\cdot g_s\big(\mathbf x\big)\Big) d\mathbf x=\int_{\mathbb R^n} g_s\big(\mathbf x\big) d\mathbf x$$  
and applying Monotone Convergence again  
$$\int_{I^{(\mathbb R^n \times \mathbb R^m)}} f dm= \lim_{s\to \infty}\int_{I^{(\mathbb R^n \times s)}} f dm= \lim_{s\to \infty}\int_{\mathbb R^n} g_s\big(\mathbf x\big) d\mathbf x=\int_{\mathbb R^n}d\mathbf x\left[ \int_{\mathbb R^m} f\big(\mathbf x,\mathbf y\big)d\mathbf y\right]$$

$\underline{\text{alternative definition for the outer measure}}$  

suppose our definition of of the outer measure, denoted $\lambda_{X\times Y}^{**}$ below, did not require the covering by measurable rectangles to be disjoint, i.e.  
$\lambda_{X\times Y}^{**}\big(E\big)=\inf_{E\subseteq \bigcup_{k=1}^\infty S_k\times T_k\vert  (S_k \times T_k)\text{ measurable rectangles}}\sum_{k=1}^\infty \mu(S_k)\cdot \mathcal v(T_k)=\inf_{E\subseteq \bigcup_{k=1}^\infty S_k\times T_k\vert  (S_k \times T_k)\text{ measurable rectangles}}\sum_{k=1}^\infty \lambda^*\big(S_k\times T_k\big)$  

then we could use the opening lemma to re-write $\bigcup_{k=1}^\infty S_k\times T_k=\bigcup_{k=1}^\infty R_k$   where the RHS is a disjoint union of measurable rectangles.  Hence  
$\sum_{k=1}^\infty \lambda^*\Big(R_k\Big) =\lambda^*\Big(\bigcup_{k=1}^\infty R_k\Big)= \lambda^*\Big(\bigcup_{k=1}^\infty S_k\times T_k\Big)\leq \sum_{k=1}^\infty\lambda^*\Big(S_k\times T_k\Big)$  
where the LHS holds because the prior work verifies that $\lambda^*$ is a measure hence is countably additive on disjoint measurable sets [and measurable rectangles are measurable per (d.)] and the RHS is countable sub-additivity for an outer measure.  Thus if we wrote $\lambda_{X\times Y}^{**}\big(E\big)$, an infimimum, in terms of a sequence of measurable rectangle coverings with monotone decreasing total measure $s_m$, then that implies a sequence of disjoint measurable rectangle coverings with total measure $d_m\leq s_m$ and hence $\lim_{m\to \infty}d_m\leq \lim_{m\to \infty}s_m= \lambda_{X\times Y}^{**}\big(E\big)\implies \lim_{m\to \infty}d_m= \lim_{m\to \infty}s_m$ by definition of the infimum, i.e. we would necessarily get the exact same outer measure. **Conclude: we may just as well define the outer measure to be based on the infimum of covering by (not necessarily disjoint) measurable rectangle coverings.**   

Also note ex 5.A.10 in Axler's MIRA would tell us that in a $\sigma$-finite space, any of two (product) outer measure definitions would result in the same measures so long as they agree on the measurable rectangles.
