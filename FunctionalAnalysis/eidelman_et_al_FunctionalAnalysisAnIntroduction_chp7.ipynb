*preamble*  
There are some technical issues with this problem as $\varphi:[a,M]\rightarrow \mathbb C$ may be complex valued hence $\varphi(A)$ won't be self-adjoint in general [this leads to errata in Prop 7.2.2 where it is stated $\Vert\varphi(A)x\Vert=\int \varphi^2(\lambda)d\langle E_{\lambda} x,x\rangle$ which can take negative [or even non-real] values and apparently $\int\vert \varphi(\lambda)\vert^2 d\langle E_{\lambda} x,x\rangle$  was meant]. But we have not learned anything about normal operators so the following argument allows us to reduce this to the self-adjoint case as follows [conceptually this is similar to how I addressed ex 6.5.9 by reducing to the cases of positive operators]:  

using (7.2.5) for any $\epsilon \gt 0$ we can find as small enough partition such that  
$\big\Vert \varphi(A) - \sum \varphi(\mu_i)\Delta E_{\lambda_i} \big\Vert \leq 3 \epsilon$    
$\implies \big\Vert \varphi(A)^* - \sum \overline{\varphi(\mu_i)}\Delta E_{\lambda_i} \big\Vert= \big\Vert \big(\varphi(A) - \sum \varphi(\mu_i)\Delta E_{\lambda_i}\big)^* \big\Vert=\big\Vert \varphi(A) - \sum \varphi(\mu_i)\Delta E_{\lambda_i} \big\Vert  \leq 3 \epsilon$  
noting that $\Delta E_{\lambda_i}$ are othogonal projections hence self-adjoint and the 2nd equality is problem 7.21 in N. Young   
$\implies \varphi(A)^*= \lim_{\vert \Delta\vert \to 0}\sum \overline{\varphi(\mu_i)}\Delta E_{\lambda_i}  = \overline{\varphi}(A)$ where $\overline\varphi(t):= \overline{\varphi(t)}$  
$\implies \varphi(A)^*\varphi(A)= \varphi(A)\varphi(A)^*$  
since for any partition  $\Delta$  
$\sum \overline{\varphi(\mu_i)}\Delta E_{\lambda_i} \sum \varphi(\mu_i)\Delta E_{\lambda_i}= \sum \vert\varphi(\mu_i)\vert^2 \Delta E_{\lambda_i} = \sum \varphi(\mu_i)\Delta E_{\lambda_i} \sum \overline{\varphi(\mu_i)}\Delta E_{\lambda_i} $   

which gives the result.

For further overkill explanation, consider  

$\Big\Vert\varphi(A)^*\varphi(A)-\sum \vert\varphi(\mu_i)\vert^2 \Delta E_{\lambda_i}\Big\Vert $   
$=\Big\Vert\varphi(A)^*\varphi(A)-\sum \overline{\varphi(\mu_i)}\Delta E_{\lambda_i} \sum \varphi(\mu_i)\Delta E_{\lambda_i}\Big\Vert $   
$=\Big\Vert\varphi(A)^*\varphi(A)-\varphi(A)^*\sum \varphi(\mu_i)\Delta E_{\lambda_i}  + \big(\varphi(A)^*\sum \varphi(\mu_i)\Delta E_{\lambda_i}-  \sum \overline{\varphi(\mu_i)}\Delta E_{\lambda_i} \sum \varphi(\mu_i)\Delta E_{\lambda_i}\big)\Big\Vert $   
$\leq \Big\Vert\varphi(A)^*\varphi(A)-\varphi(A)^*\sum \varphi(\mu_i)\Delta E_{\lambda_i} \Big\Vert  + \Big\Vert\varphi(A)^*\sum \varphi(\mu_i)\Delta E_{\lambda_i}-  \sum \overline{\varphi(\mu_i)}\Delta E_{\lambda_i} \sum \varphi(\mu_i)\Delta E_{\lambda_i}\Big\Vert $

$= \Big\Vert\varphi(A)\Big\Vert \cdot \Big\Vert\varphi(A)-\sum \varphi(\mu_i)\Delta E_{\lambda_i} \Big\Vert  + \Big\Vert\varphi(A)-\sum \varphi(\mu_i)\Delta E_{\lambda_i} \Big\Vert\cdot\Big\Vert \sum \varphi(\mu_i)\Delta E_{\lambda_i}\Big\Vert $

$\leq \Big(2\big\Vert\varphi(A)\big\Vert +3 \epsilon\Big)\cdot \Big(3 \epsilon\Big)  \quad\text{ (7.2.5))}$
