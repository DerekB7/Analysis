**2.**  
Find the spectrum of the operator $A$ in $L_2[-1,1]$ which is given by $(Af)(t)=a(t)\cdot f(t)$ where   
$a(t)=t$ for $t\in[0,1]$ and $a(t)=0$ for $t\in [-1,0]$  
(errata: the text said $a(t)=0$ for $t\in [-1,1)$ which is a contradiction)  


If $\lambda\not\in [0,1]$ then $\lambda$ is regular since   
$\big(A-\lambda I\big):f\mapsto (a-\lambda)\cdot f $ which has an inverse given by  
$g\mapsto (a-\lambda)^{-1}\cdot g $ since $a(t)-\lambda$ is continuous and non-zero everywhere hence multiplying by $(a-\lambda)^{-1}$ is a continuous mapping -- and it is bounded on $[-1,1]$ for reasons of compactness so $(a-\lambda)^{-1}\cdot g\in L_2[-1,1]$ per Cauchy-Schwarz.  

$\sigma(A)=[0,1]$    
Consider the constant map $f(t)=1$ and $\lambda \in [0,1]$    
$\implies \big(A-\lambda I\big):f\mapsto (a-\lambda)\cdot f $
which is zero at $t=\lambda$ so if $\big(A-\lambda I\big)^{-1}$ existed then it would have the mapping  
$g:\mapsto (a-\lambda)^{-1}\cdot g $ almost everywhere and while we can duck the definition at $t=\lambda$ specifically and e.g. assign it to be $1$ the issue is $\big(A-\lambda I\big)^{-1}$ cannot be defined anywhere in a neighborhood containing $\lambda$ since that set would have positive measure and the inverse mapping applies to constant $f$, above would have to be equal to $f\mapsto (a-\lambda)^{-1}\cdot f=(a-\lambda)^{-1} $ almost everywhere but that is not square integrable -- essentially for the same reason that $\int_0^1\frac{1}{x^2}=\infty$ and in some ways echoes some of the examples in ex 2.4.27

**3.**  
$A\in L(X)$ be an invertible operator. Prove that $\sigma\left(A^{-1}\right)=\big\{\lambda^{-1}: \lambda \in \sigma(A)\big\}$  

For invertible [bounded] operator $T$ we argue, for $\lambda \neq 0$  
$\lambda \text{ is a regular point of }T\implies \lambda^{-1} \text{ is a regular point of }T^{-1}$  
$\big(T-\lambda I\big)=T\big(I-\lambda T^{-1}\big)=-\lambda \cdot T\big(-\lambda^{-1}I + T^{-1}\big)=-\lambda \cdot T\big( T^{-1}-\lambda^{-1}I\big)$  
where the RHS is invertible since the LHS is, and in particular $\big( T^{-1}-\lambda^{-1}I\big)$ is invertible so $\lambda^{-1}$ is regular for $T^{-1}$.  Of course trivially $0$ is regular for $T$ and $T^{-1}$ since it is invertible.  

Letting $T:=A$ then the above yields for $\lambda\neq 0$   
$\lambda \text{ is a regular point of }A\implies \lambda^{-1} \text{ is a regular point of }A^{-1}$  

and letting $T:=A^{-1}$ then the above yields for $\lambda\neq 0$    
$\lambda \text{ is a regular point of }A^{-1}\implies \lambda^{-1} \text{ is a regular point of }A\iff \lambda^{-1} \text{ is a regular point of }A^{-1}\implies \lambda \text{ is a regular point of }A$  
where we use the automorphism of $z\mapsto z^{-1}$ on the punctured plane $\mathbb C\sim\big\{0\big\}$ to get the "$\iff$" equivalence    

conclude for $\lambda\neq 0$  
$\lambda\text{ is a regular point of }A\iff \lambda^{-1} \text{ is a regular point of }A^{-1}$  
and again trivially $A$ and $A^{-1}$ have zero as a regular point.  

Looking at set complements then, $\lambda \in \sigma\big(A\big)\iff \lambda^{-1}\in \sigma\big(A^{-1}\big)$  


**4.**  
$A\in L(X)$ and $\lambda \in \mathbb C$ an assume that there exists a sequence $(x_n)_{n=1}^\infty \subseteq X$ such that $\Vert x_n\Vert =1$ and $Ax_n-\lambda x_n\to 0$ as $n\to \infty$.  Prove  that $\lambda \in \sigma(A)$.  

*suppose for contradiction* that $\lambda$ is a regular point, i.e. that $\big(A-\lambda\big)^{-1}$ exists  
[recalling we work solely on Banach spaces so per Theorem 4.7.1, Banach Open Map Theorem, a bounded linear map is bijective *iff* it has a bounded inverse]  

then with $\big(A-\lambda I\big)x_n:=y_n$ we have for all $n$  
$\big\Vert \big(A-\lambda I\big)^{-1}y_n\big\Vert  = \big\Vert \big(A- I \lambda\big)^{-1}\big(A-\lambda I\big)x_n\big\Vert = \big\Vert x_n\big\Vert =1$  
but for any $\epsilon \gt 0$ we know $\Vert y_n\Vert \lt \epsilon$ for all $n\geq N$ (large enough)  

hence for $n\geq N$  
$\implies \big\Vert \big(A-\lambda I\big)^{-1}\frac{y_n}{\Vert y_n\Vert}\big\Vert = \frac{1}{\Vert y_n\Vert}\gt \frac{1}{\epsilon}\implies \big\Vert \big(A-\lambda I \big)^{-1}\big\Vert \gt \frac{1}{\epsilon}$ 
$\implies \big(A-\lambda I\big)^{-1}$ is unbounded, since $\epsilon \gt 0$ was arbitrary, a contradiction  


**extension:**  
we can slightly modify the above to make a (partial) converse to ex 4.8.24 [which roughly mimics exercise 3.27 in MacCluer]  

suppose $A\in B(H_1, H_2)$ and the image of $A$ is closed. Then for $v \in (\ker A)^\perp$ we have  
$\big\Vert Av \big\Vert\geq m\cdot \big\Vert v\big\Vert$ for some $m\gt 0$.  


First notice that $\text{image }A=\overline{\text{image }A}\subseteq H_2$ implies that $\text{image }A$ is a complete subspace of $H_2$ hence it is a Banach space in its own right.  Similarly $(\ker A)^\perp\subseteq H_1$ is always a closed subspace hence Banach.  
This means we can define a map $T:(\ker A)^\perp\rightarrow \text{image }A$ given by $v\mapsto Av$ which is by construction injective and surjective so a formal linear inverse $T^{-1}$ exists. And $T^{-1}\in B(H_2, H_1)$ by Banach Open Mapping theorem. Thus setting $m:=\frac{1}{\Vert T^{-1} \Vert}$ proves the claim.  Put differently, mimicking the above argument, if the claim is not true we have some sequence $v_k$ with $\Vert v_k\Vert =1$ such that $Tv_k\to 0\implies \Vert v_k\Vert = \Vert T^{-1}Tv_k\Vert= \Vert T^{-1}y_k\Vert =1$  which implies $T^{-1}$ is unbounded, contradicting the Banach Open Mapping theorem.  


**5.**  
Let $\mathbb C_{\mathbb R}$ be the space of all continuous and bounded functions $x(t)$ on $\mathbb R$ with norm $\Vert x\Vert =\sup_{t\in \mathbb R}\vert x(t)\vert$.  On the space $C_{\mathbb R}$ we define the operator $A$ by $(Ax)(t) = x(t+s)$ where $s\in \mathbb R$ is a constant.  Prove $\sigma(A) =\big\{\lambda \in \mathbb C:\vert \lambda \vert =1\big\}$  

*remark:* evidently $x:\mathbb R\longrightarrow \mathbb C$  

Note by inspection $\big\Vert A x\big\Vert=\big\Vert  x\big\Vert$ so $A$ is unitary where $\Vert A \Vert =1$ and that $A^{-1}$ exists and is given by $(Ax)(t) = x(t-s)$  which is unitary with norm 1 as well.  

$\sigma(A) \subseteq\big\{\lambda \in \mathbb C:\vert \lambda \vert =1\big\}$   
per Prop 5.0.1 if $\vert \lambda \vert \gt 1 =\Vert A\Vert$ then $\lambda \not\in\sigma(A)$, i.e. a regular point.  
and for $A^{-1}$ then $\vert \lambda\vert \lt 1\implies \vert \lambda\vert^{-1}\gt \left \Vert A^{-1}\right\Vert =1 \implies \lambda^{-1} \not\in \sigma\left(A^{-1}\right)$ per Prop 5.0.1 $\implies \lambda \not\in\sigma(A)$ per exercise 3.  

Now consider arbitrary $\lambda$ where $\vert \lambda\vert =1$, i.e. $\lambda =\exp\big(2\pi\cdot i\cdot \theta\big)$ for some $\theta \in \big(0,2\pi\big]$.  We argue $\big(A-\lambda I\big)$ has a non-trivial kernel [i.e. $\lambda$ is point spectrum]. Being linear $\big(A-\lambda I\big)\mathbf 0=\mathbf 0$, but consider $x_0(t):= \exp\big(2\pi\cdot i\cdot \frac{t}{r}\big)$ where $r:= \frac{s}{\theta}$.  This map sends the real line to the unit circle.     
$(Ax_0)(t)= \exp\big(2\pi\cdot i\cdot \frac{t+s}{r}\big)= \exp\big(2\pi\cdot i\cdot \frac{t+s}{r}\big)=\exp\big(2\pi\cdot i\cdot \frac{t}{r}\big)\cdot \exp\big(2\pi\cdot i\cdot \frac{s}{r}\big)=x_0(t)\cdot \exp\big(2\pi\cdot i\cdot \frac{s}{r}\big)=\lambda \cdot x_0(t) $  
$\implies \big(A-\lambda I\big)x_0= \lambda \cdot x_0 -\lambda \cdot x_0=0$  

**8.**  
Let $A$ be an invertible operator and let $K$ be a compact operator in a Banach space, $X$.  Prove that 

(a.) $\dim \ker\big(A+K\big)\lt \infty$  
re-write this as  
$\dim \ker\big(A+K\big)=\dim \ker\big(A(I+A^{-1}K)\big)=\dim \ker\big(I+A^{-1}K\big)\lt \infty$  
Working from right to left:  
where the RHS holds by Prop 5.2.4, with $\lambda :=-1$ and  the 2nd to right equality holds because invertible linear maps preserve finite dimensions.  Also recall $A^{-1}K$, as well as  $KA^{-1}$, is compact per Prop 4.3.4  

(b.) $\text{codim } \text{Image}\big(A+K\big)\lt \infty$  
re-write this as  
$\text{codim } \text{Image}\big(A+K\big)=\text{codim } \text{Image}\big(A(I+A^{-1}K)\big)=\text{codim } \text{Image}\big(A(A^{-1}K+I)\big)\lt\infty$   
*remark: the below argument actually yields an independent proof of exercise 5.3.13 Fredholm's third theorem*  

We prove this as follows.  Let $T$ be compact with $\dim \ker \big(T + I\big)=m\lt \infty$ [here $\lambda:=-1$ with Prop 5.2.4 applied].  Letting $L:=\text{Image }\big(T + I\big)$ we have $L$ is closed per Prop 5.2.5 

Base Case: $m=1$    
Since $\dim \ker \big(T + I\big)=1$ we have $\ker \big(T+I\big)=\text{span}\big\{w\big\}$ and  some $y \not\in L$ per Prop 5.2.7.  Per [Hahn-Banach] Corollary 3.1.7 there is an $f \in X^*$ such with norm $1$ such that $f(w) \gt 0$.  

$B:= \big(T+I\big) +f(x)\cdot y$ noting that $Bw= f(w)\cdot y\neq 0\implies y \in \text{image }\big(B\big)$  
$v \in \ker B\implies -\big(T+I\big)v = f(v)\cdot y$  
$\implies -\big(T+I\big)v \in L \cap \text{span}\big\{y\big\}\implies -\big(T+I\big)v = 0= f(v)\cdot y$  $\implies v \in \ker \big(T+I\big)\implies v =\alpha \cdot w\implies 0=f(v)\cdot y=0=f(\alpha\cdot w)\cdot y=\alpha\cdot f( w)\cdot y\implies \alpha =0 \implies v=0$  
$\implies \dim \ker B =0\implies \text{image }B=X \quad\text{(prop 5.2.10)}$  

$\implies X = \text{image }\big(T+I\big)\oplus\text{span}\big\{y\big\}$  
$\implies X \big/\big(T+I\big)\cong \text{span}\big\{y\big\}$  
i.e. this sets the stage for an induction that $\dim \big/\big(T+I\big) = \dim \ker \big(T+I\big)$  

Inductive Case: $m\geq 2$  
The argument runs almost the same as in the Base Case, again reducing the the dimension of the kernel by one and finishing with induction hypothesis.  
Since $\dim \ker \big(T + I\big)=m$ we have $\ker \big(T+I\big)=\text{span}\big\{w_1, \dotso, w_{m-1}, w_m\big\}$ and some $y_m \not\in L$ per Prop 5.2.7.  Per [Hahn-Banach] Corollary 3.1.7 there is an $f \in X^*$ such with norm $1$ such that $f(w_m) \gt 0$ and $f\big(\text{span}\big\{w_1, \dotso, w_{m-1}\big\}\big)=\big\{0\big\}$ since $\text{span}\big\{w_1, \dotso, w_{m-1}\big\}$ is closed [ex 3.3.3].  

$B:= T+I +f(x)\cdot y_m$  [where $T+  f(x)\cdot y_m$ is the sum of two compact operators so compact per Prop 4.3.4]     
$v \in \ker B\implies -\big(T+I\big)v = f(v)\cdot y_m$  
$\implies -\big(T+I\big)v \in L \cap \text{span}\big\{y_m\big\}\implies -\big(T+I\big)v = 0= f(v)\cdot y_m$  $\implies v \in \ker \big(T+I\big)\implies v =\sum_{k=1}^m \lambda_k \cdot w_k\implies 0=f\big(\sum_{k=1}^m \lambda_k \cdot w_k\big)\cdot y_m = \sum_{k=1}^m \lambda_k \cdot f\big( w_k\big)\cdot y_m = \lambda_m \cdot f\big(w_m\big)\cdot y_m $  
$\implies \lambda_m =0 \implies v \in \text{span}\big\{w_1, \dotso, w_{m-1}\big\}$  
$\implies \dim \ker B \leq m-1 \implies \dim \ker B = m-1 \quad\text{(check by setting }v:=w_j\text{ for }1\leq j\leq m-1\text{)}$  

and by calling on induction hypothesis we get  
$\implies X = \text{image }\big(B\big)\oplus \text{span}\big\{y_1, \dotso, y_{m-1}\big\}=\text{image }\big(T+I\big)\oplus \text{span}\big\{y_m\big\}\oplus \text{span}\big\{y_1, \dotso, y_{m-1}\big\}$  
where the 2nd equality comes from checking  $B(w_m) =f(w_m)\cdot y_m\neq 0$ and for any $v \in \text{image}(T+I)$ we have $(T+I)x =v$ for some $x$ which implies $B(x+\alpha \cdot w_m)=v$ for suitable choice of $\alpha$ so $\text{image }B \supseteq \text{image}(T+I)\oplus \text{span}\big\{y_m\big\}$  . Conversely if $v\in \text{image }B$ then $Bx=v =(T+I)x +f(x)\cdot y_m\implies \text{image }B \subseteq \text{image}(T+I)\oplus \text{span}\big\{y_m\big\}$   
$\implies X =\text{image }\big(T+I\big)\oplus \text{span}\big\{y_1, \dotso, y_{m-1},y_m\big\}$  
$\implies X \big/\text{image}\big(T+I\big)\cong \text{span}\big\{y_1, \dotso, y_{m-1},y_m\big\}$  
$\implies X \big/\text{image}\big(T+I\big) = \dim \text{span}\big\{y_1, \dotso, y_{m-1},y_m\big\} = m =\dim \ker \big(T+I\big)$  

Finally, let $A^{-1}K:= T$  
$\implies X =\text{image }\big(A^{-1}K+I\big)\oplus \text{span}\big\{y_1, \dotso, y_{m-1},y_m\big\}$  
$\implies AX =X =\text{image }\Big(A\big(A^{-1}K+I\big)\Big)\oplus \text{span}\big\{Ay_1, \dotso, Ay_{m-1},Ay_m\big\}=\text{image }\Big(A\big(A^{-1}K+I\big)\Big)\oplus \text{span}\big\{z_1, \dotso, z_{m-1},z_m\big\}$  
$\implies X \big/\text{image }\Big(A\big(A^{-1}K+I\big)\Big)\cong \text{span}\big\{z_1, \dotso, z_{m-1},z_m\big\}$  
which has dimension $m\lt \infty$ as required by the problem.  

*remark:*  
The technique here is very much that same as what we'd do in linear algebra.  E.g. the approach for $A \in \mathbb F^{n\times n}$: if $0\neq v \in \ker A$, we can reduce the kernel size of $A$ by one [and of course increase its rank by one] as follows: construct invertible matrix $V$ with its first column being $v$ and any other (chosen) basis vectors for $\ker A$ being columns $v_k$ for $2\leq k\leq \dim \ker A=r$ of $V$ then append standard basis vectors as relevant for other columns to get an invertible matrix.  Let $W:=V^{-1}=\begin{bmatrix}\mathbf w_1^T   \\\vdots \\ \mathbf w_n^T\end{bmatrix}$ and $\implies \dim \ker \big(A+\mathbf v\mathbf w_1^T\big)=1+\dim \ker A $  

which we can check since  
$\big(A+\mathbf v\mathbf w_1^T\big)V = AV + \mathbf v\mathbf w_1^TV = V \begin{bmatrix} \mathbf 0 &*\\ \mathbf 0 & A_{n-r}\end{bmatrix}+V \begin{bmatrix} D_r &\mathbf 0\\ \mathbf 0 & \mathbf 0\end{bmatrix}= V \begin{bmatrix} D_r &*\\ \mathbf 0 & A_{n-r}\end{bmatrix}$  
where $D_r$ is diagonal with the top left entry $=1$ and all else equal to zero and $\det\big(A_{n-r}\big)\neq 0$.  By sub-additivity of rank we have $\text{rank}\big(A+\mathbf v\mathbf w_1^T\big)\leq \text{rank}A +1$ and this is met with equality since the bottom $n-r$ rows are linearly independent (since $A_{n-r}$ is invertible) and the top row is linearly independent of those by inspection [alternatively consider the $n-r+1\times n-r+1$ principal submatrix containing $A_{n-r}$ and the top left entry of $D_r$ which has determinant $=1\cdot \det\big(A_{n-r}\big)\neq 0$]  

**deriving Fredholm's Third Theorem**  
the above argument implies Fredholm's Third Theorem as follows.  First note that for compact $T'$ and $\lambda \neq 0$ we may re-write $\text{image}\big(T'-\lambda I\big)=\text{image}\big(-\lambda \cdot\big(\frac{T'}{-\lambda}+I \big)\Big)=\text{image}\big(\frac{T'}{-\lambda}+I \big)=\text{image}\big(T+I \big)$ 

so it suffices to consider $\big(T+I\big)$ as was done above [i.e. WLOG $\lambda =-1$].  Now referring to page 81, consider solutions to  
$Tx - \lambda \cdot x = v = (T+I)x $  
$\implies v \in \text{image}\big(T+I\big)\iff f_i(v)=0\text{ for all }i\in\big\{1,\dotso, k\big\}$ 
where $\big\{f_1,\dotso, f_k\big\}$ forms a basis for $\ker (T^*)_\overline{\lambda}=\ker T^*_{-1}$ [i.e. annihilators of the image of $T+I$ in line with Prop 5.2.8(i)]   

Applying ex 3.3.9 we can derive [necessarily linearly independent] $\big\{x_1, \dotso, x_k\big\}\in X$ such that  
$f_i(x_j)= \delta_{i,j}$  
$\implies \text{span}\big\{x_1, \dotso, x_k\big\}\cap \text{image }\big(T+I\big)=\big\{0\big\}$ by checking the subspaces under the image of  $f\in\big\{f_1,\dotso, f_k\big\}$  
$\implies \text{image }\big(T+I\big)\oplus \text{span}\big\{x_1, \dotso, x_k\big\}\subseteq X =\text{image }\big(T+I\big)\oplus \text{span}\big\{y_1, \dotso, y_{m-1},y_m\big\}$   
$\implies k\leq m$  
by quotienting each side by $\text{image }\big(T+I\big)$.  


Re-using the Gram type matrix in ex 3.3.9 we get  
$G=\begin{bmatrix}
    f_1(y_1) & \dots& f_{k-1}(y_1) &f_{k}(y_1) \\
    f_1(y_2) &   \dots & f_{k-1}(y_2)& f_k(y_2)\\
    \vdots & \ddots & \ddots & \vdots\\
    f_1(y_{m-1}) & \dots & f_{k-1}(y_{m-1})& f_k(y_{m-1})\\
    f_1(y_{m}) & \dots & f_{k-1}(y_{m}) & f_k(y_{m}) \\
    \end{bmatrix}$  
where $G\in \mathbb C^{m\times k}$. Now consider the left nullspace, i.e. any $\mathbf a$ such that 
$\mathbf a^TG =\mathbf 0^T\implies f_i\big(v\big)=0 \text{ for }1\leq i\leq m$  with $v:= \sum_{j=1}^m a_j \cdot y_j$  
$\implies v \in \text{image }\big(T+I\big)$  by commentary at bottom of page 81 [i.e. since the $f_i$ don't kill it, $(T+I)x=v$ is valid]   
$\implies v \in \text{image }\big(T+I\big)\cap \text{span}\big\{y_1, \dotso, y_{m-1},y_m\big\}\implies v =0$  
since $X=\text{image }\big(T+I\big)\oplus \text{span}\big\{y_1, \dotso, y_{m-1},y_m\big\}$  
$\implies \mathbf a =\mathbf 0\text{ since }0=v= \sum_{j=1}^m a_j \cdot y_j$ and  the $y_j$ are linearly independent  
$\implies \dim \ker G^T=0\implies m\leq k\quad \text{(e.g. by rank-nullity for  }G^T\text{)}$  

conclude: $\dim \ker \big(T+I\big)= m =k =\ker \big(T+I\big)^* $

*remark:*  
a different way to argue $m=k$ would be to observe  
$v \not\in \text{image}\big(T+I\big)\iff  [v]\neq [0] \in X\big/\text{image}\big(T+I\big) \iff f_i(v)\neq 0\text{ for some }i\in\big\{1,\dotso, k\big\}$ so define $E_i:= \ker f_i$   
Now consider $0\neq v \in \text{span}\big\{y_1,\dotso, y_m\big\}\implies [0]\neq [v]\in X\big / \big(\bigcap_{i=1}^k E_i\big)$ i.e. the quotient maintains linear independence of $y_m$ but we know   
$\dim X\big / \big(\bigcap_{i=1}^k E_i\big)\leq k$ per ex 1.6.3 $\implies m\leq k$. Using $\big\{x_1, \dotso, x_k\big\}$ one should be able to argue that in fact $\dim X\big / \big(\bigcap_{i=1}^k E_i\big)= k$  and then that  $X\big / \big(\bigcap_{i=1}^k E_i\big) \cong X\big/\text{image}\big(T+I\big)$ yielding $m=k$.  


**11.**  
Let $S_R$ and $S_L$ be the right and left shift operators respectively on $\mathscr l_2$  i.e.  
$S_R\left( \begin{bmatrix}x_1 \\ x_2  \\ x_3\\\vdots \end{bmatrix}\right)=\begin{bmatrix}0\\x_1 \\ x_2  \\\vdots \end{bmatrix}$

$S_L\left(\begin{bmatrix}x_1 \\ x_2  \\ x_3\\\vdots \end{bmatrix}\right)=\begin{bmatrix}x_2  \\ x_3\\ x_4\\\vdots \end{bmatrix}$  

Find the spectrum of these operators  

Note that trivially $0\in\sigma(S_R)$ and $0\in\sigma(S_L)$ since $\mathbf e_1 \not\in \text{image }S_R$ and $\mathbf e_1 \in \ker S_L$.  $S_R$ is unitary, but not surjective, so $\Vert S_R\Vert =1$
 and also $\Vert S_L\Vert =1$ though it is not unitary [check e.g. that it is contracting but preserves the norm of $\mathbf e_2$] hence per Prop 5.0.1 any $\vert \lambda \vert \gt 1$ is a *regular point*.  Now consider any $\lambda \in B^*\big(0,1\big)\cup \partial B\big(0,1\big)=\overline B\big(0,1\big)\sim\big\{0\big\}$ i.e. the punctured closed unit ball.  
 



$C_\lambda := S_R -\lambda I = \lambda \cdot 
\left[\begin{matrix}1 & 0 & 0 & 0 & 0&   \dots \\
-\lambda^{-1} & 1 & 0 & 0 & 0 &  \dots\\ 
0 & -\lambda^{-1} & 1 & 0 & 0 &  \dots\\
0 & 0 & -\lambda^{-1} & 1 & 0 &  \dots\\
\vdots & \vdots & \vdots & \ddots & \ddots & \ddots\\
\end{matrix}\right]$  

suppose $\mathbf e_1 \in \text{image }C_\lambda\iff \mathbf e_1=C_\lambda x$ for some $x$   
$\implies x =\begin{bmatrix}\lambda^{-1} \\ \lambda^{-2} \\ \lambda^{-3}\\ \lambda^{-4}\\\vdots \end{bmatrix}$  
$\implies \big\Vert x \big \Vert=\infty$ which is a contradiction since $x \in \mathscr l_2$  
conclude $C_\lambda$ is not surjective (hence not invertible) so $\lambda \in \sigma(S_R)$  

For the Left shift operator  
$B_\lambda := S_L -\lambda I = \lambda \cdot 
\left[\begin{matrix}1 & -\lambda^{-1} & 0 & 0 & 0&   \dots \\
0 & 1 &- \lambda^{-1} & 0 & 0 &  \dots\\
