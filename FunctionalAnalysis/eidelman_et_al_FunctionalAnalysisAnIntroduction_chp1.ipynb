**3.**  
Prove that if $\dim (E/E_i)=1$ for $1\leq i \leq n$ then $\dim E\big / \big(\cap_{i=1}^n E_i\big)\leq n$  
While projections are mentioned until next chapter, they are a standard tool in abstract & linear algebra so I make use of them now, and the fact that $E/E_k = [\alpha_k \cdot x_k]=\alpha\cdot x_k + E_k\cong \alpha_k \cdot x_k = 
\text{image }P^{(k)}$ for well chosen projection $P^{(k)}$. Refer to Pryce ex 10S for basic algebraic properties of projections in functional analysis.    

Do this by induction on $n$ where $n=1$ trivially holds.  
**Base Case:** $n=2$  
suppose for contradiction that we have (at least) $n+1= 3$ linearly independent vectors in $E/(E_1\cap E_2)$, say $\big\{[x_0], [x_1], [x_2]\big\}$  and $V:=\text{span}\big\{x_0, x_1, x_2\big\}$.  Then consider the restrictions $P^{(1)}_{\vert V}:V\rightarrow V$, and $P^{(2)}_{\vert V}:V\rightarrow V$.  
Each have 1 dimensional images hence 2 dimensional kernels by rank-nullity and $3=\dim V  \geq \dim \ker P^{(1)}_{\vert V} + \dim \ker P^{(2)}_{\vert V}- \dim\Big(\ker P^{(1)}_{\vert V}\cap\ker P^{(2)}_{\vert V}\Big)=4-\dim\Big(\ker P^{(1)}_{\vert V}\cap\ker P^{(2)}_{\vert V}\Big)$ $\implies 1\leq \dim\Big(\ker P^{(1)}_{\vert V}\cap\ker P^{(2)}_{\vert V}\Big)$  
$\implies \text{non-zero }v\in V\in \ker P^{(1)}\cap \ker P^{(2)}\cong E_1\cap E_2$ contradicting the fact that $V$ is 3 dimensional. I.e. $v$ is non-zero in $V$ but $\cong [0] \in E/(E_1\cap E_2)$   
Put differently $\big\{v, x_1, x_2\big\}$ form a basis for $V\big[$linear independence is implied by differing behavior under $P^{(1)}_{\vert V}$ and $P^{(2)}_{\vert V}\big]$ but $\big\{v, x_1, x_2\big\} \cong \big\{[0], [x_1], [x_2]\big\}$ so checking dimensions, conclude $2=3$ which is a contradiction.   

**Inductive Case:**  
suppose for contradiction that we have (at least) $n+1$ linearly independent vectors in $E\big / \big(\cap_{i=1}^n E_i\big)$, say $\big\{[x_0], [x_1], [x_2], \dotso, [x_n]\big\}$  and $V:=\text{span}\big\{x_0, x_1, x_2, \dotso, v_n\big\}$.  Then consider the restrictions $P^{(1)}_{\vert V}:V\rightarrow V$, and $P^{(2)}_{\vert V}:V\rightarrow V$. Where $\text{image }P^{(1)}\cong E/E_1$ and $\text{image }P^{(2)}\cong E\big / \big(\cap_{i=2}^n E_i\big)$ so by induction hypothesis $\dim \text{image }P^{(2)}\leq n-1$.  Thus rank-nullity tells us $\dim \ker P^{(2)}_{\vert V}=n$ and $\dim \ker P^{(2)}_{\vert V}\geq 2$.  

$n+1=\dim V \geq  \dim \ker P^{(1)}_{\vert V} + \dim \ker P^{(2)}_{\vert V}- \dim\Big(\ker P^{(1)}_{\vert V}\cap\ker P^{(2)}_{\vert V}\Big)\leq n+2\dim\Big(\ker P^{(1)}_{\vert V}\cap\ker P^{(2)}_{\vert V}\Big)$ $\implies 1\leq \dim\Big(\ker P^{(1)}_{\vert V}\cap\ker P^{(2)}_{\vert V}\Big)$  
$\implies \text{non-zero }v\in V\in \ker P^{(1)}\cap \ker P^{(2)}\cong \bigcap_{i=1}^n E_i$ contradicting the fact that $\dim V= n+1$. I.e. $v$ is non-zero in $V$ but $\cong [0] \in E/(E_1\cap E_2)$

**13.**  
Prove that if $p\geq q \geq 1$ then $L_p([a,b])\subseteq L_q([a,b])$   for every finite segment $[a,b]$   

*remark:* this is closely related to ex 3.6.8 in Nelson.  Though the Lebesgue integral is not formally used until the 2nd 'semester' of this book, I use it now.  

Suppose $f \in L_p([a,b])$ and set $r:=\frac{p}{q}\geq 1$ and Hölder conjugate $s$ such that $\frac{1}{r}+\frac{1}{s}=1$ and $\vert f(x)\vert^q = \vert g(x)\vert$    

$$\int_a^b\big \vert  f \big\vert^q dm =\int_a^b 1 \cdot \big \vert  g \big\vert dm\leq \left(\int_a^b 1^s dm \right)^\frac{1}{s}\cdot\left(\int_a^b\big \vert  g \big\vert^r dm \right)^\frac{1}{r}=m\big([a,b]\big)^\frac{1}{s}\cdot\left(\int_a^b  \big(\big \vert  f \big\vert^q\big)^\frac{p}{q} dm \right)^\frac{1}{r}=m\big([a,b]\big)^\frac{1}{s}\cdot\big\Vert f \big \Vert_p^\frac{1}{r}$$   
raising each side to $\frac{1}{q}$ gives the result

**15.**  
Let $p,q,r$ be numbers such  that $p,q,r\gt 1$ and $\frac{1}{p}+\frac{1}{q}+\frac{1}{r}=1$.  Let $f\in L_p(a,b)$, $g\in L_q(a,b)$, and $h\in L_r(a,b)$.  Prove that $fgh\in L_1(a,b)$ and $\big\Vert fgh  \big\Vert_1 \leq \big\Vert f  \big\Vert_p\big\Vert g  \big\Vert_q \big\Vert h  \big\Vert_r $

remark: this is essentially the integral (instead of sum) form of ex 9.9 in Steele's *The Cauchy-Schwarz Masterclass*.  Since $\big\Vert f  \big\Vert_p \big\Vert g  \big\Vert_q \big\Vert h  \big\Vert_r \lt \infty$ by assumption, the problem reduces to showing  $\big\Vert fgh  \big\Vert_1 \leq \big\Vert f  \big\Vert_p\big\Vert g  \big\Vert_q \big\Vert h  \big\Vert_r $ as that implies $fgh \in L_1(a,b)$.  Applying Hölder's inequality twice yields     


$$\big\Vert fgh  \big\Vert_1 = \int_a^b \vert f\vert \cdot \vert gh\vert dm \leq \big\Vert f  \big\Vert_p\cdot \big\Vert gh  \big\Vert_s=\big\Vert f  \big\Vert_p\cdot\left(\int_a^b  \vert g\vert^s \cdot \vert h\vert^s dm \right)^\frac{1}{s}\leq \big\Vert f  \big\Vert_p\cdot\big\Vert g  \big\Vert_q\cdot\big\Vert h  \big\Vert_s$$  

Note that $s$ was chosen to be a Hölder conjugate to $p$, i.e. $\frac{1}{p}+\frac{1}{s}=1$,  and $\frac{1}{s} = \frac{1}{q}+\frac{1}{r}\iff 1 = \frac{s}{q}+\frac{s}{r}$ so Hölder tells us  

$$\left(\int_a^b  \vert g\vert^s \cdot \vert h\vert^s dm\right)^\frac{1}{s} \leq \left(\big\Vert \vert g\vert^s\big\Vert_{\frac{q}{s}}\cdot \big\Vert \vert h\vert^s\big\Vert_{\frac{r}{s}}\right)^\frac{1}{s} =\left)\left(\int_a^b  \vert g\vert^q  dm\right)^\frac{s}{q}\cdot \left(\int_a^b  \vert h\vert^r  dm\right)^\frac{s}{r}\right)^\frac{1}{s}=\big\Vert g  \big\Vert_q\cdot\big\Vert h  \big\Vert_r$$

It seems to me that a more basic argument is to observe that if $\Vert x_n-x_m\Vert \lt \epsilon$ for all $n,m \geq N$ then the $k$th component is real Cauchy and hence it converges to a limit $L_j$ and and $\vert x_n^{(j)}-L_j\vert\to 0$ which implies a vector $L$ having its jth component be $L_j$. [This is step 1 in the Cauchy sequence recipe on page 17]  

we know $L\in l^\infty$ since the components of $x_n$ are uniformly bounded (by $\infty$ norm) so $\big\Vert L \big \Vert \leq \big \Vert x_n \big \Vert +\epsilon$ for any $\epsilon\gt 0$ for $n$ large enough.  [This is step 2 in the Cauchy sequence recipe on page 17]      

Finally, we need to show $\Vert x_n -L\Vert\to 0$.  Suppose not, then there is some $\epsilon \gt 0$ such that  
$\Vert x_k -L\Vert\geq \epsilon$ for $k \in S$ where $\vert S\vert =\infty$.  Thus  $\Vert x_k^{(j)} -L_j\Vert\geq \epsilon$ for infinitely many $(k,j)$ tuples and for $k\in S$ [with appropriate $j$] we have   

$\epsilon-\Vert L_j-x_n^{(j)} \Vert \leq \Vert x_k^{(j)} -L_j\Vert -\Vert L_j-x_n^{(j)} \Vert\leq \Vert x_k^{(j)} -L_j + L_j-x_n^{(j)} \Vert=\Vert x_k^{(j)} -x_n^{(j)} \Vert\lt \frac{\epsilon}{2}$  

but the LHS has a limit as $n\to \infty$ of $\epsilon$ while the RHS trivially has a limit of $\frac{\epsilon}{2}$ and that is contradiction.   

To be clear, the above reads: $\Vert x_k^{(j)} -x_n^{(j)} \Vert\lt \frac{\epsilon}{2}$ for $k, n \geq K$  by virtue of $(x_n)$ being a Cauchy sequence *and* the bound holds for all $j\in \mathbb N$ since we are using the sup norm (the convergence occurs uniformly), so after selecting $k\in S$ such that $k\geq K$, select and fix the associated $j$  [matching the tuple $(k,j)$] and now using that same $j$ throughout, and finally: consider *any* $n\geq k$.

Prove that if $M$ is a closed subspace and $N$ is a finite dimensional subspace [each] of a normed linear space then the sum $M+N$ is closed. [Hint consider $\dim N=1$ and induct]  

*remarks:* I originally was not planning on doing this exercise however it is useful in dealing with some technical nits in ex 2.4.27 and more importantly, it seems to be heavily tied in with some chapter 3 ideas.  E.g. with $M:=\big\{0\big\}$ and $N:=E$ then this exercise implies the answer for ex 3.3.3.  On the other hand the case of $\dim N=1$ nicely ties in with Hahn-Banach theorem and I give a proof in that vein, using chapter 2 and 3 ideas, below.  


If it enough consider the case where $M+N=M\oplus N$.  In particular induct on the dimension and use bunching.  If $M+N\neq M\oplus N$ then we can find a non-zero vector $v$ in common and write this as $M+N'$ where $N'$ has dimension one less [i.e. make a basis for $N$ starting with $v$ then $N'$ has that same basis with $v$ removed]. Further the inductive case, via bunching, is for $M\oplus N = \big(M \oplus \text{span}\{v'\}\big)\oplus N'$ where $M\oplus\text{span}\{v'\}$ is closed by the Base Case and $\dim N'=\dim N-1$ so the result follows by  induction hypothesis.  I.e. this is one of those less common inductive problems where the inductive case is trivial and the Base Case takes some work.  

Base Case:  
Consider the subspace $V=M\oplus N = M\oplus\big\{\lambda \cdot v\big\}\subseteq X$ where $\dim N=1$.  This implies a linear functional $f:V\rightarrow \mathbb F$ with $\ker f=M$ and $f(v)=1$ [the spirit of this comes from [Theorem 2.3.2, though that doesn't seem strictly needed, we can directly verify this is a linear map].  We know $\ker f$ is closed in $V$ since it is closed in $X\supseteq V$ [check any sequence $m_k\in M$ that converges in $X$] hence by ex 2.4.24 $f$ is bounded/continuous, i.e. $f\in V^*$. By Hahn-Banach [Theorem 3.1.2] there is an extension to $F\in X^*$.   

Now consider any sequence $w_k\in V$ that converges to $w\in X$.  We have $v_k =\lambda_k \cdot v + m_k$ by the direct sum formulation and $F\big(w_k\big)= \lambda_k$.  Then $\lim_{k\to\infty}\lambda_k=\lim_{k\to\infty} F\big(w_k\big)=F\big(\lim_{k\to\infty}  w_k\big)=F\big(w\big)$, i.e. $\lim_{k\to\infty}\lambda_k= F\big(w\big)=\lambda$.  

Thus $w_k -\lambda_k \cdot v = m_k$ has a limit and and $\lim_{k\to\infty}m_k =m\in M$ since $M$ is closed.  Conclude $w=\lim_{k\to\infty} w_k  =\lim_{k\to\infty}\big(\lambda_k \cdot v + m_k\big)=\lim_{k\to\infty}\lambda_k \cdot w + \lim_{k\to\infty} m_k= \lambda \cdot v + m= m+\lambda \cdot v \in M\oplus N$

**23.**  
Let $(x_n)$ be a Cauchy sequence in a normed space $E$.  Prove that  
(a) if $(x_n)$ contains a subsequence $(x_{n_k})$ converging to a vector $x\in E$ then the sequence $(x_n)$ converges to $x$  
select $\epsilon \gt 0$ and for all $n_k\geq n_K$ we have $\Vert x_{n_k}-x\Vert \lt \frac{\epsilon}{2}$ by definition of convergence and  
$\Vert x_n - x\Vert = \Vert x_n - x_{n_{r}} + x_{n_{r}}  - x\Vert\leq  \Vert x_n - x_{n_{r}}\Vert + \Vert x_{n_{r}}  - x\Vert \leq \frac{\epsilon}{2}+ \frac{\epsilon}{2}=\epsilon$

$\Vert x_{n-k} - L\Vert \lt \frac{1}{2}\cdot 2^{-k}$ and $\Vert x_{n-k+1} - L\Vert \lt \frac{1}{2}\cdot 2^{-k}$  
then  
$\Vert x_{n-k} - x_{n-k+1}\Vert \leq \Vert x_{n-k} - L\Vert +\Vert L - x_{n-k+1}\Vert \lt 2^{-k}$

$\Vert\hat x_{n-k} - \hat L\Vert \lt \frac{1}{2}\cdot 2^{-k}$ and $\Vert \hat x_{n-k+1} -\hat L\Vert \lt \frac{1}{2}\cdot 2^{-k}$
