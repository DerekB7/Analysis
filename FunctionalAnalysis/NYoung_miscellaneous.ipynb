**7.18** Show that if $A,B\in B(E)$ are invertible *iff* both $AB$ and $BA$ are invertible [where invertible is taken to mean has a bounded inverse]    
The forward  directions is immediate  

For the other direction consider e.g.  
$A^{-1}:=(BA)^{-1}B$ which is bounded and since $AB$ is invertible we know $A$ is surjective  [since $A$ has a right inverse given by $B(BA)^{-1}$] hence any $v\in V$ may be written as $Ax=v$ for some $x$ hence   

$A^{-1}v=(BA)^{-1}Bv = (BA)^{-1}B(Ax) = x$  as required; put differently if $x \in \ker A\implies 0 = Ax\implies 0=(BA)^{-1}B0 =(BA)^{-1}B(Ax) = x$ hence $A$ is injective and the earlier stated right inverse $B(BA)^{-1}$ is a left inverse as well hence it is *the inverse* of $A$.  

By nearly identical reasoning, setting $B^{-1}:= (AB)^{-1}A$ gives the result for $B$'s inverse, or just observe that $B=(BA)(A^{-1})$ the product of two invertible operators $\implies B^{-1}=A(BA)^{-1}$ 

**more abstract argument:**    
set $C:=(BA)^{-1}B$ where $CA =I\implies AC = A\big(CA\big)C=\big(AC\big)^2$   
[underlying idea for the proof: any idempotent element that has a right (or left) inverse must be the identity element]  


let $R$ be a (choice of) right inverse of $\big(AC\big)$ with $R:=AB(AB)^{-1}=I$   
$I = A\big((BA)^{-1}B A\big)B(AB)^{-1}= A\big((BA)^{-1}B\big)\big( AB(AB)^{-1}\big) =  ACR = \big(AC\big)^2R =AC\big(ACR\big)=AC $  
note: the last 3 equalities are technically not needed since we know $R=I$ so $ACR=AC$, though I left them in since idempotence and usage of some right inverses was the motivation behind this approach  

conclude: $CA=I=AC$ and $C$ is the (unique) inverse of $A$

*alternatively if we wanted to avoid using the square-root* we could mimic the approach mentioned in ex 7.29 and consider the semi-inner product given by  $(x,y)\mapsto\langle A x, y\rangle$ and consider for arbitrary $x,y\in H$ that $\langle A A^\frac{k-1}{2}x,A^\frac{k-1}{2}y\rangle =0\implies \text{image }A^\frac{k-1}{2}\subseteq \ker A $  since any $v\in \text{image }A^\frac{k-1}{2}$ may be written as $v =A^\frac{k-1}{2}x$  and consider arbitrary $y\in H$ then

$0=Av=AA^\frac{k-1}{2}x = A^\frac{k+1}{2}x\implies A^\frac{k+1}{2}=0$  
$\implies \frac{k+1}{2}\geq k\quad \text{(minimality of k)}\implies k+1\geq 2k\implies k=1$

where all $\lambda_n\gt 0$ since $A^*A$ is injective on $\overline{\text{image }A^*A}=\big(\ker A^*A\big)^\perp$

[projection formula-- in particular $w:=z-\sum_{k=1}^\infty \langle z,\varphi_n\rangle \cdot \varphi_n$ then $w\perp \ker A^*A$ and $w\perp \text{span} \big\{\varphi_j\big\}\implies w \perp\overline {\text{image }A^*A}\implies w=0$]

$Ax = Az = A\big(\sum_{n=1}^\infty \alpha_n\cdot \varphi_n\big)= \sum_{n=1}^\infty \alpha_n\cdot A\varphi_n  = \sum_{n=1}^\infty \lambda_n^\frac{1}{2}\cdot\alpha_n\cdot\lambda_n^\frac{-1}{2} \cdot A \varphi_n$  
$= U\big(\sum_{n=1}^\infty \alpha_n\cdot \lambda_n^\frac{1}{2}\cdot\varphi_n\big)  = U \big(A^*A\big)^\frac{1}{2}\big(\sum_{n=1}^\infty \alpha_n\cdot\varphi_n\big)=U \big(A^*A\big)^\frac{1}{2}z = U \big(A^*A\big)^\frac{1}{2}x $  
where at the end we use $\ker \sqrt{A^*A}=\ker A^*A$ which holds by problem 7.29.

$\big\Vert Ux\big\Vert^2= \langle Ux,Ux \rangle=\langle Uz,Uz \rangle= \langle  z,U^*Uz \rangle=\langle \sum_{n=1}^\infty \alpha_n\cdot \varphi_n , \sum_{n=1}^\infty \alpha_n \big(\lambda_n^\frac{-1}{2}\big)^2\cdot A^*A \varphi_n\rangle=\langle \sum_{n=1}^\infty \alpha_n\cdot \varphi_n , \sum_{n=1}^\infty \alpha_n \varphi_n\rangle=\sum_{n=1}^\infty \vert \alpha_n\vert^2$  
$= \big\Vert z\big\Vert^2\leq \big\Vert y\big\Vert^2 +  \big\Vert z\big\Vert^2=\big\Vert x\big\Vert^2$  
and set e.g. $x:= \varphi_1$ to meet the upper bound with equality  


Thus we have a polar decomposition for compact $A$ on some [not necessarily separable] Hilbert space $H$. Now define $B_r$ as a rank $r$ approximation of $\sqrt{A^*A}$ i.e.  
$B_r x = B_r z = B_r\big(\sum_{n=1}^\infty \alpha_n\cdot \varphi_n\big)=\sum_{n=1}^r \alpha_n\cdot\lambda_n^\frac{1}{2}\cdot \varphi_n$  
where $B_r$ is compact since it has finite rank and is self-adjoint, by checking $\langle B_r x, B_r x\rangle \geq 0$ for all $x$ in our [complex] Hilbert space.  Then  $\sqrt {A^*A}- B_r$ has maximum eigenvalue $\lambda_{r+1}^\frac{1}{2}$ and since all eigenvalues are non-negative this means it is the max modulus eigenvalue of a compact self-adjoint operator hence $\big\Vert \sqrt{A^*A } B_r \big\Vert =\lambda_r^\frac{1}{2}$ by Theorem 8.10 [in Eidelman et al this is Property 6.2.1]  

$\implies \big\Vert U B_r-A\big\Vert = \big\Vert U\big(B_r- \sqrt{A^*A}\big)\big\Vert\leq \big\Vert U\big\Vert \cdot \big\Vert B_r- \sqrt{A^*A}\big\Vert = 1\cdot \lambda_r^\frac{1}{2} \to 0$  

Conclude: every compact operator on a Hilbert spaces is the [norm] limit of a sequence of finite rank operators.  [Note: this does not hold on general Banach spaces due work of Enflo].
