$f\in C[0,1]$ since if $z_k\to z$ then, we may select some $\epsilon \gt 0$ and deduce   
$\big \vert f(z_k)-f(z)\big\vert =\big \vert f(z_k) - f_n(z_k) +f_n(z_k) - f_n(z) + f_n(z)-f(z)\big\vert\leq \big \vert f(z_k) - f_n(z_k)\big\vert +\big\vert f_n(z_k) - f_n(z)\big\vert + \big\vert f_n(z)-f(z)\big\vert\leq \epsilon$

$C[0,1]$ is sometimes defined as a quotient space (recall e.g. Nelson ex 3.6.2) though in example 1.4 of page 2 in MacClure it is not. If we instead considered say $C[\overline {B(0,1)}]$ for unit ball $B(0,1)\in \mathbb R^n$ [for continuous complex valued functions] we'd still have this result as we can select $x\in \overline {B(0,1)}$ and $f,g\in C[\overline {B(0,1)}]$ then for any $v$ and any $\delta \gt 0$ if $f(x)-g(x)\neq 0$ for all $x\in B(v,\delta)$ [formally in the relative topology for $v\in \partial B(0,1)$] then $f$ and $g$ disagree on a set with positive measure hence are in different equivalence classes. Conclude that if $f,g\in C[\overline {B(0,1)}]$ are equivalent [i.e. agree almost everywhere] and are both continuous then $f=g$ everywhere, i.e. each equivalence class only has 1 continuous representative.

**(ii.)**  Show that in the norm defined by $\big\Vert f \big\Vert=\big\Vert f \big\Vert_\infty+\big\Vert f' \big\Vert_\infty$, that $C^1[0,1]$ is a Banach space.   

Suppose $f_n$ is a Cauchy sequence.  Then for any $\epsilon \gt 0$  
$\big\Vert f_n-f_m \big\Vert=\big\Vert f_n-f_m \big\Vert_\infty+\big\Vert f_n'-f_m' \big\Vert_\infty\lt \epsilon$
for all $n,m\geq N$.  Thus $f_n$ and $f_n'$ are themselves Cauchy sequences in $C[0,1]$ and per ex 1.2 $f_n \to f\in C[0,1]$ and $f_n' \to g\in C[0,1]$ where the convergence is in norm in each case.  

It remains to show that $g=f'$   
By the Fundamental Theorem of Calculus we have  

$$f_n(x) = \int_0^x f_n'(t)dt\implies f(x) = \lim_{n\to\infty}f_n(x) = \lim_{n\to\infty} \int_0^x f_n'(t)dt= \int_0^x \lim_{n\to\infty} f_n'(t)dt=\int_0^x g(t)dt$$     
where the integral and limit interchange is justified by uniform convergence of $f_n'\to g$  
$\implies f'(x) = g(x)$ for $a\in [0,1]$  where the derivative at $0$ and $1$ is to be interpreted as a left and right derivative respectively.   

For refresher on the FTC ref https://math.berkeley.edu/~ogus/Math_1A/lectures/fundamental.pdf  



**1.4**  
Show that $\mathscr l^1$ is complete.  
Consider Cauchy sequence $ \mathbf a^{(n)}=\begin{bmatrix}a_1^{(n)} \\ a_2^{(n)}  \\ a_3^{(n)}\\ \vdots \end{bmatrix}\in \mathscr l^1$  

This implies the $k$th component is Cauchy so $\lim_{n\to\infty}a_k^{(n)}=a_k$ (by completeness of $\mathbb C$) which implies a limiting vector $\mathbf a$.  

- - - - 
We can argue that $\mathbf a \in \mathscr l^1$  
*using measure theory*  
by observing that for say $\epsilon:=1$ there is some $N$ such that $\big\Vert \mathbf a^{(n)}-\mathbf a^{(m)}\big\Vert \lt 1$ for all $n,m\geq N$ which implies $\big\Vert \mathbf a^{(n)}\big\Vert \leq M\lt \infty$ for all $n$.  Using the Counting measure on $\mathbb N$ and Fatou's Lemma we deduce $\big\Vert \mathbf a\big\Vert \leq \liminf_{n\to\infty} \big\Vert \mathbf a^{(n)}\big\Vert \leq M$    

*alternatively for an intro real analysis argument:*  
(the below mimics the closing argument at the end for $\big\Vert \mathbf a_n -\mathbf a\big\Vert \to 0$ but omits some of the details)

Suppose for contradiction that $\sum_{k=1}^\infty \vert a_k\vert =\infty$  
Select $\epsilon \gt 0$.  Being a Cauchy sequence, there is some $N$ such that $\big\Vert \mathbf a^{(n)}-\mathbf a^{(m)}\big\Vert \lt \frac{\epsilon}{4}$ for all $n,m\geq N$.  Since $\big\Vert \mathbf a^{(N)}\big\Vert\lt \infty$, there is some $K\geq 2$ such that and $\sum_{k=K}^\infty \left\vert a^{(N)}_k\right \vert\lt \frac{\epsilon}{4}$
$\implies \sum_{k=K}^\infty \left\vert  a^{(m)}_k\right\vert \lt \frac{\epsilon}{2}$ for all $m\geq N$ (ref reverse triangle inequality argument for details)    

$\sum_{k=1}^\infty \vert a_k\vert =\infty\implies \epsilon \leq \sum_{k=K}^r \vert a_k\vert$   for some $r \gt K$ but  
$\epsilon \leq \sum_{k=K}^r \vert a_k\vert = \sum_{k=K}^r \vert a_k -a_k^{(m)}+a_k^{(m)}\vert\leq \sum_{k=K}^r \vert a_k -a_k^{(m)}\vert +\sum_{k=K}^r \vert a_k^{(m)}\vert $    
and taking limit superiors this reads  
$\epsilon \leq \limsup_{m\to\infty}\big(\sum_{k=K}^r \vert a_k -a_k^{(m)}\vert +\sum_{k=K}^r \vert a_k^{(m)}\vert\big)\leq \limsup_{m\to\infty}\big(\sum_{k=K}^r \vert a_k -a_k^{(m)}\vert\big) +\limsup_{m\to\infty}\big(\sum_{k=K}^r \vert a_k^{(m)}\vert\big)\lt 0 +\frac{\epsilon}{2} $    
which is absurd.  Note the final inequality holds since for all $m\geq N$ we have $\sum_{k=K}^r \vert a_k^{(m)}\vert\leq \sum_{k=K}^\infty  \vert a_k^{(m)}\vert\lt\frac{\epsilon}{2}$  
- - - -  
Finally $\big\Vert \mathbf a_n -\mathbf a\big\Vert \to 0$   
Select $\epsilon \gt 0$.  There is some $N$ such that $\big\Vert \mathbf a^{(n)}-\mathbf a^{(m)}\big\Vert \lt \frac{\epsilon}{4}$ for all $n,m\geq N$.  Further since $\big\Vert \mathbf a\big\Vert\lt\infty$ and $\big\Vert \mathbf a^{(N)}\big\Vert\lt \infty$, there is some $K\geq 2$ such that $\sum_{k=K}^\infty \left\vert a_k\right \vert\lt \frac{\epsilon}{4}$ and $\sum_{k=K}^\infty \left\vert a^{(N)}_k\right \vert\lt \frac{\epsilon}{4}$.  

Let $P$ kill the 1st $K-1$ components and act as the identity on the rest.  For $m\geq N$  
$\sum_{k=K}^\infty \left\vert a^{(N)}_k- a^{(m)}_k\right \vert =\big\Vert P\mathbf a^{(N)}-P\mathbf a^{(m)}\big\Vert \leq \big\Vert \mathbf a^{(N)}-\mathbf a^{(m)}\big\Vert \lt \frac{\epsilon}{4}$  
$\implies \sum_{k=K}^\infty \left\vert  a^{(m)}_k\right \vert=\big\Vert P\mathbf a^{(m)}\big\Vert  \lt \frac{\epsilon}{4}+\big\Vert P\mathbf a^{(N)}\big\Vert \lt \frac{\epsilon}{2}\quad \text{(reverse triangle inequality)}$  

Since $\lim_{n\to\infty}a_k^{(n)}=a_k$ point-wise, this means that for $1\leq k\leq K-1$ there is an $N_k$ such that $\big\Vert a_k^{(n)}-a_k\big\Vert \lt \frac{\epsilon}{4}\frac{1}{K-1}$  for all $n\geq N_k$. Finally set $N':= \max\big(N, N_1,N_2,\cdots, N_{K-1}\big)$ and for all $n\geq N'$ we have  
$\big\Vert \mathbf a^{(n)}-\mathbf a\big\Vert\leq\big\Vert  P(\mathbf a^{(n)}-\mathbf a)\big\Vert +\big\Vert (I-P)(\mathbf a^{(n)}-\mathbf a)\big\Vert\leq \big\Vert  P\mathbf a^{(n)}\big\Vert +\big \Vert P\mathbf a\big\Vert +\big\Vert (I-P)\mathbf a^{(n)}-(I-P)\mathbf a\big\Vert\lt \frac{\epsilon}{2}+\frac{\epsilon}{4}+\frac{\epsilon}{4}=\epsilon$  
(where the 1st inequality is actually an equality though this isn't needed)

**1.5**  
Show that a metric space is complete if every Cauchy sequence has a convergent subsequence.  

Consider Cauchy sequence $\big\{x_n\big\}\implies x_{n_k}\to x\in X$ for some subsequence.  Then select $\epsilon\gt 0$ and for all $n, n_k\geq N_0$ we have $d\big(x_n,x_{n_K}\big)\lt \frac{\epsilon}{2}$ and for all $n_k \geq N_1$ we have $d\big(x_{n_K},x\big)\lt \frac{\epsilon}{2}$ so for all $n,n_k\geq N:=\max\big(N_0,N_1\big)$ we have by triangle inequality   

$d\big(x_n,x\big)\leq d\big(x_n,x_{n_K}\big)+ d\big(x_{n_K},x\big)\lt \epsilon$  
$\implies x_n\to x$  


*remark:*  
This is useful in interpreting completeness in $\mathbb R$ in terms of the typical development in intro real analysis: first the least upper bound property, then Bolzano-Weierstrass property.  Cauchy sequences are bounded hence they have a convergent subsequence (Bolzano-Weierstrass) so $\mathbb R$ is complete, i.e. all Cauchy sequences converge.  Conversely if we know all Cauchy sequences converge, we can (via a bisection argument) derive the least upper bound property as the supremum can be written as the limit of a Cauchy sequence.

$G:=\displaystyle \left[\begin{matrix}\langle v, v\rangle & \langle w, v\rangle\\ \langle v, w\rangle & \langle w, w\rangle\end{matrix}\right]$ 
note $G$ is Hermitian and  
$\mathbf x^* \big(G \mathbf x\big)=\mathbf x^*\displaystyle \left[\begin{matrix}\langle  x_1\cdot v + x_2\cdot w, v\rangle\\ \langle  x_1\cdot v+ x_2\cdot w, w\rangle\end{matrix}\right]=\big\langle x_1\cdot v + x_2\cdot w, x_1\cdot v+x_2\cdot w\big\rangle\geq 0$  
$\implies G\succeq \mathbf 0\implies \langle v, v\rangle \cdot \langle w, w\rangle -\big\vert \langle v, w\rangle \big\vert^2 =\det \big(G\big)\geq 0$  
(either via eigenvalues or via congruence to a PSD diagonal matrix)

i.e. $\big\{v,w\big\}$ is linearly dependent and the inner product is non-negative (hence the scalar multiple is real non-negative)

Then per Prop 1.18   
$$\lim_{n\to\infty}\big\langle f_n, P\mathcal X_{z,r}\big\rangle+0 = \lim_{n\to\infty}\big\langle f_n, P\mathcal X_{z,r}\big\rangle +\lim_{n\to\infty}\big\langle f_n, Q\mathcal X_{z,r}\big\rangle = \lim_{n\to\infty}\big\langle f_n, \mathcal X_{z,r}\big\rangle= \lim_{n\to\infty}\frac{1}{\pi}\int_{B(z,r)} f_n(z)dA =\lim_{n\to\infty} r^2\cdot f_n(z)$$  
and by weak convergence we also have  
$$\lim_{n\to\infty}\big\langle f_n, P\mathcal X_{z,r}\big\rangle=\big\langle f, P\mathcal X_{z,r}\big\rangle =\big\langle f, \mathcal X_{z,r}\big\rangle= \frac{1}{\pi}\int_{B(z,r)} f(z)dA =r^2\cdot f(z)$$   
so using the linear functional $\varphi_{z,r} \cong \big\langle \bullet ,\frac{1}{r^2} P\mathcal X_{z,r}\big\rangle$ gives the result

(b.) show that on $L^p[0,1]$ with $0\lt p\lt 1$, the following is not a norm     

$$\big\Vert f\big\Vert_p=: \left(\int_0^1 \vert f\vert^pdx\right)^\frac{1}{p}$$  


This is not a $p-$norm on $\mathbb C^2$ and we can mimic that proof by considering $A:= \left[0,\frac{1}{2}\right)$ 

$\big\Vert \mathcal X_A + (1-\mathcal X_A)\big\Vert_p=\big\Vert 1 \big\Vert_p= 1\gt\left(\frac{1}{2}\right)^\frac{1}{p} +\left(\frac{1}{2}\right)^\frac{1}{p} =\big\Vert \mathcal X_A\big\Vert_p + \big\Vert \mathcal 1- X_A \big\Vert_p $  

where the inequality holds for rational $p=\frac{m}{n}$ with $m\lt n$ since  
$\left(\frac{1}{2}\right)^\frac{1}{p}=\left(\frac{1}{2}\right)^\frac{n}{m}\lt \frac{1}{2}\iff \left(\frac{1}{2}\right)^n\lt \left(\frac{1}{2}\right)^m$  

However to make this stick for non-rational $p$ we could argue by writing $\frac{1}{p}=\lim_{k} r_k$ where $r_k\in \mathbb Q_{\gt 1}$ is an increasing sequence  yielding $\frac{1}{2}\gt \left(\frac{1}{2}\right)^{r_1}\gt \left(\frac{1}{2}\right)^{r_2}\gt \left(\frac{1}{2}\right)^{r_3}\gt \cdots \implies\frac{1}{2}\geq \left(\frac{1}{2}\right)^{r_1} \gt \lim_{k\to\infty}\left(\frac{1}{2}\right)^{r_k}= \left(\frac{1}{2}\right)^\frac{1}{p} $   
hence $\left(\frac{1}{2}\right)^\frac{1}{p}$ is the limit of a monotone decreasing sequence and the result follows.  Alternatively write $\left(\frac{1}{2}\right)^\frac{1}{p}=\exp\left(\frac{1}{p}\cdot\ln \left(\frac{1}{2}\right)\right)=\exp\left(\frac{1}{p}\cdot -\ln \left(2\right)\right)$  
hence $\frac{1}{p}\cdot -\ln \left(2\right)\lt -\ln \left(2\right)$ since $\frac{1}{p}\gt 1$ and $\ln(2)\gt 0$ then use the fact that the exponential function is strictly monotone increasing [1st derivative]  

The result should also be implied by a strict concavity (negative convexity argument, i.e. the fact that the 2nd derivative of $x\mapsto x^p$ is negative for $0\lt p\lt 1$ and $x\gt 0$.
