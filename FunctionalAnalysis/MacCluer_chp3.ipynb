**Notes on isometries on Banach spaces**    

The notion of an isometry $\psi$ comes up in context of reflexive spaces though the ideas are not fully flushed out. Using the inner product notation from the bottom of page 42, suppose we have  
$x^*(x)=\big\langle x,x^*\big\rangle=\big\langle \psi^{-1} x^{**},x^*\big\rangle=\big\langle x^{**},(\psi^{-1})^* x^*\big\rangle=\big\langle x^{**},\psi^{-*} x^*\big\rangle$  
where $\psi: X\rightarrow \hat X\subseteq X^{**}$ a surjective linear isometry [ref page 52]    
$\implies \psi^*: (\hat X)^*\rightarrow X^*$  

Linearity is immediate from the adjoint definition on page 52.  Invertibility follows [re-labelling $\psi^{-*}$ as s $\varphi^*$ helps clean up assumption of invertibility] since  
$\big\langle  x,\psi^*\varphi{^*} x^*\big\rangle=\big\langle \psi x,\psi^{-*} x^*\big\rangle= \big\langle \psi x,\varphi{^*} x^*\big\rangle=\big\langle \varphi\psi x, x^*\big\rangle =\big\langle x, x^*\big\rangle$  
and by uniqueness of the adjoint $\psi^*\varphi{^*}=\text{id}_{X^*}$

$\big\langle  x,\psi^*\varphi{^*} x^*\big\rangle=\big\langle \psi x,\psi^{-*} x^*\big\rangle= \big\langle \psi x,\varphi{^*} x^*\big\rangle=\big\langle \varphi\psi x, x^*\big\rangle =\big\langle x, x^*\big\rangle$  

but, using the RHS of the 1st line and the fact that $\psi\varphi=\text{id}_{\hat X}$  
$x^*(x)= \big\langle x^{**},\psi^{-*} x^*\big\rangle=\big\langle x^{**},\varphi^{*} x^*\big\rangle=\big\langle \text{id}_{\hat X} x^{**},\varphi^{*} x^*\big\rangle=\big\langle \psi\varphi x^{**},\varphi^{*} x^*\big\rangle=\big\langle \varphi x^{**},\varphi^*\psi^* x^*\big\rangle =\big\langle  x,\varphi^*\psi^* x^*\big\rangle=x^*(x)=\big\langle x,x^*\big\rangle$  
$\implies \varphi^*\psi^* =\text{id}_{(\hat X)^*}$ by uniqueness of the adjoint  
$\implies \varphi =\psi^{-*}$  

Finally it remains to prove that $\psi^*$ is an isometry or equivalently that $\psi^{-*}$ is an isometry (since their composition is the relevant identity this should suffice)  

Revisiting the first line.  Select arbitrary $x^*\in X$ with $\Vert x^*\Vert =1$.. Then there is a sequence of $x_k\in X$ with $\Vert x_k\Vert =1$ such that 
$\big\vert \big\langle x_k^{**},\psi^{-*} x^*\big\rangle\big\vert=\big \vert \big\langle \psi^{-1} x_k^{**},x^*\big\rangle\big \vert  =\big\vert\big\langle x_k,x^*\big\rangle\big\vert= \big\vert x^*(x_k)\big\vert \to 1^+$  

$\implies\big \Vert \psi^{-*}x^* \big\Vert\geq 1$ since $\psi^{-*}$ sends $x^*$ to an element with norm at least one    
If that inequality is strict, then there is some unit length $x^{**}$ such that 
    $1\lt \big\vert \big\langle x^{**},\psi^{-*} x^*\big\rangle\big\vert =\big\vert \big\langle \psi^{-1}x^{**},x^*\big\rangle\big\vert  =\big\vert \big\langle x,x^*\big\rangle\big\vert \leq 1\quad \text{(absurd)}$  
The final equality holds since $\psi^{-1}$ is an isometry hence $\Vert x\Vert =1$ and the final inequality is because $\Vert x^*\Vert =1$  

Conclusion: after a large amount of symbol manipulation we confirm that $\psi^*$ inherits invertibility and being an isometry from $\psi$

**3.2**  
On a vector space $X$, a *real* linear functional $\phi$ is one that respects the abelian group homomorphism and scaling by $\alpha \in \mathbb R$   

**(a.)** for $z \in \mathbb C$ we have $z=a+bi$ for $a,b\in \mathbb R$ so  
$z=a + bi = \text{Re }(z) -i\cdot \text{Re }(-b+ai) =\text{Re }(z) -i\cdot \text{Re }(iz)$   

**(b.)** Suppose $X$ is a complex vector space and $\varphi$ is a complex-linear functional on $X$.  Define $u:X\longrightarrow \mathbb R$ by $u(x)=\text{Re }\varphi(x)$.  Note: $u$ is a real linear functional. If $\varphi(x)=z$ then $\varphi(x)=z= \text{Re }(z) -i\cdot \text{Re }(iz)=\text{Re }\varphi(x) -i\cdot \text{Re }\varphi(ix)=u(x) - i\cdot u(ix)$ per part (a.)  

**(c.)**  Conversely if $u$ is a real linear functional then define $\varphi:X\longrightarrow \mathbb C$ by $\varphi(x)= u(x) - i\cdot u(i\cdot x)$.  For the abelian group homomorphism, consider $v,x\in X$   

$\varphi(x+v)= u(x+v) - i\cdot u(i\cdot (x+v))= u(x) - i\cdot u(i\cdot x)+u(v)- i\cdot u(i\cdot v)=\varphi(x)+\varphi(v)$  

for $\alpha \in \mathbb R$   
$\varphi(\alpha \cdot x)= u(\alpha \cdot x) - i\cdot u(i\cdot \alpha \cdot x)= \alpha \cdot u( x) - i\cdot u( \alpha \cdot i\cdot x)= \alpha \cdot \Big(u( x) - i\cdot u( i\cdot x)\Big)=\alpha \cdot \varphi( x)$   

for $\beta \in i\cdot\mathbb R$ so $\beta = i\cdot \alpha\implies \alpha = -i \cdot \beta$   
$\varphi(\beta \cdot x)= u(\beta \cdot x) - i\cdot u(i\cdot \beta \cdot x)= u(\alpha \cdot i\cdot x) - i\cdot u(-\alpha \cdot x)= \alpha \cdot u( i\cdot x) - \alpha \cdot i\cdot u(- x)$  $= \alpha \cdot u( i\cdot x) - \alpha \cdot i\cdot u(- x) = \alpha \cdot u( i\cdot x) + \alpha \cdot i\cdot u( x)= (-i\cdot\beta) \cdot u( i\cdot x) + (-i\cdot \beta) \cdot i\cdot u( x)= \beta \cdot \Big(-i\cdot u( i\cdot x) + u( x)\Big)=\beta \cdot \varphi(x)$   

The above properties are enough to establish that $\varphi$ respect scaling by $\mathbb C\ni \lambda =\alpha+\beta$ then   
$\varphi(\lambda \cdot x) =\varphi(\alpha \cdot x+ \beta\cdot x) = \varphi(\alpha \cdot x)+ \varphi(\beta\cdot x)  =\alpha \cdot\varphi( x)+ \beta\cdot\varphi( x)=(\alpha +\beta) \cdot\varphi( x)=\lambda \cdot \varphi( x)$   


**(d.)**  
Show that $\big\Vert \varphi \big\Vert = \big\Vert u \big\Vert$  
For any $x\in X$ we have $\big\Vert \varphi(x)\big\Vert =\big\vert a+bi\big\Vert= \sqrt{a^2 + b^2}\geq \vert a\vert  = \big\Vert u(x)\big\Vert$   
$\implies \big\Vert \varphi\big\Vert \geq  \big\Vert u\big\Vert$ since e.g. for any $\epsilon \gt 0$ we have some $\Vert v\Vert =1 $ such that $\big\vert \varphi(v)\big\vert\geq \big\vert u(v)\big\vert=\big\Vert u\big\Vert -\epsilon$  

Now let $\big\{x_k\big\}$ with $\Vert x_k\Vert =1$ such that  $\varphi(x_k)\to \big\Vert \varphi \big\Vert$ so  
$u(x_k)=\text{Re}\varphi(x_k)=\varphi(x_k)\to \big\Vert \varphi\big\Vert\implies \big\Vert \varphi\big\Vert \leq  \big\Vert u\big\Vert$

Let $L_{\mathbb R}^\infty$ be the space of real-valued essentially bounded function on $[0,1]$ with respect to the Lebesgue measure.  Let $K$ be the subspace of constant functions.  Define $f:K\longrightarrow \mathbb R$ by $f(k)=k$ where on the LHS $k$ denotes the constant function with value $k$ [this is an abuse of notation though somewhat improved from MacCluer trying to use $c$ instead of $k$ here which clashes with the 1 step extension process discussed on pages 52 and 53].  Let $g_0(x)=x$ and set $N:=\big\{k+\alpha\cdot g_0: k\in K, \alpha\in \mathbb R\big\}$.  The proof of the one-step extension process in the Hahn-Banach theorem tells you how to find all linear $F:N\longrightarrow \mathbb R$ so that $F$ extends $f$ and $\big\Vert F\big\Vert =\big\Vert f\big\Vert$.  Find all such $F$.  


note that for $k \in K$ we have  

$\begin{cases}
\big\Vert k-g_0\big\Vert_\infty&=\vert k\vert+1 &\text{ for } k\leq 0 \\
&=1-k &\text{ for } 0\leq k \leq \frac{1}{2}\\
&=k &\text{ for } \frac{1}{2}\leq k\\
\end{cases}$  
and applying (3.3) with $F(g_0)=c$ and recalling $f(k)=k$ we have  

$k-\big\Vert k-g_0\big\Vert \leq c\leq k+\big\Vert k-g_0\big\Vert$ for all $k\in K$ 

when $k\leq 0$ this $\implies 2k -1 \leq c\leq k+\vert k \vert +1 = 1$  
when $k\geq \frac{1}{2}$ this $\implies 0\leq c\leq 2k\quad \text{(upper bound is at least one as before, but new lower bound of zero)}$  
when $0\leq k\leq\frac{1}{2}$ this $\implies 2k-1\leq c\leq 1\quad \text{(upper bound of one and lower bound of zero both unaltered)}$  

$\implies F(g_0) \in \big[0,1\big]$, i.e. any such choice works.  

**remarks:**  
$F$ is a linear functional, hence a linear map, and its defined on a 2-dimensional vector space $N$ so it is completely specified by its action on basis vectors for $N$.  More generally if we have some subspace $K$ and we consider $K\oplus\text{span }\big\{g_0\big\}$ then we define $F$ on $K\oplus\text{span }\big\{g_0\big\}$ by $F(k)=f(k)$ for $k\in K$ and $F(\alpha g_0)=\alpha \cdot c$ -- that is what the above exercise is getting at for the one step extension process.  In a different setting we go through this sort of idea of extending a vector space by a one dimensional subspace in ex 3.8.

**3.8**  
Prove Corollary 3.6 [summarized below]  

*claim:*  
Let $X$ be a normed linear space with $x_0\in X$ and $M$ a [not necessarily closed] subspace in $X$.  Suppose that $\inf_{y\in \overline M}\big\Vert x_0-y\big\Vert =\text{dist}\big(x_0, \overline M\big)=d\gt 0$.  There exists $\varphi \in X^*$ with $\varphi(x_0)=1$ and $\varphi(M)=0$ with $\Vert \varphi\Vert =\frac{1}{d}$.  In particular $x_0\in \overline M$ *iff* there is no bounded linear functional on $X$ that is zero on $M$ and nonzero on $x_0$. [Note: $\big\Vert x_0\big\Vert \geq \inf_{y\in M}\big\Vert x_0-y\big\Vert =d$]    

*proof:*  
$\text{dist}\big(x_0, M\big)=d\gt 0\implies x_0\not\in \overline M$ [recall exercise 3.6 in McCleary]  

So consider the normed vector (sub)space $V:=\overline M \oplus \text{span}\big\{x_0\big\}\subseteq X$.  Write $v = u +w$ with $u\in \overline M$ and $\text{span}\big\{x_0\big\}\ni w=\alpha \cdot x_0$.  And consider $f\in V^*$ specified by $f\big(x_0\big)=1$ and $\overline M = \ker f$    

$f\big(v\big)=f \big(u+w\big)=f \big(u\big)+f\big(w\big) = 0 +f\big(w\big)=\alpha \cdot f\big(x_0\big)=\alpha$  
Linearity of $f$ is immediate, and we show the norm (boundedness) below.    

Consider  $\big\Vert x_0\big\Vert =\text{dist }\big(x_,\big\{\mathbf 0\big\}\big) \geq \text{dist }\big(x_,\overline M\big)=\inf_{y\in \overline M}\big\Vert x_0-y\big\Vert =d= \lim_{k\to \infty} \big\Vert x_0 - y_k\big\Vert$  where the infimum is written as the limit of a sequence of $y_k\in \overline M$ on the right 

$\implies 1 = f\big(x_0\big) = \lim_{k\to \infty} f\big(x_0 - y_k\big)$  
where for any $\epsilon \gt 0$ we know $d\leq \big\Vert x_0-y_k\big\Vert \leq d+\epsilon$ for all $k$ large enough and notice this implies     
$\frac{1}{d+\epsilon}\leq \big\Vert x_0-y_k\big\Vert^{-1}=f\left(\frac{x_0 - y_k}{\big\Vert x_0-y_k\big\Vert}\right)\implies \frac{1}{d}\leq \big\Vert f\big\Vert $ since $\epsilon$ was arbitrary  

Now consider $\lambda \in \mathbb C$   
$v = u + \lambda \cdot x_0$  
if $\lambda  =0$ there is nothing to do as $f\big(v\big)=0$ so consider non-zero $\lambda$ and $-\lambda \cdot y:= u\in \overline M$   
$\vert \lambda\vert \cdot d=\vert \lambda\vert \cdot \inf_{y\in \overline M}\big\Vert x_0-y\big\Vert \leq \vert \lambda\vert \cdot\big\Vert  x_0-u  \big\Vert = \big\Vert \lambda\cdot (x_0-u)  \big\Vert  = \big\Vert v \big\Vert $   
$\big\vert f(v)\big\vert=\big\vert f(\lambda \cdot x_0)\big\vert= \vert \lambda\vert$  


$\implies \big\Vert f\big\Vert =\sup\left\{\frac{\vert f(v)\vert}{\Vert v\Vert}:  v\neq \mathbf 0\right\}=\max\left(\sup\left\{\frac{\vert f(v)\vert}{\Vert v\Vert}:  \lambda =0, v\neq \mathbf 0 \right\},\sup\left\{\frac{\vert f(v)\vert}{\Vert v\Vert}:  \lambda \neq 0,v\neq \mathbf 0 \right\}\right) =\sup\left\{\frac{\vert f(v)\vert}{\Vert v\Vert}:  \lambda \neq 0,v\neq \mathbf 0 \right\}$  
$\leq \sup\left\{\frac{\vert f(v)\vert}{\vert \lambda\vert \cdot d}:  \lambda \neq 0\right\}=\frac{\vert \lambda\vert}{\vert \lambda\vert \cdot d}=\frac{1}{d}$  
The 1st equality is ex 1.16, the 2nd equality comes from bi-partitioning, and the 3rd equality comes from the fact that $\sup\left\{\frac{f(v)}{\Vert v\Vert}:  \lambda =0, v\neq \mathbf 0 \right\} =0 $.  The final inequality comes from the above work which shows$\frac{1}{\Vert v \Vert}\leq \frac{1}{\vert \lambda\vert \cdot d}$.  

By Hahn-Banach $f$ has an extension $\varphi$ to $X$ with $\big\Vert \varphi\big\Vert =\big\Vert f\big\Vert =\frac{1}{d}$  

*Re: the final line of the claim in the first paragraph ("in particular...")*    
For all $f\in X^*$ such that $M\subseteq \ker f\implies \overline M\subseteq \overline {\ker f}=\ker f$ where we use $f$ is continuous so its kernel is closed   

if $x_0\in \overline M$ then trivially $f\big(x_0\big)=0$     
else $x_0\not\in \overline M\implies \text{dist}\big(x_0, \overline M\big)=d\gt 0$ and per the above, consider setting $f:=\varphi$ so $f\big(x_0\big)=\varphi\big(x_0\big)=1\neq 0$

**corollary: finite dimensional vector spaces are closed**  
Consider finite dimensional subspace $V\subseteq X$ and induct on $\dim V = d$

Base Case: $d=0$   
The zero vector space $\big\{\mathbf 0\big\}$ is trivially a topologically closed set-- in fact it is closed even in spaces that aren't Hausdorf but only have the weaker $T_1$ axiom [recall McCleary Prop 3.9, though I don't think we consider anything non-Hausdorff in this book].  In a normed (hence metric) space any point $p\neq \mathbf 0$ has an open ball around it not meeting zero hence the complement of $\big\{\mathbf 0\big\}$ is open so $\big\{\mathbf 0\big\}$ is closed.  
    

Inductive Case: $d\in \mathbb N$     
Let $\big\{v_j\big\}_{j=1}^d$ be a basis for $V$.  For $j\in \big\{1,\dotso, d\big\}$ apply Corollary 3.6 to $v_j$ and $W_j:=\text{span}\Big(\big\{v_1, \dotso , v_d\big\}\setminus \big\{v_j\big\}\Big)$ and we know that $W_j$ is closed by induction hypothesis [note in the case $d=1$ then define $W_1:=\big\{\mathbf 0\big\}$].  This gives us a Biorthogonal system of $\big\{f_j\big\}_{j=1}^d\in X^*$ such that $f_j\big(v_k\big) =\delta_{j,k}$ 

Now define $P:X\longrightarrow X$ with $P\big(x\big)=\sum_{k=1}^d f_k\big(x\big)\cdot v_k$ and via triangle inequality confirm $P$ is bounded plus algebraically check $P^2=P$, i.e. $P$ is a projection and so is $Q:=I-P$.  By construction  $\text{image }P\subseteq \text{span}\big\{v_1, v_2, \dotso, v_d\big\}$ and since each basis vector is in the image of $P$ -- i.e. $Pv_k = v_k$ then the subspace $\text{image }P\supseteq \text{span}\big\{v_1, v_2, \dotso, v_d\big\}$ so they are equal.  

Conclude: $V=\text{span}\big\{v_1, v_2, \dotso, v_d\big\}=\text{image }P=\ker \big(I-P\big) = Q^{-1}\Big(\big\{\mathbf 0\big\}\Big)$ is the pre-image of a closed set under a continuous map hence is closed.

**corollary: all linear maps with a finite dimensional domain are continuous**  

Let $\dim V = n$ and consider linear $S:V\longrightarrow W$  

$m=\dim\text{ image T}\leq n$ so $\text{ image T}$ is closed and define 
$T:V\longrightarrow \text{image }S$, where $\text{ image }S\subseteq W$ inherits its norm from $W$    
and $T$ is given by $Tv :=Sv$ for $v\in V$.  Note that $T$ is bounded *iff* $S$ is bounded.     

$\mathbf B=\bigg[\begin{array}{c|c|c|c|c} v_1 & v_2 &\cdots & v_{n}\end{array}\bigg]$ and $\mathbf B'=\bigg[\begin{array}{c|c|c|c|c} w_1 & w_2 &\cdots & w_{m}\end{array}\bigg]$   
(recall that $m\leq n$ since linear maps cannot increase rank/dimension)     

$T\mathbf B =\mathbf B'D$ where $D$ is a short and wide 'diagonal' matrix with with $d_{k,k}=1$ for $1\leq k\leq m$ and is zero otherwise.   

$\big[$These bases are implied by Rank Normal Form, so $A=Q^{-1}DP$ from chapter 4 of Artin, though since we work over $\mathbb C$ we could use SVD instead.  E.g.  
$T\mathbf B_1 =\mathbf B_2A=\mathbf B_2Q^{-1}DP\implies T\mathbf B=T\big(\mathbf B_1P^{-1}\big) =\big( \mathbf B_2Q^{-1}\big)D=\mathbf B'D\big]$  

For $1\leq j\leq m$ we have $Tv_1 = w_1$ so consider linear functional given by  $f_j(v_k)=\delta_{jk}$ [recall from finite dimensional linear algebra a linear map is completely specified by its action on a basis] which is bounded per the previous optional corollary [alternatively use Hahn-Banach Corollary 3.6, i.e. ex 3.8, to create/justify from scratch]

$\implies x\mapsto f_j(x)\cdot w_j$ is a bounded map with the norm of $\Vert f_j\Vert \cdot \Vert w_j\Vert$ 
$\implies Tv =\sum_{j=1}^m f_j(v)\cdot w_j$  
where the equality holds since both sides are linear and equal when evaluated on the basis vectors in $\mathbf B$   
$\implies \big\Vert T \big\Vert\leq \sum_{j=1}^m \big\Vert f_j(v)\big\Vert \cdot \big\Vert w_j\big\Vert\lt \infty$  
Conclude: $T$ is bounded hence $S$ is bounded and thus continuous.    


**corollary: all norms are topologically equivalent over a finite dimensional vector space**  
[where as usual per this book, we work over $\mathbb C$, though $\mathbb R$ works here as well]  
The idea here is similar in spirit to Theorem 3.22 (Two-Norm Theorem)  

Let $X$ be a finite dimensional vector space and consider the normed spaces  
$\big(X, \big\Vert \bullet \big\Vert_1\big)$ and $\big(X, \big\Vert \bullet \big\Vert_2\big)$  

Define $T:\big(X, \big\Vert \bullet \big\Vert_1\big)\longrightarrow \big(X, \big\Vert \bullet \big\Vert_2\big)$ given by $Tx:=x$.  Then $T$ is continuous, hence bounded, per the preceding corollary and $\big\Vert x\big\Vert_2=\big\Vert Tx\big\Vert_2\leq \big\Vert T\big\Vert \cdot \big\Vert x\big\Vert_1$, so $ \big\Vert \bullet \big\Vert_1$ is a stronger norm than $\big\Vert \bullet \big\Vert_2$.  Running the argument in reverse on with $T^{-1}$ we deduce $\big\Vert \bullet \big\Vert_2$ is a stronger norm than $\big\Vert \bullet \big\Vert_1$, i.e. they are topologically equivalent.  

note: this result is proven via different means in the next chapter (Theorem 4.2).

By the same argument with $P$ we have $\text{image }S=\text{span }\big\{\mathbf e_j\big\}_{j=1}^d$.  

$S$ is a surjective bounded linear map between Banach spaces and per Open Mapping Theorem (Theorem 3.16) $S$ sends opens sets in $X$ to open sets in $V$.  Suppose we have find some open set $O\in X$ where $O\subseteq \text{span }\big\{v_j\big\}_{j=1}^{d-1}$.  Then we have open set

**corollary: Riesz' Lemma**  
Let $M$ be a proper closed subspace of an infinite dimensional normed space $X$. For any $\epsilon \in \big(0,1\big)$ there exists $x\in X$ with $\big\Vert x\big\Vert =1$ and $\text{dist}\big(x,M\big)\gt 1-\epsilon$.  

Since $M\subsetneq X$ there is some $x_0$ such that $\text{dist}\big(x,M\big)= d\gt 0$.  Per the above Corollary 3.6, there is some $\varphi\in X^*$ with $\big\Vert \varphi\big\Vert =\frac{1}{d}$ and $M\subseteq \ker \varphi$.  Now let $\Phi:=d\cdot \varphi$ and as before $\Phi$ kills $M$ and $\big\Vert \Phi\big \Vert =1$.  This means there is a sequence of $x_k\in X$ with $\big\Vert x_k\big\Vert =1$, and using the above $\epsilon \in \big(0,1\big)$,   

$\implies 1-\epsilon\leq \big\vert \Phi(x_k)\big\vert\leq 1 \text{ for all }k\geq K$  
$\implies 1=\big\Vert x_K\big\Vert = \text{dist}\big(x_K,\big\{0\big\}\big) \geq \text{dist}\big(x_K,M\big)\geq 1-\epsilon$  
where if the RHS was not true, then there would be some $y \in M$ such that $\big\Vert x_K-y\big\Vert =\lambda \lt 1-\epsilon$ and  define $v:= \lambda^{-1}\big(x_K-y\big)$ where $\big\Vert v\big\Vert =1$  
$\implies 1 =  (1-\epsilon)^{-1}\cdot (1-\epsilon) \lt  \lambda^{-1}\cdot (1-\epsilon)  \leq \lambda^{-1}\cdot \big\vert \Phi (x_K)\big\vert = \lambda^{-1}\cdot \big\vert \Phi (x_K-y)\big\vert=  \big\vert \Phi (v) \big\vert  \quad\text{(absurd)}$  
[Technical note: $\lambda \neq 0$ since norms are positive definite and $\Phi(x_K-y)=\Phi(x_K)\neq 0$]  

$\underline{\text{intuition remarks:}}$   
The above argument basically is that $\Phi$ implies the 'same' linear functional on $X/M$, given by $\psi$, and $\big\Vert \psi \big\Vert =1$ so there must be a sequence $\big\Vert [x_k]\big\Vert\to 1$ where $\big\vert \psi\big([x_k]\big)\big\vert\to 1^-$ and $\big\Vert [x_k]\big\Vert=  \text{dist}\big(x_k,M\big)$ which is exactly what we want.  Conceptually this ties in with ex 3.29(b) since $(X/M)^*\cong M^\perp$, where the above $\varphi \in M^\perp$ since $M\subseteq \ker \varphi$ hence there is an invertible linear isometry between it and an element in $(X/M)^*$.  However, it seemed cleanest to give a direct argument here and avoid using the more abstract approach of ex 3.29(b).  

*Strengthening the claim:*     
if $X$ is **reflexive** (e.g. if $\dim X\lt \infty$ or $\mathscr l^p$ for $p \in (1,\infty)$ or of course for Hilbert spaces) we can strengthen this to $\text{dist}\big(x,M\big)= 1$.  
Being *reflexive* implies there is some $\Vert x\Vert =1$ such that $\Phi(x)=1$ (ref Corollary 3.7 or Eidelman et al ex 3.3.7). And re-running the above argument, we now have   
$\implies 1=\big\Vert x\big\Vert = \text{dist}\big(x,\big\{0\big\}\big) \geq \text{dist}\big(x,M\big)\geq 1$  
where if the RHS was not true, then there would be some $y \in M$ such that $\big\Vert x-y\big\Vert =\lambda \lt 1$ and  define $v:= \lambda^{-1}\big(x-y\big)$ where $\big\Vert v\big\Vert =1$  
$\implies 1  \lt \lambda^{-1} =\lambda^{-1}\cdot \Phi (x) = \Phi (v) =\big\vert \Phi (v) \big\vert  \quad\text{(absurd)}$  

**corollary: the closed unit ball is not compact [using the norm topology] for infinite dimensional normed spaces**  
Select some unit length vector and call it $x_1$.  

For $n\geq 2$   
$X\supsetneq V_n:=\text{span}\big\{x_1, \dots, x_{n-1}\big\}$ where $V_n$ is closed since it is finite dimensional  
[the above argument or Prop 4.3 in MacCluer]  

Apply Riesz's lemma with $\epsilon:=\frac{1}{2}$ to find some $\big\Vert x_n\big\Vert=1$ such that $\text{dist}\big( x_n, V_n\big)\geq \frac{1}{2}$.   
Conclude we have a bounded sequence $x_n\in \overline B\big(0,1\big)$ that has no convergent subsequence, since for any $j\in \mathbb N$ and any $n\gt j$ we have  

$x_j\in V_n$ and $\text{dist}\big( x_n, \big\{x_j\big\}\big) \geq \text{dist}\big( x_n, V_n\big)\geq \frac{1}{2}$ hence $\big\{x_j\big\}$ has no Cauchy subsequence thus no convergent subsequence.  Conclude that the Bolzano-Weierstrass property was not met for a set--i.e. $\overline B\big(0,1\big)$--  in a metric space thus that set is not compact.

**3.25**  
Show that the quotient $X\diagup M$ on Banach space $\big(X,\big\Vert \bullet\big\Vert\big)$ by a closed subspace $M$ is a Banach space.  

Begin by showing $\big\Vert x\big\Vert_{X\diagup M}= \inf\big\{\big\Vert x+m\big\Vert : m\in M\big\}$ is a norm on $X\diagup M$.  Note: technically since we are dealing with equivalence classes in the quotient space, it is better to write $\big\Vert [x]\big\Vert_{X\diagup M}=\inf\big\{\big\Vert x+m\big\Vert : m\in M\big\}$.  

Positive definiteness [on the equivalence classes] and homoegeniety of $\big\Vert \bullet \big\Vert_{X\diagup M}$ is immediate.  Triangle inequality follows since for $[x],[y]$ (where $[x]$ refers to the equivalence class of $x$ in the quotient space) we have  

$\big\Vert [x]+[y]\big\Vert_{X\diagup M}\leq \big\Vert x+m_x+y+m_y\big\Vert\leq \big\Vert x+m_x\big\Vert +\big\Vert y+ m_y\big\Vert\leq \Big(\big\Vert [x]\big\Vert_{X\diagup M}+\frac{\epsilon}{2}\Big)+\Big(\big\Vert [y]\big\Vert_{X\diagup M}+\frac{\epsilon}{2}\Big)=\big\Vert [x]\big\Vert_{X\diagup M}+\big\Vert [y]\big\Vert_{X\diagup M}+\epsilon$  
$\implies\big\Vert [x]+[y]\big\Vert_{X\diagup M}\leq \big\Vert [x]\big\Vert_{X\diagup M}+\big\Vert [y]\big\Vert_{X\diagup M}$  
since $\epsilon \gt 0$ was arbitrary  


- - - - -  
Unfortunately showing that $\big(X\diagup M,\big\Vert \bullet\big\Vert_{X\diagup M}\big)$ is complete [i.e. Banach] ends up being a bit involved.  The approach in Bachman & Narci (pages 129-131) is perhaps close to what MacCluer had in mind though it is a bit in the weeds.  

- - - - - - 
The proof in Pryce (pages 59-60) seems more intuitive to me and uses the correspondence between sequences and series.  Pryce says it suffices to show that if $X\diagup M$ (or any normed space $V$) has the property that every absolutely convergent series converges, then said space must be Banach; this is Theorem 4.21 in Pryce and proven as follows:  
If $V$ is a normed space and $\big\{v_k\big\}$ is an arbitrary Cauchy sequence (note the technique reminiscent of the proof of Nelson Theorem 3.1.13--sometimes called Beppo-Levi): select $N_1$ such that $\Vert v_k-v_j\Vert\lt 2^{-1}$ for $k,j\geq N_1$ and then for integer $m\geq 2$ select $N_m\geq N_{m-1}$ such that $\Vert v_k-v_j\Vert\lt 2^{-m}$ for $k,g \geq N_m$ and let $s_1 = v_{N_1}$ and $s_2 = \big(v_{N_2} -v_{N_1}\big) + v_{N_1}= V_{N_2}$ and $s_m = \big(v_{N_{m}} -v_{N_{m-1}}\big)+\big(v_{N_{m-1}} -v_{N_{m-2}}\big)+\dotso +\big(v_{N_{2}} -v_{N_{1}}\big)+v_{N_{1}}= v_{N_{m}}=v_{N_{1}} +\sum_{r=2}^m \big(v_{N_r}-v_{N_{r-1}}\big)$  

Then $\lim_{m\to\infty}\big\Vert s_m\big\Vert \leq \big\Vert v_{N_{1}}\big\Vert +\sum_{r=2}^\infty \big\Vert v_{N_r}-v_{N_{r-1}}\big\Vert=\big\Vert v_{N_{1}}\big\Vert+1$  
i.e. the series absolutely converges hence it converges by assumption on normed $V$, thus  $\lim_{m\to\infty} s_m=\lim_{m\to\infty}v_{N_{m}}$ exists, i.e. our Cauchy sequence $\big\{v_k\big\}$ was found to have a convergent subsequence hence $V$ is complete [MacCluer ex 1.5]. 

recall the 'other side of the coin':    
in a Banach space absolutely convergent series must converge since their sequence of partial sums form a Cauchy sequence, i.e. $s_m:=\sum_{k=1}^m x_k$ then for $j\gt m\geq N$ we have 
$\big\Vert s_j-s_m\big\Vert=\big\Vert\sum_{k=m+1}^j x_j\big\Vert\leq \sum_{k=m+1}^j  \big\Vert x_j\big\Vert\leq \sum_{k=m+1}^\infty  \big\Vert x_j\big\Vert\to 0$ by selecting large enough $N$  
- - - - -  
With Pryce's Theorem 4.21 in hand, suppose we have arbitrary $[x_n]\in X\diagup M$ that absolutely converges as a series in $X\diagup M$, i.e. $\sum_{n=1}^\infty \big\Vert [x_n]\big\Vert_{X/\diagup M}\lt \infty$.  If we are are able to show that that implies the series $\sum_{n=1}^\infty [x_n]$ itself converges, then that must mean $X\diagup M$ is Banach.  

So we may choose $x_n\in X$ (i.e. distinct representatives in the equivalence classes-- we could alternatively say representatives $v_n$ such that $[v_n]=[x_n]$ for $n\in \mathbb N$ though this adds more notation)   
such that $\big\Vert x_n\big\Vert \leq \big\Vert [x_n]\big\Vert_{X\diagup M} +2^{-n}\implies \big\Vert\sum_{n=1}^\infty x_n\big\Vert\leq \sum_{n=1}^\infty \big\Vert x_n\big\Vert\leq \sum_{n=1}^\infty \big(\big\Vert [x_n]\big\Vert_{X\diagup M} + 2^{-n}\big)= 1+\sum_{n=1}^\infty \big\Vert [x_n]\big\Vert_{X\diagup M}\lt \infty$  

Since $X$ is Banach then $\sum_{n=1}^\infty x_n=x\in X$ ('other side of the coin' above).  
Using the quotient map $\Pi$ (page 66) then we get $\Pi: x\mapsto [x]$ and $\Pi$ is a contracting map since $\big\Vert [x]\big\Vert_{X\diagup M}=\inf_{m\in M}\big\Vert x +m\big\Vert\leq \big\Vert x\big\Vert$ by definition of the quotient norm (none of this relies on completeness).  Thus with partial sums $s_r =\sum_{n=1}^r x_n$  we have $\Pi s_r =\sum_{n=1}^r [x_n]$ and   
$[x]=\Pi x = \Pi \sum_{n=1}^\infty x_n = \Pi\lim_{r\to\infty }s_r =\lim_{r\to\infty }\Pi s_r=\sum_{n=1}^\infty [x_n]$ and since $\sum_{n=1}^\infty [x_n]$ converges to $[x]$ we conclude that $X\diagup M$ is Banach.

$\Big(T\big(\alpha\varphi\big)\Big)\big(x\big)= \big(\alpha \varphi\big)\big([x]\big)= \alpha \cdot\big(\varphi\big)\big([x]\big)= \alpha \cdot\Phi\big(x\big)= \alpha \cdot \Big(T\big(\varphi\big)\Big)\big(x\big)$  
Conclude:  $T$ is linear    

$\big\Vert \Phi\big\Vert \leq \big\Vert \varphi \big\Vert_{_{(X\diagup M)^*}}$   
$\big\Vert \Phi\big\Vert =\sup \big\{\vert \Phi(x)\vert:\big\Vert x\big\Vert  \leq 1\big\}=\sup \big\{\vert \varphi([x])\vert:\big\Vert x\big\Vert   \leq 1\big\}\leq \sup \big\{\vert \varphi([x])\vert:\inf_{y\in M }\big\Vert x-y\big\Vert \leq 1\big\}= \big\Vert \varphi \big\Vert_{_{(X\diagup M)^*}}$  

$\big\Vert \varphi \big\Vert_{_{(X\diagup M)^*}}\leq \big\Vert \Phi\big\Vert $   
Consider sequence $[x_k]$ with norm $1$ such that $\big\vert \varphi[x_k]\big\vert \to \big\Vert \varphi\big\Vert_{(X\diagup M)^*}$.  This implies for any $\delta\gt 0$ that there is some $y_k \in M$ such that $1+\delta\geq \big\Vert x_k-y_k\big\Vert \geq \inf_{y\in M }\big\Vert x_k-y\big\Vert=\big\Vert [x_k]\big\Vert_{(X\diagup M)^*} =1$ and  
$\big\vert\Phi\big(x_k-y_k\big)\big\vert=\big\vert \varphi[x_k]\big\vert \to \big\Vert \varphi\big\Vert_{(X\diagup M)^*}\implies \big\Vert \Phi\big\Vert \geq \frac{1}{1+\delta}\cdot\big\Vert \varphi\big\Vert_{(X\diagup M)^*}\implies \big\Vert \Phi\big\Vert \geq \big\Vert \varphi\big\Vert_{(X\diagup M)^*}$  
The second implication holds since $\delta \gt 0$ was arbitrary  

The first implication holds since via re-scaling we have $\big\Vert \frac{1}{1+\delta}\big(x_k+y_k\big)\big\Vert\leq 1$ for all $k\in \mathbb N$ hence with $A:=\bigcup_{k=1}^\infty\big\{x_k+y_k\big\}$  
$\big\Vert \Phi \big\Vert =\sup\big\{\big\vert \Phi(x)\big\vert:\Vert x\Vert \leq 1\big\}\geq \sup\big\{\big\vert \Phi(v)\big\vert: v\in A\big\} \geq \frac{1}{1+\delta}\cdot\big\Vert \varphi\big\Vert_{(X\diagup M)^*}$ where $A$ is a subset of the closed unit ball and the supremum of a countable ordered set is at least as big as the $\limsup$ taken  on that countable ordered set which is equal to the limit of the sequence [since the limit exists]  

*re: surjectivity*  
Consider $\Phi \in M^\perp\subseteq X^*\implies$ there is some $\varphi\in (X\diagup M)^*$ given by $\varphi\big([x]\big)= \Phi\big(x+y\big)$ for any $y\in M$.  This is a well defined linear functional since e.g. $\varphi\big([x+x']\big)= \Phi\big(x + y+ x'+y'\big)= \Phi\big(x + y\big) +\Phi\big( x'+y'\big)=\varphi\big([x]\big)+\varphi\big([x']\big)$ where we have arbitrary $y,y'\in M$.  A similar argument holds for scaling.   

To confirm that $\varphi$ is continuous, hence actually $\in (X\diagup M)^*$ consider that if $\varphi$ were unbounded, then as in the above we'd have $[x_k]$ with length one and $\big\vert\varphi[x_k]\big\vert \to \infty$.  So for any $\delta \gt 0$ there is $y_k \in M$ such that $1+\delta\geq \big\Vert x_k-y_k\big\Vert \geq \inf_{y\in M }\big\Vert x_k-y\big\Vert=\big\Vert [x_k]\big\Vert_{(X\diagup M)^*} =1$   
$\implies (1+\delta)\cdot\big\Vert \Phi\big\Vert \geq \big\vert\Phi\big(x_k-y_k\big)\big\vert=\big\vert\varphi[x_k]\big\vert \to \infty\quad \text{(absurd)}$

**A chapter 5 type of approach to part (b.) [with some outside help]**  
$e_s\big(M_g f\big)=g(s)\cdot f(s)$ so let $\lambda = g(s)$  
$\implies \big\vert\lambda^k\cdot f(s)\big\vert=\big\vert e_s\big(M_g^k f\big)\big\vert\leq \big\Vert f\big\Vert \cdot \big\Vert e_s\big\Vert \cdot \big\Vert M_g^k\big\Vert$  

Since no $e_s$ is identically zero (ref page 18), select some $f$ such that $f(s)\neq 0$  
$\implies \big\vert\lambda\big\vert\leq \left(\frac{\big\Vert f\big\Vert \cdot \big\Vert e_s\big\Vert}{\vert f(s)\vert}\right)^\frac{1}{k} \cdot \big\Vert M_g^k\big\Vert^\frac{1}{k}$  
$\implies \big\vert g(s)\big\vert=\big\vert\lambda\big\vert\leq \lim_{k\to\infty }\big\Vert M_g^k\big\Vert^\frac{1}{k}=r\big(M_g\big)\quad \text{(Gelfand formula from chp 5 [Theorem 5.15])}$  
since $\left(\frac{\big\Vert f\big\Vert \cdot \big\Vert e_s\big\Vert}{\vert f(s)\vert}\right)^\frac{1}{k}\to 1$ ( taking the kth root of a positive constant)  

$\implies \sup\big\{\big\vert g(s)\big\vert :s\in S\big\}\leq r\big(M_g\big)$  

Thus to find an example where the inequality 
$\sup\big\{\big\vert g(s)\big\vert :s\in S\big\}\lt \big\Vert M_g\big\Vert$, it suffices to find a case where  

$r\big(M_g\big)\lt \big\Vert M_g\big\Vert$  

Awkwardly, page 18 gives a short list of functional Banach spaces that do not seem to have examples of strict inequality.  However $C^1[0,1]$ from ex 1.3 is a functional Banach space though it was not stated to be such (on page 18 where the examples were given by MacCluer).  We can confirm it is one since  
for $\Vert f\Vert =1$ we have  
$\big\vert e_s(f)\big\vert=\big\vert f(s)\big\vert \leq \big\Vert f\big\Vert_\infty \leq \big\Vert f\big\Vert_\infty  +\big\Vert  f'\big\Vert_\infty=\big\Vert f\big\Vert =1\implies \big\Vert e_s\big\Vert \leq 1$ so each $e_s$ is bounded.  


And for every $s\in [0,1]$ there is some $f\in C[0,1]$ such that $f(s)=e_s(f)\neq 0$ -- e.g. let $f$ be a polynomial, especially a simple one like $\mathbf 1$ or $x^k$ or $x^k-\mathbf 1$, thus no $e_s$ is identically $0$ and by the definition at the bottom of page 18 we deduce that $C[0,1]$ is a functional Banach space.  

E.g. let $C^1[0,1] \ni g(x)=x$ (the identity map).  


$\big\Vert g\big\Vert = 1 + 1=2$ and $\big\Vert \mathbf 1\big\Vert = 1+0=1\implies \big\Vert M_g\mathbf 1\big\Vert =\big\Vert g\big\Vert =2\implies 2\leq \big\Vert M_g\big\Vert$ and in fact this is met with equality since if $\big\Vert f\big\Vert =1$ we have  
$\big\Vert M_gf\big\Vert =\big\Vert g\cdot f\big\Vert=\big\Vert g\cdot f\big\Vert_\infty + \big\Vert (g\cdot f)'\big\Vert_\infty = \big\Vert g\cdot f\big\Vert_\infty + \big\Vert g'\cdot f + g\cdot f'\big\Vert_\infty$  
$\leq  \big\Vert g\cdot f\big\Vert_\infty + \big\Vert g'\cdot f\big\Vert_\infty  +\big\Vert g\cdot f'\big\Vert_\infty= \big\Vert g\cdot f\big\Vert_\infty + \big\Vert f\big\Vert_\infty  +\big\Vert g\cdot f'\big\Vert_\infty$   
$\leq \big\Vert g\big\Vert_\infty\big\Vert f\big\Vert_\infty + \big\Vert f\big\Vert_\infty  +\big\Vert g\big\Vert_\infty\big\Vert  f'\big\Vert_\infty=\big\Vert f\big\Vert_\infty + \big\Vert f\big\Vert_\infty  +\big\Vert  f'\big\Vert_\infty$   
$\leq 2\big\Vert f\big\Vert=2$  
$\implies \big\Vert M_g\big\Vert =2$  
In fact we can directly answer the question right here since the preceding shows $\big\Vert M_g\big\Vert =2$ but $\big\vert g(s)\big\vert \leq 1$ for all $s\in [0,1]$ hence $\sup\big\{\vert g(s)\vert :s\in[0,1]\big\}=1\lt 2= \big\Vert M_g\big\Vert$.  However the argument through the spectral radius is given below.  


Finally, suppose $\lambda \in \mathbb R\setminus \big[0,1\big]$ then  
$\big(M_g-\lambda \mathbf 1\big)f= \big(g-\lambda\big)\cdot f$  
and the multiplicative inverse, $\big(g-\lambda\big)^{-1}$ exists, noting that

**3.32**  
Suppose that $A$ is a linear map from a Hilbert space $\mathcal H$ into itself that satisfies  
$\big\langle x, Ay\big\rangle =\big\langle Ax, y\big\rangle$ for all $x,y \in \mathcal H$.  Show that $A$ is bounded.  

remark: a common approach is via Banach-Steinhaus (Uniform Boundedness).  I.e. using Riesz-Frechet and Theorem 3.15, we have set $S:=\big\{Th: \vert h\vert =1\big\}$ and for $\varphi \in H^*$ there is some $h\in \mathscr H$ such that for all $x \in S$  

$\big\vert\varphi\big(x\big)\big\vert=\big\vert\big\langle Th, y\big\rangle\big\vert =\big\vert\big\langle h, Ty\big\rangle\big\vert\leq \big\Vert h\big\Vert \big\Vert Ty\big\Vert =\big\Vert Ty\big\Vert $   
hence by Theorem 3.15 we know $S$ is bounded hence $T$ is bounded  


But, I prefer doing this instead instead with the Closed Graph Theorem, as follows.  In line with ex 3.21 it suffices to consider arbitrary $x_k\to \mathbf 0$ and $Ax_k \to b$ where we need to show $b=\mathbf 0$.  

Set $y:=b$ and this reads   
$\big\langle Ax_k ,b\big\rangle = \big\langle x_k ,Ab\big\rangle$ for all $k$ and using continuity of the inner product this yields   
$\big\Vert b\big\Vert^2=\big\langle b ,b\big\rangle  = \big\langle \lim_{k\to\infty} Ax_k ,b\big\rangle =\lim_{k\to\infty}\big\langle Ax_k ,b\big\rangle = \lim_{k\to\infty}\big\langle x_k ,Ab\big\rangle=\big\langle \lim_{k\to\infty} x_k ,Ab\big\rangle=\big\langle \mathbf 0 ,Ab\big\rangle=0$ $\implies b=\mathbf 0$  



**encore:**  
A similar problem is to suppose $T:\mathcal H\rightarrow \mathcal H$ where $\mathcal H$ is a *real* Hilbert space, and suppose $\big\langle x,Tx\big\rangle \geq 0$ for all $x$.  Argue that this implies $T$ is continuous.  

Note: that when working over $\mathbb C$ then $\big\langle x,Tx\big\rangle \in \mathbb R\implies \big\langle Tx,y\big\rangle=\big\langle x,Ty\big\rangle $ for $x,y \in \mathcal H$ but per e.g. https://math.stackexchange.com/questions/561636/show-that-a-positive-operator-on-a-complex-hilbert-space-is-self-adjoint/ so we'd get $T$ is continuous by the preceding problem but things are different in $\mathbb R$ so we need a different approach for this encore.  

Note that $\big\langle x,y\big\rangle_c:= \big\langle x,Ty\big\rangle$ gives a real positive semi-definite bilinear form though it is not a semi-inner product per se since it isn't known to be symmetric.  Using ex 3.21 it suffices to show that if arbitrary $x_k\to \mathbf 0$ and $Tx_k \to b$ then $b=\mathbf 0$.  The way forward involves symmetrization to get a variant of Cauchy-Schwarz then applying that. 

$C:=\displaystyle \left[\begin{matrix}\big\langle b,b\big\rangle_c & \big\langle x_k, b\big\rangle_c\\ \big\langle b,x_k\big\rangle_c& \big\langle x_k,x_k\big\rangle_c\end{matrix}\right]$   
The fact that $\big\langle \bullet, \bullet \big\rangle_c$ is a real positive semi-definite bilinear form implies for any $\mathbf v\in \mathbb R^2$ that $\mathbf v^T C \mathbf v =\big\langle (v_1\cdot b+v_2\cdot x_k),(v_1\cdot b+v_2\cdot x_k)\big\rangle_c  \geq 0$  

$G:= \Big(\frac{1}{2}\big(C+C^T\big)\Big)$ which also implies $G\succeq \mathbf 0$   
$\implies\big\langle b,b\big\rangle_c\cdot \big\langle x_k,x_k\big\rangle_c - \Big(\frac{1}{2}\big\langle b,x_k\big\rangle_c + \big\langle x_k,b\big\rangle_c\big)\Big)^2 =\det\big(G\big) \geq 0 $  
which simplifies to  
$\left\vert\big\langle b,T x_k\big\rangle + \big\langle x_k,Tb\big\rangle\right\vert = \left\vert\big\langle b,x_k\big\rangle_c + \big\langle x_k,b\big\rangle_c\right\vert\leq 2\cdot\big\langle b,b\big\rangle_c^\frac{1}{2}\cdot \big\langle x_k,x_k\big\rangle_c^\frac{1}{2}=2\cdot\big\langle b,Tb\big\rangle^\frac{1}{2}\cdot \big\langle x_k,Tx_k\big\rangle^\frac{1}{2} $  
and taking limits yields  
$\big\Vert b\big\Vert^2= \lim_{k\to\infty} \left\vert\big\langle b,T x_k\big\rangle + \big\langle x_k,Tb\big\rangle\right\vert \leq \lim_{k\to\infty} 2\cdot\big\langle b,Tb\big\rangle^\frac{1}{2}\cdot\big\langle x_k,Tx_k\big\rangle^\frac{1}{2}=2\cdot\big\langle b,Tb\big\rangle^\frac{1}{2}\cdot\big\langle \lim_{k\to\infty} x_k, \lim_{k\to\infty} Tx_k\big\rangle^\frac{1}{2}=0 $  
since $\big\langle \lim_{k\to\infty} x_k, \lim_{k\to\infty} Tx_k\big\rangle^\frac{1}{2}=\big\langle 0, b\big\rangle^\frac{1}{2}=0$ 

Conclude: $ b=\mathbf 0$ and $T$ is continuous.  

*extension:*  
This argument immediately generalizes to prove the Banach space equivalent claim of:  
For a real Banach space $X$, let $T:X\rightarrow X^*$ observe $(Tx)(x)\geq 0$ for all $x\in X$ which implies $T$ is bounded.  

Using the notation from pages 63-64 of Eidelman et al, we have for $x,y\in X$  
$\big\langle x,y\big\rangle_c:= \big\langle x,Ty\big\rangle=(Ty)(x)$  
which is a positive semi-definite bilinear form on $X$, and by Closed Graph Theorem if $x_k\to 0$ and $\lim_{k\to\infty }Tx_k = f\in X^*$ then we need to show $f=0$.  Selecting $b\in X$ such that $ f(b)=\Vert f\Vert $ [recall Eidelman ex 3.3.9] the argument runs as    

$G=\frac{1}{2}\big(C+C^T\big)\succeq \mathbf 0$ as before and    
$\big\Vert f\big\Vert =f\big(b\big)= \lim_{k\to\infty} \left\vert\big\langle b,T x_k\big\rangle + \big\langle x_k,Tb\big\rangle\right\vert \leq \lim_{k\to\infty} 2\cdot\left\vert\big\langle b,Tb\big\rangle\right\vert^\frac{1}{2}\cdot\left\vert\big\langle x_k,Tx_k\big\rangle\right\vert^\frac{1}{2}$  
where $\lim_{k\to\infty}\big\langle x_k,Tx_k\big\rangle=0$ since $f_k:=Tx_k$  
for any $\epsilon \gt 0$ [say $\epsilon:=1$] and all $k\geq K$ we have  
$\Vert f_k-f\Vert \leq \epsilon\implies \Vert f_k(x_k)-f(x_k)\Vert=\Vert (f_k-f)(x_k)\Vert \leq \epsilon\cdot\Vert x_k\Vert$ and taking limits sends the upper bound to zero (since $x_k\to 0$ by assumption)  
$0=\lim_{k\to\infty}f_k(x_k)-\lim_{k\to\infty} f(x_k)=\lim_{k\to\infty}f_k(x_k)-f(0)=\lim_{k\to\infty}f_k(x_k)$.  

Similarly $\big\langle b,T x_k\big\rangle =f_k\big(b\big)$ and $\big\Vert f_k(b)-f(b)\big\Vert\leq \big\Vert f_k-f\big\Vert\cdot \big\Vert b\big\Vert\to 0\implies \lim_{k\to\infty}f_k\big(b\big)=f\big(b\big)$  

**3.33**  
Prove if $X$ is Banach and $T\in \mathscr B(X)$ is such that the quotient vector space $X\diagup TX$ is finite dimensional, then $\text{image }T$ is closed. 

There are 3 parts to the official problem as it is set up as a guided proof.  

note: this is also ex 2.27 in Brezis, though that exercise does not have hints, it does have a solution in the 'solutions' section.  It does not, however factor through part (a.) of getting to an injective map first so the approach in MacCluer seems a bit nicer overall.   


**(a.)**  
Per ex 3.26  
$A:X\diagup \ker T\rightarrow Y$  given by $A\big(x+ \ker T\big) =Tx$ where ($x\in X$ is an arbitrary representative of an equivalence class) is a well-defined, injective and bounded linear map where $\text{image }A=\text{image }T$ since the above equality shows that each point in the image of $A$ is given by $Tx$ for arbitrary $x\in X$ and conversely for $y\in \text{image }T$ we have $y=Tx = A\big(x+ \ker T\big)=A[x]\in \text{image }A$.  Thus the argument reduces to the case where assuming **WLOG** that $T$ is injective -- if not just run the argument on $A$ instead.  

**(b.)**  
Suppose $\dim X\diagup TX=1$, which means there is some $y\in X$ such that $X = \big\{Tx + \alpha \cdot y: x\in X, \alpha \in \mathbb C\big\}$.  The map $S:X\times \mathbb C\longrightarrow X$ given by $S\big(x,\alpha\big)=Tx +\alpha\cdot  y$ is bounded below and thus has closed image (ex 2.12).  

Using the 1-norm from  ex 3.19 (a) we know that $X\times \mathbb C$ is complete. 

Note that for $\big\Vert x, \alpha\big\Vert_1 =1\implies \max\big(\big\Vert x\big\Vert, \big\vert \alpha\big\vert\big) \leq 1 $ 
$\implies \big\Vert S\big(x, \alpha\big)\big\Vert_1= \big\Vert Tx\big\Vert + \big\vert \alpha\big\vert\leq \big\Vert T\big\Vert + 1 \implies S \text{ is bounded}$  

Now by inspection $S$ is linear and injective on $(x, \alpha)$ [part (a.) is key here] and $S$ is surjective since, as we observed $X = \big\{Tx + \alpha \cdot y: x\in X. \alpha \in \mathbb C\big\}$.  Open Mapping Theorem / Corollary 3.17 applies and $S^{-1}\in B(X)$ exists hence for $\big\Vert x_k\big\Vert =1$ we have $\big\Vert Tx_k\big\Vert =\big\Vert Sx_k\big\Vert \geq \delta \gt 0$ for some small enough $\delta\gt 0$ (otherwise $S^{-1}$ sends points arbitrarily close to $0$ to the boundary of the unit ball, implying $S^{-1}$ is unbounded-- this is closely related to ex 3.27).  Conclude $\text{image }T$ is closed per the argument in ex 2.12.    


**(c.)**  
Prove the full result:    
Let $\dim X\diagup TX=n$ with basis $\big\{y_1,\dotso, y_n\big\}$ which means that $X = \big\{Tx + \sum_{k=1}^n\alpha_k \cdot y_k: x\in X, \alpha_k \in \mathbb C\big\}$ 

The map $S:X\times \underbrace{\mathbb C\times \dotso \times \mathbb C}_{\mathbb C \text{ occurs n-times}}\longrightarrow X$ given by $S\big(x,\alpha_1, \dotso, \alpha_n\big)=Tx +\sum_{k=1}^n\alpha_k\cdot  y_k$ is linear and bijective by the same observations as before. The domain $X\times \underbrace{\mathbb C\times \dotso \times \mathbb C}_{\mathbb C \text{ occurs n-times}}$, equipped with the 1-norm, is complete since $\big(X\times \underbrace{\mathbb C\times \dotso \times \mathbb C}_{\mathbb C \text{ occurs n-1-times}}\big)\times \mathbb C$ is the product of two complete spaces per ex 3.19(a) [formally we know $X\times \underbrace{\mathbb C\times \dotso \times \mathbb C}_{\mathbb C \text{ occurs n-1-times}}$ is complete via an induction hypothesis].  And of course $S$ is bounded since  
$\big\Vert \big(x, \alpha_1,\dotso, \alpha_n\big)\big\Vert_1=1\implies \max\big(\big\Vert x\big\Vert, \big\vert \alpha_1\big\vert,\dotso, \big\vert\alpha_n\big\vert\big)\leq 1$  
$\implies\big\Vert S\big(x, \alpha_1,\dotso, \alpha_n\big)\big\Vert= \big\Vert Tx\big\Vert +\sum_{k=1}^n \big\vert \alpha_k\big\vert\leq \big\Vert T\big\Vert + \sum_{k=1}^n 1 =\big\Vert T\big\Vert +n\implies S \text{ is bounded}$  

Finally, as before, Corollary 3.17 then tells us that $S^{-1}\in B(X)$ exists hence for $\big\Vert x_k\big\Vert =1$ we have $\big\Vert Tx_k\big\Vert =\big\Vert Sx_k\big\Vert \geq \delta \gt 0$ (otherwise $S^{-1}$ is unbounded).  Conclude $\text{image }T$ is closed per the argument in ex 2.12.     


The **Brezis** [ex 2.27] problem statement is similar albeit without the hints/guide.  Let $E,F$ be Banach and $T\in B(X,Y)$.  Assume $\text{image }T$ has finite co-dimension, i.e. there is a finite dimensional subspace $W$ such that $W+\text{image }T=Y$ and $W\cap \text{image }T=\big\{\mathbf 0\big\}$.  Prove that $\text{image }T$ is closed.  

The solution is let $V=X\times W$ [evidently with the '1 norm' though this was not stated] and consider the map  

$S:G\rightarrow Y$  given by $S\big(x, w\big) = Tx + w$   
$S$ is linear and surjective so by open mapping theorem we have   
$S\Big(X \times  \big\{\mathbf 0\big\}^C\Big) =S\Big(X \times \big(W-\setminus \big\{\mathbf 0\big\}\big)\Big)=\text{image }T+ \big\{\mathbf 0\big\}^C$  
is open since it is the image of an open set (since the product of two open sets is open) under an open map,  Thus its complement i.e.  $\Big(\text{image }T+ \big\{\mathbf 0\big\}^C\Big)^C=\text{image }T + \big\{\mathbf 0\big\}=\text{image }T$ is closed.  

*remark:*  
This argument leans on topology a little more than the MacCluer argument and is more direct in that it does not first fact through the quotient map.  However the MacCluer argument (a.) by using the quotient map is a nice exercise in using ex 3.26 and (b.) hammers on whether a map is bounded below or not, which turns out to be an ongoing theme through the rest of the book [e.g. for normal operators / elements], so pedagogically speaking the MacCluer approach pays dividends down the line in a way that the Brezis solution wouldn't.

**3.36**  
A sequence $\big\{h_k\big\}\in \mathscr H$ is said to be a *Bessel sequence* if for every $h \in \mathscr H$   
$\sum_{n=1}^\infty\big\vert \langle h,h_n\rangle\big\vert^2 \lt \infty$  

A sequence $\big\{g_n\big\}$ is said to be a *Riesz-Fischer sequence* if given any $\begin{bmatrix}c_1 \\ c_2  \\ c_3\\\vdots \end{bmatrix}\in \mathscr l^2$ there exists (at least one) vector $g \in \mathscr H$ such that  
$\big\langle g,g_n\big\rangle =c_n$ for all $n$  

Note that an orthonormal basis is both a Bessel sequence and a Riesz-Fischer sequence.  

**(a.)**  
Show that if $\big\{h_k\big\}$ is a Bessel sequence, then there exists $M\lt \infty$ so that  
$\sum_{n=1}^\infty\big\vert \langle h,h_n\rangle\big\vert^2 \leq M\cdot \big\Vert h\big\Vert^2$  for all $h\in \mathscr H$.  
Hint: apply the closed graph theorem to the map $S:\mathscr H\rightarrow \mathscr l^2$ defined by $Sx=\begin{bmatrix}\langle x,h_1\rangle \\ \langle x,h_2\rangle  \\ \langle x,h_3\rangle\\\vdots \end{bmatrix}$ 

In line with ex 3.21 it suffices to consider $x_k\to \mathbf 0$ and $Sx_k\to y$ and we need to show $y=\mathbf 0$. For (orthonormal) standard basis vector $e_n\in\mathscr l^2$  

$\big\langle x_k,h_n\big\rangle=\big\langle Sx_k,e_n\big\rangle\implies 0=\big\langle \mathbf 0,h_n\big\rangle=\big\langle \lim_{k\to\infty} x_k,h_n\big\rangle= \lim_{k\to\infty}\big\langle x_k,h_n\big\rangle=\lim_{k\to\infty}\big\langle Sx_k,e_n\big\rangle=\big\langle \lim_{k\to\infty} Sx_k,e_n\big\rangle=\big\langle y,e_n\big\rangle$  
$\implies y=\mathbf 0$ either by inspecting the coordinates, or Theorem 1.33(b).  Conclude $S$ is continuous, hence bounded and   
$ \sum_{n=1}^\infty\big\vert \langle h,h_n\rangle\big\vert^2=\big\Vert Sh\big\Vert^2 \leq \big\Vert S\big\Vert^2\cdot \big\Vert h\big\Vert^2$  
so set $M:=\big\Vert S\big\Vert^2$  

**additional work for use in ex 3.37**  
Now consider the adjoint   
$S^*:\mathscr l^2\rightarrow \mathscr H$, and again with $Sx=\begin{bmatrix}\langle x,h_1\rangle \\ \langle x,h_2\rangle  \\ \langle x,h_3\rangle\\\vdots \end{bmatrix}$  we have    
$\big\langle x,h_n\big\rangle=\big\langle Sx,e_n\big\rangle=\big\langle x,S^* e_n\big\rangle\implies S^* e_n = h_n$ since $0=\big\langle x,\big(h_n- S^* e_n\big)\big\rangle$ for all $x \in \mathscr H$.  Now using the fact that $\big\{e_n\big\}$ form an orthonormal basis for $\mathscr l^2$ then $v=\sum_{n=1}^\infty\alpha_n\cdot e_n$  

$\implies S^*v=S^*\sum_{n=1}^\infty\alpha_n\cdot e_n=\sum_{n=1}^\infty\alpha_n\cdot S^*e_n=\sum_{n=1}^\infty\alpha_n\cdot h_n$  
where the 2nd to last equality is justified by continuity (and linearity) of the adjoint   
note: $S^*$ is the infinite dimensional version of $B$ from ex 2.20  

Finally consider $S^*S:\mathscr H\rightarrow \mathscr H$  
$S^*Sx= S^*\begin{bmatrix}\langle x,h_1\rangle \\ \langle x,h_2\rangle  \\ \langle x,h_3\rangle\\\vdots \end{bmatrix}=S^*\sum_{n=1}^\infty\langle x,h_n\rangle\cdot e_n=\sum_{n=1}^\infty\langle x,h_n\rangle\cdot S^* e_n=\sum_{n=1}^\infty\langle x,h_n\rangle\cdot h_n$  

$\implies S^*S=T\quad \text{ (in ex 3.37 a)}$  
So that $T$ has norm $\big\Vert S^*S\big\Vert =\big\Vert S\big\Vert^2$  (Prop 2.14) hence it is bounded.  
[Note that Prop 2.14 was written for $T\in B(\mathscr H)$ but as I commented in ex 2.8(b), aside from notation, it runs just as well for $T\in B(\mathscr H,\mathscr H')$]

**(b.)**  
Show that if $\big\{g_n\big\}$ is a *Riesz-Fischer sequence*, then there exists $m\gt 0$ such that given $\big\{c_n\big\}\in \mathscr l^2$, the equations in (3.5) hold for at least one solution $g$ satisfying $m\Vert g\Vert^2\leq \sum_{n=1}^\infty \vert c_n\vert^2$    

Hint: the closed graph theorem again, applied to the appropriate map  
$T:\mathscr l^2\rightarrow \mathscr H\diagup N$  

where $N:=\big( \overline{\text{span}\big\{g_1, g_2, g_3,\dotso,\big\}}\big)^\perp$  

Let $W:=\overline{\text{span}\big\{g_1, g_2,\dotso\big\}}$ and $\mathscr H =W\oplus W^\perp$ 
with orthogonal projection [topological form] $P:\mathscr H\longrightarrow W$  

for each $v\in \mathscr l^2$ we have some $g^{(v)}$ such that  
$v=\begin{bmatrix}c_1^{(v)}\\ c_2^{(v)}  \\ c_3^{(v)}\\\vdots \end{bmatrix}=\begin{bmatrix}\big\langle g^{(v)},g_1\big\rangle\\ \big\langle g^{(v)},g_2\big\rangle \\\big\langle g^{(v)},g_3\big\rangle\\\vdots \end{bmatrix} =\begin{bmatrix}\big\langle Pg^{(v)},g_1\big\rangle\\ \big\langle Pg^{(v)},g_2\big\rangle \\\big\langle Pg^{(v)},g_3\big\rangle\\\vdots \end{bmatrix}$

Define $G^{(v)}:=Pg^{(v)}$ which is the unique element in $x\in W$ that satisfies $\big\langle x,g_k\big\rangle =c_k^{(v)}$ for all $k$--i.e. if there way another element $F^{(v)}$ then we'd have 
$\big\langle (G^{(v)}- F^{(v)}),g_k\big\rangle=0$ for all $k$ hence $(G^{(v)}- F^{(v)})\in W^\perp$ and $(G^{(v)}- F^{(v)})=P\big( g^{(v)}-f^{(v)}\big)\in W$, i.e. $(G^{(v)}- F^{(v)})\in W\cap W^\perp\implies (G^{(v)}- F^{(v)})=\mathbf 0$.  

Now define the mapping $T:\mathscr l^2\longrightarrow W$ 
note: since we are dealing with Hilbert spaces, recall ex 3.26(c) which implies $ \mathscr H\diagup N\cong N^\perp =W$ so this co-domain is isometrically isomorphic to the one suggested in the hint  

$Tv=T\begin{bmatrix}c_1^{(v)}\\ c_2^{(v)}  \\ c_3^{(v)}\\\vdots \end{bmatrix}=T\begin{bmatrix}\big\langle G^{(v)},g_1\big\rangle\\ \big\langle G^{(v)},g_2\big\rangle \\\big\langle G^{(v)},g_3\big\rangle\\\vdots \end{bmatrix}=G^{(v)}$  
Linearity  is immediate [since the inner product is linear in its 1st argument].  Both spaces are Banach [in fact Hilbert] so we can argue that $T$ is continuous by Closed Graph Theorem-- and in particular the formulation in ex 3.21.  

Suppose $v_m\to \mathbf 0$ and $G^{(v_m)} =Tv_m\to y$; we need to show this implies $y=\mathbf 0$. (Optional: since $v_m\to \mathbf 0$ [in norm] this implies $v_m\to_w \mathbf 0$ [weakly; ref ex 1.28(b)].)  
$\big\langle y,g_n\big\rangle=\big\langle \lim_{m\to\infty} G^{(v_m)},g_n\big\rangle=\lim_{m\to\infty}\big\langle G^{(v_m)},g_n\big\rangle = \lim_{m\to\infty}\big\langle v_m,e_n\big\rangle=\big\langle \lim_{m\to\infty} v_m,e_n\big\rangle =\big\langle\mathbf 0,e_n\big\rangle = 0$  

Note the middle equality $\big\langle G^{(v_m)},g_n\big\rangle = \big\langle v_m,e_n\big\rangle$ comes from looking at the $n$th component of $v_m$.  More formally: where $e_n$ are the standard basis vectors for $\mathscr l^2$ [which form an orthonormal basis for $\mathscr l^2]$ then for $v\in \mathscr l^2$  we have  $v=\begin{bmatrix}\big\langle G^{(v)},g_1\big\rangle\\ \big\langle G^{(v)},g_2\big\rangle \\\big\langle G^{(v)},g_3\big\rangle\\\vdots \end{bmatrix} =\sum_{n=1}^\infty \big\langle v,e_n\big\rangle\cdot e_n=\sum_{n=1}^\infty \big\langle G^{(v)},g_n\big\rangle\cdot e_n $  
and taking $\big\langle \bullet, e_j\big\rangle $ gives the equality $\big\langle v,e_j\big\rangle=\big\langle G^{(v)},g_j\big\rangle $   

Now $y\in W$, since $W$ is closed and since $\text{span}\big\{g_1, g_2, g_3,\dotso,\big\}$ is dense in $W$ this means $y$ is orthogonal to every element in $W$, i.e. $y\in W^\perp$.  Conclude $y \in W\cap W^\perp\implies y=\mathbf 0$ as required.  

Thus for each $v\in \mathscr l^2$ we have  
$\big\Vert Pg^{(v)}\big\Vert =\big\Vert G^{(v)}\big\Vert=\big\Vert Tv\big\Vert\leq \big\Vert T\big\Vert\cdot \big\Vert v \big\Vert=\big\Vert T\big\Vert\cdot \sqrt{\sum_{n=1}^\infty \vert c_n^{(v)}\vert^2}$ 
and dividing each side by $\big\Vert T\big\Vert$ and then squaring each side gives the result.


**3.37**  
A sequence of distinct vectors $\big\{h_n\big\}\in \mathscr H$ for separable $\mathscr H$ is called a **frame** if there exist finite positive constants $M_1$ and $M_2$ with  

**note:** separability seems to be used in situations involving frames but doesn't seem necessary for this problem  

$M_1\Vert h\Vert^2\leq \sum_{n=1}^\infty\big\vert \langle h,h_n\rangle\big\vert^2 \leq M_2\Vert h\Vert^2 $  
for all $h\in \mathscr H$.  Observe that if $\big\{h_n\big\}$ is a frame then $\big\{h_n\big\}$ is a Bessel sequence [ref ex 3.36] and that whenever $\big\{h_n\big\}$ is a Bessel sequence, the 2nd inequality in this definition must hold.

**(a.)** Suppose $\big\{h_n\big\}$ is a frame and define $T$ by  
$Th=\sum_{n=1}^\infty \big\langle h,h_n\big\rangle \cdot h_n$  
Show that $T\in B(\mathscr H)$

$\underline{\text{A different argument}}$  
If we did not do ex 3.36, then the tricky thing is showing that the operator is well defined, i.e. that the series for the operator converges.  To this end we show the sequence of partial sums is Cauchy.  Select $d\in \mathbb N$ and for $m\gt d$, set $v_m:= \frac{\sum_{n=d+1}^m \big\langle h,h_n\big\rangle \cdot h_n   }{\big\Vert\sum_{n=d+1}^m \big\langle h,h_n\big\rangle \cdot h_n \big\Vert}$  [and if the numerator is zero then just set $v_m$ equal to some arbitrary unit length vector]. As an aside while using $v_m$ is very simple and direct, it *could* be interpreted through Hahn-Banach Corollary 3.6 (with $M=\big\{\mathbf 0\big\}$, ignoring the case numerator of $v_m$ being zero) so there is some $\varphi \in \mathscr H^*$ where $\big\Vert\varphi\big\Vert=1$ and $\varphi\big(\sum_{n=d+1}^m \big\langle h,h_n\big\rangle \cdot h_n \big)= \text{dist }\big(\sum_{n=d+1}^m \big\langle h,h_n\big\rangle \cdot h_n , M\big)=\big\Vert \sum_{n=d+1}^m \big\langle h,h_n\big\rangle \cdot h_n \big\Vert $, and of course via Riesz-Frechet we know we can find a representer for this linear functional, which is given by $v_m$.    


$\big\Vert\sum_{n=1}^m \big\langle h,h_n\big\rangle \cdot h_n - \sum_{n=1}^d \big\langle h,h_n\big\rangle \cdot h_n\big\Vert = \big\Vert\sum_{n=d+1}^m \big\langle h,h_n\big\rangle \cdot h_n \big\Vert = \big\langle \sum_{n=d+1}^m \big\langle h,h_n\big\rangle \cdot h_n, v_m \big\rangle =  \sum_{n=d+1}^m \big\langle h,h_n\big\rangle \cdot\big\langle  h_n, v_m \big\rangle$  
$\leq \sum_{n=d+1}^\infty \big\vert \big\langle h,h_n\big\rangle\big\vert \cdot\big\vert\big\langle  h_n, v_m \big\rangle\big\vert$  
$\leq \Big(\sum_{n=d+1}^\infty \big\vert \big\langle h,h_n\big\rangle\big\vert^2\Big)^\frac{1}{2} \cdot \Big(\sum_{n=d+1}^\infty\big\vert \big\langle  h_n, v_m \big\rangle\big\vert^2\Big)^\frac{1}{2}$  
$\leq \Big(\sum_{n=d+1}^\infty \big\vert \big\langle h,h_n\big\rangle\big\vert^2\Big)^\frac{1}{2} \cdot \Big(\sum_{n=1}^\infty\big\vert \big\langle  v_m, h_n \big\rangle\big\vert^2\Big)^\frac{1}{2}$  
$\leq  \Big(\sum_{n=d+1}^\infty \big\vert \big\langle h,h_n\big\rangle\big\vert^2\Big)^\frac{1}{2} \cdot  \sqrt M_2 $  

The inequalities are triangle inequality [and non-negativity], Cauchy-Schwarz, non-negativity, and the frame definition combined with the fact that $\Vert v_m\Vert =1$.  Note that for any $\epsilon \gt 0$ then $\sum_{n=d+1}^\infty \big\vert \big\langle h,h_n\big\rangle\big\vert^2\lt \frac{\epsilon^2}{M_2}$ for any $d$ large enough since $\sum_{n=1}^\infty \big\vert \big\langle h,h_n\big\rangle\big\vert^2$ converges absolutely.  This confirms that the partial sums form a Cauchy sequence and since $\mathscr H$ is complete, they converge.  
 
We can confirm that $T$ is bounded by applying ex 3.32.  I.e. for arbitrary $v,h\in \mathscr H$ we have  
$\big\langle Th,v\big\rangle = \big\langle \sum_{n=1}^\infty \langle h, h_n\rangle \cdot h_n,v\big\rangle=\sum_{n=1}^\infty \big\langle h, h_n\big\rangle \cdot\big\langle  h_n,v\big\rangle=\sum_{n=1}^\infty \big\langle  h_n,v\big\rangle\cdot \big\langle h, h_n\big\rangle$ and   
$\big\langle h,Tv\big\rangle = \big\langle h,  \sum_{n=1}^\infty \langle v, h_n\rangle \cdot h_n\big\rangle= \sum_{n=1}^\infty \overline{\big\langle v, h_n\big\rangle} \cdot\big\langle h,  h_n\big\rangle=\sum_{n=1}^\infty \big\langle h_n,v\big\rangle \cdot\big\langle h,  h_n\big\rangle$  
$\implies \big\langle Th,v\big\rangle=\big\langle h,Tv\big\rangle$ for arbitrary $v,h\in \mathscr H$ hence $T$ is bounded.  Note since $T$ is bounded we have for arbitrary $v,h \in\mathscr H$  
$\big\langle h,T^*v\big\rangle =\big\langle Th,v\big\rangle=\big\langle h,Tv\big\rangle\implies T=T^*$ by uniqueness of the adjoint [Theorem 2.12], which answers part (c.).

$\big\Vert \sqrt{T_m-T_d}\big\Vert$ can be easily bounded since $\big\Vert T_m-T_d\big\Vert \leq 2M_2$ so $\big\Vert \sqrt{T_m-T_d}\big\Vert\leq \sqrt{2M_2}$ (Prop 2.14) for any $m,d$.  

$\implies\big\Vert (T_m-T_d)h\big\Vert^2=\big\Vert \sqrt{T_m-T_d}\sqrt{T_m-T_d}h\big\Vert^2\leq 2M_2\cdot \big\Vert \sqrt{T_m-T_d}h\big\Vert^2 = 2 M_2\cdot \big\langle (T_m-T_d)h,h\big\rangle\to 0$ as $d\to \infty$   
Conclude: $\big\{T_dh\big\}$ *is a Cauchy sequence* thus $\lim_{d\to\infty} T_d h$ exists for each $h\in \mathscr H$ by completeness.  Per Banach-Steinhaus (Theorem 3.12) this implies bounded linear map $T$ given by $Th=\lim_{d\to\infty} T_d h= \lim_{d\to\infty}\sum_{n=1}^d \big\langle h,h_n\big\rangle \cdot h_n$  which is the operator defined in this problem.    


$\underline{\text{notes on the square-root}}$  
The square-root can of course be interpreted in the more general sense for bounded positive operators on a Hilbert space (or the chapter 4 sense for compact operators).  However since each $T_d$ is self-adjoint and of finite rank we can actually use results from finite dimensional linear algebra to justify the square-root.   

Using ex 2.16 (and  ex 1.12), write   
$\mathscr H= \ker (T_m-T_d)\oplus (\ker(T_m-T_d))^\perp= \ker (T_m-T_d)\oplus \overline{\text{image }(T_m-T_d)^*}= \ker (T_m-T_d)\oplus \overline{\text{image }(T_m-T_d)}$  $= \ker (T_m-T_d)\oplus \text{image }(T_m-T_d)$   

where $(T_m-T_d)^*=(T_m-T_d)$, i.e. it is self-adjoint, and we know $ \text{image }(T_m-T_d)$ is closed by ex 1.21 (b) [run Gram schdmit on $\text{image }(T_m-T_d)$ to get an orthnormal basis for it and apply the exercise] or the corollary to ex 3.8.   

Each of those are $(T_m-T_d)$-invariant subspaces and this means $(T_m-T_d)\vert_{\ker (T_m-T_d)}=\mathbf 0$ which has a trivial square root and $T\vert_{\text{image }(T_m-T_d)} = A$ where $A$ is a positive semi-definite (in fact definite) Hermitian operator from a finite dimensional subspace to itself hence $\sqrt A\succeq \mathbf 0$ exists from finite dimensional linear algebra and thus specifies a square-root for $(T_m-T_d)$ on $\mathscr H$.   


*Alternatively:* at the cost of more notation, we can completely bypass the use of the square-root and write $(T_m- T_d)=B_{m,d}B_{m,d}^*$ in the above argument, though it is a bit clunkier.  

*Yet another approach* is to toggle the representations $BB^*$ vs $B^*B$ from ex 2.20, where the latter is a Gram Matrix.   Then with $m\gt d$, for fixed $h\in \mathscr H$  
$(T_m-T_d) h=\sum_{n=d+1}^m \big\langle h,h_n\big\rangle \cdot h_n$  
$\implies \big\Vert (T_m-T_d) h\big\Vert^2 =\sum_{n=d+1}^m \sum_{j=d+1}^m  \overline {\big\langle h,h_j\big\rangle} \cdot \big\langle  h_n,  h_j\big\rangle\cdot\big\langle h,h_n\big\rangle =\mathbf x_{m,d}^*G_{d,m}\mathbf x_{m,d}$  
$\implies \big\Vert (T_m-T_d) h\big\Vert^2\leq \big\Vert G_{d,m}\big\Vert \cdot \big\Vert\mathbf x_{m,d}\big\Vert^2\leq M_2\cdot \big\Vert\mathbf x_{m,d}\big\Vert^2\to 0 \text{ as d}\to\infty$  

Noting that here $G_{d,m}$ is a principal submatrix of $G_m$ hence $\big\Vert G_{d,m}\big\Vert \leq \big\Vert G_m\big\Vert=  \big\Vert B_m^*B_m \big\Vert^2=\big\Vert B_m^*\big\Vert^2=\big\Vert B_m\big\Vert^2 =\big\Vert B_m B_m^*\big\Vert= \big\Vert T_m\big\Vert  \leq M_2$  
where the 1st inequality holds by Cauchy Interlacing (or constrained quadratic form argument) and the middle equalities are Prop 2.14.  And of course $\big\Vert x_{m,d}\big\Vert^2=  \sum_{n=d+1}^m \big\vert \big\langle h,h_n\big\rangle \big\vert^2\to 0\text{ as d}\to\infty$  since $\sum_{n=d+1}^m \big\vert \big\langle h,h_n\big\rangle \big\vert^2$ converges

The 2nd equality is ex 2.16, the 3rd equality is that $T$ is self-adjoint and the final equality is ex 1.12.  Thus $\text{image T}$ is dense in $\mathscr H$ so $T$ is invertible.

The 2nd equality is because $B$ is continuous and the 3rd equality is because $\big(BA\big)$ is continuous.  The implication that $By=\mathbf 0\implies y=\mathbf 0$ holds because $B$ is injective.
