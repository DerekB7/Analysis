For those familiar with complex analysis, probably the expedient route here is via holomorphic functional calculus, though some care is needed since we need to consider boundary behavior. Working over a **complex Hilbert** space:   

Typically for an analytic function applied to operator $B$ we'd write   
$$f(z) = \frac{1}{2\pi}\int_{\Gamma}f(\lambda)(\lambda I-B)^{-1}d\lambda$$  
where $\Gamma$ is a cycle with winding number $n\big(\Gamma, w\big)=1$ for $w\in \sigma(B)$ and $n\big(\Gamma, w\big)=0$ for $w\not \in\text{domain  of f}$.  The issue is that $B$ need not be invertible, i.e. it is possible that $0\in\sigma(B)$, and $0$ is not in the interior of any continuous square root function (e.g. that would imply a winding number of $\frac{1}{2}$ for a closed curve around zero, but winding numbers of closed curves must be integer valued.).  

We can **instead work through a power series** expansion of an analytic square root and navigate subtleties regarding its boundary behavior.  It is convenient to work with the classical binomial/power series   
$f(z)=(1+z)^\frac{1}{2}=1 + \sum_{k=1}^\infty c_k z^k$ and its polynomial approximations $f_m(z):=1 + \sum_{k=1}^m c_k z^k$ where  
$c_1=\frac{1}{2}$ and $c_k =(-1)^{k+1}\frac 1{2k}\cdot\frac{1\cdot 3\cdot5\dots (2k-3)}{2\cdot 4\cdot6\dots (2k-2)}$  
The *key insight* is that the $c_k$ are real with an alternating sign such that $c_k\lt 0$ for even $k$ and $c_k\gt 0$ for odd $k$. We need to show that $f$ converges absolutely for $\vert z\vert =1$, i.e. that $f(-1)=1 + \sum_{k=1}^\infty c_k (-1)^k=1 - \sum_{k=1}^\infty \vert c_k\vert$ is finite.  

The key insight means that $f_m(-1)$ is a monotone decreasing sequence in $m$, and the limit $f(-1)$ exists $\iff \big\{ f_m(-1)\big\}$ is a bounded sequence. Define $g:[-1,0]\rightarrow \mathbb R$ given by $g(z) = 1-f(z) =\sum_{k=1}^\infty \vert c_k\vert\cdot \vert z\vert^k$ and $g_m(z)=1-f_m(z)=\sum_{k=1}^m \vert c_k\vert\cdot \vert z\vert^k$ note $0\leq g_{m}(z)\leq g_{m+1}(z)\leq \dots \leq g(z)$ [monotonicity] for $z\in (-1,0)$.  With $z_n :=\frac{1}{n}-1$ consider arbitrary $m\in \mathbb N$, and select $\epsilon \gt 0$  
$\implies 0\leq g_m(-1) \leq \big\vert g_m(-1)-g_m(z_n)\big\vert+g_m(z_n)\leq \big\vert g_m(-1)-g_m(z_n)\big\vert+ g(z_n)=\big\vert g_m(-1)-g_m(z_n)\big\vert+\left(1-\sqrt{\frac{1}{n}}\right)\lt \epsilon+1$  
by triangle inequality, then monotonicity, and the final inequality holds for all $n$ large enough since $g_m$ is a polynomial function (continuous). We have $0\leq g_m(-1)\leq 1$, a bounded sequence in $m$ hence $0 \leq f_m(-1)\leq 1$ the desired bounded sequence $\implies f(-1)$ exists and $0\leq f(-1)\leq 1$, i.e. $f(-1)=(1+-1)^\frac{1}{2}=1 - \sum_{k=1}^\infty \vert c_k\vert \geq 0\implies 1\geq\sum_{k=1}^\infty \vert c_k\vert$. (In fact this is met with equality though only the inequality is needed.)    

Conclude for $z\in \overline B\big(0,1\big)$ we have well defined $f(z)$ such that $1+z=f(z)\cdot f(z)=\Big(1 + \frac{1}{2} z +\big(\sum_{k=2}^\infty c_k z^k\big)\Big)^2 = 1+z + 2\cdot\big(1+\frac{z}{2}\big)\big(\sum_{k=2}^\infty c_k z^k\big) +\big(\sum_{k=2}^\infty c_k z^k\big)^2=1+z + h(z)$   
$\implies 0=h(z)=2\cdot\big(1+\frac{z}{2}\big)\big(\sum_{k=2}^\infty c_k z^k\big) +\big(\sum_{k=2}^\infty c_k z^k\big)^2$ hence after simplification, every coefficient of the power series of $h$ is equal to zero.   

- - -  
*Remark:* the above series information is useful elsewhere as e.g. it implies $p_n(x)=1 +\sum_{k=1}^\infty c_k (x^2-1)^k$ converges uniformly to $\vert x\vert$ on $[-1,1]$ which can be useful in proving Stone-Weierstrass.  

- - -  
Now for your bounded positive self-adjoint operator, we can re-scale by a positive number as need such that we may assume **WLOG** that  
$\mathbf 0\preceq B\preceq I\implies -I\preceq B-I\preceq \mathbf 0\implies \Vert B-I\Vert \leq 1$    
[ref addendum for more information on the RHS implication]  

Then $A:=\big(B\big)^\frac{1}{2}=\big(I + (B-I)\big)^\frac{1}{2}=f\big(B-I\big)=1 + \sum_{k=1}^\infty c_k (B-I)^k$  
and the series on the right converges since $\big\{f_m\big((B-I)\big)\big\}$ is Cauchy, i.e. for $n\geq m\geq N$ we have $\big\Vert f_m\big((B-I)\big)-f_n\big((B-I)\big)\big\Vert \leq \sum_{k=m+1}^n \vert c_k\vert \cdot \big\Vert B-I\big\Vert^k\leq \sum_{k=N}^\infty \vert c_k\vert \to 0$ as $N\to\infty$  

And as above $A^2=\Big(I + \sum_{k=1}^\infty c_k (B-I)^k\Big)^2 = \big(I + (B-I) + h(B)\big)=\big(I + (B-I)\big)=B$  
Note that when we say $f_m\big(B-I\big)\to f_m\big(B-I\big) = A$ we are referring to **convergence in norm** not something weaker like strong convergence. If we ignore the $\preceq$ line and just let $C=B-I$ be some operator that happens to observe $\Vert C\Vert \leq 1$ then all above the above applies verbatim to bounded operators on Banach spaces.  The below specializes to Hilbert spaces and finishes off the answer.   

*As an aside:* notice if we tried to apply this to a non-zero nilotent matrix $B$-- consider a $n\times n$ nilpotent matrix that has rank $n-1$-- even though all the spectrum of $B$ is zero, the above fails since $\Vert B-I\Vert \gt 1$.  Further such a $B$ cannot have a square-root $R$ otherwise $R^2=B$ and $\text{rank}\big(R\big)\in \big\{n-1,n\big\}$; if the rank is $n$ then $R$ is invertible and so is $B$ (absurd) and if the rank is $n-1$ then $\text{rank}\big(R\big)=n-1=\text{rank}\big(B\big)= \text{rank}\big(R^2\big)$ so eigenvalue $0$ is semi-simple (absurd).  This shows that even though $1=\sup_{\lambda \in \sigma(B-I)}\vert \lambda\vert =\lim_{k\to \infty} \Vert A^k\Vert^\frac{1}{k}$, the **spectral radius is not a sufficient consideration for convergence of a power series in context of boundary behavior.**  [It is of course sufficient when considering interior points, i.e. actual points of analyticity.] 

- - - -  
To confirm that $A=f\big(B-I\big)$ is a **positive** self-adjoint operator, for $\Vert x\Vert =1$ we have   
$\big\langle f\big(B-I\big)x,x\big\rangle=\big\langle I + \sum_{k=1}^\infty c_k (B-I)^k x,x\big\rangle= 1 -\sum_{k=1}^\infty \vert c_k\vert \big\langle  (I-B)^k x,x\big\rangle\geq 1 -\sum_{k=1}^\infty \vert c_k\vert \geq 0$  
The 2nd equality is justified by the 'key insight' (and linearity and continuity of inner products) and the 1st inequality uses $\mathbf 0 \preceq (I-B)^k\preceq  I$ since $I-B$ is positive and $\big\Vert (I-B)^k\big\Vert\leq \big\Vert I-B\big\Vert^k \leq 1$  
$\implies A=f\big(B-I\big)\succeq \mathbf 0$  
- - - - - 
**regarding uniqueness:**    
suppose $A^2=B=C^2$ for positive $A,C$. The map $T:B(H)\rightarrow B(H)$ given by $S\mapsto CS-SC$ is bounded with $\Vert T\Vert \leq 2\Vert C\Vert$ so $\ker T$ is closed, necessarily containing $C^2=B$ as well as $f_m\big(B-I\big)$ and its limit $f\big(I-B\big)=A\implies AC=CA$.  Write $H= V \oplus \ker(A+C)$, with $V:=\overline{\text{image }(A+C)}$, each of which are $A$ and $C$ invariant (commutativity). This implies $A_{\vert \ker (A+C)}=\mathbf 0$ since $A+C\succeq A\succeq \mathbf 0$ so $x\in \ker(A+C)\implies0=\langle (A+C)x,x\rangle\geq \langle Ax,x\rangle=\big\Vert A^\frac{1}{2}x\big\Vert^2\geq 0\implies \mathbf 0= A^\frac{1}{2} \mathbf 0=A^\frac{1}{2} \big(A^\frac{1}{2} x\big)=Ax$.  And similarly $C_{\vert \ker (A+C)}=\mathbf 0=A_{\vert \ker (A+C)}$.    
Finally $\mathbf 0 = B_{\vert V}-B_{\vert V}=(A_{\vert V}+C_{\vert V})(A_{\vert V}-C_{\vert V})\implies \mathbf 0 = A_{\vert V}-C_{\vert V}$ since $(A_{\vert V}+C_{\vert V})=(A+C)_{\vert V}$ is injective, so for $x \in V$ we have  $\mathbf 0=\mathbf 0x =(A_{\vert V}+C_{\vert V})(A_{\vert V}-C_{\vert V})x =(A_{\vert V}+C_{\vert V})y\implies y =\mathbf 0$. Conclude $A=C$.  

**addendum**  
The one key step is that  
$\mathbf 0\preceq B\preceq I\implies -I\preceq B-I\preceq \mathbf 0\implies \Vert B-I\Vert \leq 1$    
typically this is inferred by equating the operator norm and numerical radius for self-adjoint operators [ref e.g. Lemma 4.18 in MacCluer or Prop 6.1.5 of Eidelman et al for a direct argument or Pryce 18O for a roots of unity argument in a more general setting that implies the result].  However we could argue as follows:  

Now suppose for contradiction  that $\big\Vert I-B\big\Vert \gt 1$  
$\implies \big\Vert I-\tau B\big\Vert \gt 1$ for any $\tau\in \big(1-\delta, 1\big)$ for $\delta \gt 0$ small enough by continuity of norms.  Select and fix a satisfactory $\tau$    
$C:= I + (\tau B)+(\tau B^2)+\cdots\quad \text{(Neumann Series)}$  
$\implies   C(I-\tau B)=I$  
$\implies \big\langle Cv,v\big\rangle = \big\Vert v\big\Vert^2 + \sum_{k=1}^\infty \big\langle (\tau B)^k v,v\big\rangle=\geq \big\Vert v\big\Vert^2$  

For odd $k$, then $k=2m+1$ and recalling that $(\tau B)\succeq \mathbf 0$        
$\big\langle (\tau B)^k v,v\big\rangle=\big\langle (\tau B) \big((\tau B)^m v\big),\big((\tau B)^m v\big)\big\rangle\geq 0$  

And of course for even $k$, then $k=2m$    
$\big\langle (\tau B)^k v,v\big\rangle=\big\langle (\tau B)^m v,(\tau B)^m v\big\rangle=\big\Vert (\tau B)^m v\big\Vert^2\geq 0$  

Now by our assumption, there must be some unit length $x$ with $\big\Vert x\big\Vert \lt 1$ such that $\big\Vert (I-\tau B)x\big\Vert = 1$ and write $v=(I-\tau B)x$  
$\implies 1\gt \big\Vert x\big\Vert =\big\Vert C(I-\tau B)x\big\Vert =\big\Vert Cv\big\Vert\cdot 1\geq \big\langle Cv,v\big\rangle \geq \big\Vert v\big\Vert^2= 1$  
by Cauchy-Schwarz, and this is a contradiction  

Conclude $\big\Vert B-I\big\Vert \leq 1$ and the rest of the above proof for the square-root goes through as written.  

$\underline{\text{As an encore: }}w\big(T\big)=\sup\big\{\big\vert \langle Tx,x\rangle\big\vert :\big\Vert x\big\Vert=1\big\}= \big\Vert T\big\Vert\quad\text{ for self-adjoint}T$    


The easy part of the numerical radius argument is (by Cauchy-Schwarz) for any $T\in B(\mathscr H)$   
$w\big(T\big)=\sup\big\{\big\vert \langle Tx,x\rangle\big\vert :\big\Vert x\big\Vert=1\big\}\leq \big\Vert T\big\Vert$    

Now for self adjoint $T\in B(\mathscr H)$, we know  
$w\big(T^2\big)=\sup\big\{\big\vert \langle T^2x,x\rangle\big\vert :\big\Vert x\big\Vert=1\big\}=\sup\big\{\big\vert \langle Tx,Tx\rangle\big\vert :\big\Vert x\big\Vert=1\big\}= \big\Vert T\big\Vert^2$    

Using the above square root formulation, we know for *self-adjoint positive* $T\in B(\mathscr H)$ that $\sqrt{T}$ uniquely exists as a positive self-adjoint operator.     
$\implies w\big(T\big)= w\big(\sqrt{T}^2\big)=\big\Vert \sqrt{T}\big\Vert^2=\big\Vert T\big\Vert$    
where the final equality is Prop 2.14 in MacCluer [Cauchy-Schwarz based]  

Finally, for *self-adjoint* $T\in B(\mathscr H)$ since $T$ is normal we have Polar Decomposition [ref Brezis Problem 44(6) for a nice walk-through]  
$T=UP=U\sqrt{T^2}$ where $PU=UP$ and $U$ is unitary such that $U$ acts as the identity on $\ker T=\ker T^*$.  And $T^2= U^2P^2=U^2T^2\implies (I-U^2)T^2=\mathbf 0$.  Now write $\mathscr H = \ker T\oplus (\ker T)^\perp= \ker P\oplus (\ker P)^\perp =\ker T\oplus \overline{\text{image }P}$  
We already know that $U$ acts as an involution [group theory definition $U^2=\text{id}$ not $C^*$] on $\ker T$  

For $v\in \text{image }P$ then $v =Px\implies (I-U^2)v=(I-U^2)Px=\mathbf 0$  

More generally, for $v\in \overline{\text{image }P}$ then $v=\lim_{k\to\infty}v_k$ where $v_k =Px_k$  
$(I-U^2)v_k=(I-U^2)Px_k=\mathbf 0\implies (I-U^2)v=\mathbf 0$  by taking limits   
$\implies U^2v = v$ for all $v\in \ker T^\perp=\overline{\text{image }P}$  
Conclude $U^2=I\implies U^*=U^{-1}=U$ on $\mathscr H$ and  
$S:=\frac{U+I}{2}$ where $S^2=S$  and $S=S^*$ so $S$ is an orthogonal projection  

$\implies \mathscr H = \ker S \oplus \text{image }S=\ker S\oplus \ker S^\perp$ where for $v\in \ker S\iff Uv =-v$ and $v\in \text{image }S\iff Uv=v$  
Note that $SP=PS$ so each of these subspaces are $P$-invariant and since $US=SU$ they are of course $T$-invariant as well .  

So in terms of operator matrices we have  
$T:=\displaystyle \left[\begin{matrix}T_{\vert \ker S} & \mathbf 0\\ \mathbf 0& T_{\vert \text{image } S}\end{matrix}\right]$ and for any $v\in \mathscr H$   
$\big\Vert Tv\big\Vert^2 = \big\Vert \big(S+(I-S)\big)Tv\big\Vert^2=\big\Vert  STv\big\Vert^2+\big\Vert (I-S)Tv\big\Vert^2=\big\Vert  TSv\big\Vert^2+\big\Vert T(I-S)v\big\Vert^2= \big\Vert  T_{\text{image } S}Sv\big\Vert^2+\big\Vert T_{\vert \ker S}(I-S)v\big\Vert^2$  
$\leq \big\Vert  T_{\text{image } S}\big\Vert^2\cdot \big\Vert  Sv\big\Vert^2+\big\Vert T_{\vert \ker S}\big\Vert^2\cdot \big\Vert  (I-S)v\big\Vert^2$  
$\leq \max\big(\big\Vert  T_{\text{image } S}\big\Vert^2, \big\Vert T_{\vert \ker S}\big\Vert^2\Big)\cdot \Big( \big\Vert  Sv\big\Vert^2+ \big\Vert  (I-S)v\big\Vert^2\Big)$  
$=\max\big(\big\Vert  T_{\text{image } S}\big\Vert^2, \big\Vert T_{\vert \ker S}\big\Vert^2\Big)\cdot \big\Vert  v\big\Vert^2$  
$\implies \big\Vert T\big\Vert\leq \max\big(\big\Vert  T_{\text{image } S}\big\Vert, \big\Vert T_{\vert \ker S}\big\Vert\Big)$  

WLOG suppose $ \max\big(\big\Vert  T_{\text{image } S}\big\Vert, \big\Vert T_{\vert \ker S}\big\Vert\Big)= \big\Vert  T_{\text{image } S}\big\Vert$
then consider a sequence of unit length vectors $v_k \in \text{image }S$ such that $\big\Vert Tv_k\big\Vert \to \big\Vert  T_{\text{image } S}\big\Vert\implies   \big\Vert T\big\Vert=\big\Vert  T_{\text{image } S}\big\Vert$  

I.e. the norm of $T$ is 'reachable' via a sequence solely in the space on which $T$ is is positive [or negative if in $\ker S$].  

$\implies \big\Vert T\big\Vert=\big\Vert  T_{\text{image } S}\big\Vert=w\big( T_{\text{image } S}\big)=\sup\big\{\big\vert \langle Tx,x\rangle\big\vert :\big\Vert x\big\Vert=1 \text{ & } x\in \text{image }S\big\}\leq \sup\big\{\big\vert \langle Tx,x\rangle\big\vert :\big\Vert x\big\Vert=1 \big\}=w\big( T\big)\leq  \big\Vert T\big\Vert$   
The 2nd equality is because $T_{\text{image } S}$ is positive [if instead we restricted to $\ker S$ we'd notice $-1\cdot T_{\text{ker } S}$ is positive]  
The 1st inequality is because the supremum occurs over a bigger set and the 2nd inequality was the originally mentioned easy direction implied by Cauchy-Schwarz.  

- - - - -  
Conclusion: for positive self-adjoint $T\in B(\mathscr H)$, 
$w\big(T\big)= \big\Vert T\big\Vert$    
This is an easy consequence of $w\big(T^2\big)= \big\Vert T\big\Vert^2$ and applying that to $\sqrt{T}$.  

For self-adjoint $T\in B(\mathscr H)$ we still have $w\big(T\big)= \big\Vert T\big\Vert$ since we can restrict to the subspace on which $T$ is positive-- or restrict to the subspace on which it is negative and write $w\big(T\big)=w\big(-T\big)= \big\Vert- T\big\Vert= \big\Vert T\big\Vert$.    



**Additional note:**  
Our self-adjoint $T\in B\big(\mathscr H\big)$ had the operator matrix form of  
$T=\displaystyle \left[\begin{matrix}T_{\vert \ker S} & \mathbf 0\\ \mathbf 0& T_{\vert \text{image } S}\end{matrix}\right]=\displaystyle \left[\begin{matrix}U_{\vert \ker S}P_{\vert \ker S} & \mathbf 0\\ \mathbf 0& U_{\vert \text{image } S}P_{\vert \text{image } S}\end{matrix}\right]=\displaystyle \left[\begin{matrix}-IP_{\vert \ker S} & \mathbf 0\\ \mathbf 0& IP_{\vert \text{image } S}\end{matrix}\right]=\displaystyle \left[\begin{matrix}\mathbf 0 & \mathbf 0\\ \mathbf 0& P_{\vert \text{image } S}\end{matrix}\right]+\displaystyle \left[\begin{matrix}-P_{\vert \ker S} & \mathbf 0\\ \mathbf 0& \mathbf 0\end{matrix}\right]$   

which is a way of stating the common result that $T=M+N$ where $M,N$ are positive and $MN=\mathbf 0=NM$. We can of course re-write this as  
$T= PU = PUS+PU(I-S)= PIS+P(-I)(I-S)=PS-P(I-S)=SPS-(I-S)P(I-S)=M+N$  

Note that $M=SPS = SP = \frac{U+I}{2}P= \frac{1}{2}\big(T + P\big)$ and $M=(I-S)P(I-S) = (I-S)P = \frac{-U+I}{2}P= \frac{1}{2}\big(-T + P\big)$  
which is the more common way of stating this result.  


**Corollary:**  
If $T\in B(\mathscr H)$ is self-adjoint then $\big\Vert T\big\Vert \text{ or }-\big\Vert T\big\Vert \in \sigma(T)$.  Recall from the Neumann (geometric) series argument that if $\vert \lambda \vert\gt  \big\Vert T\big\Vert$ then $\big(\lambda I-T\big)$ is invertible, so what this says is that the maximum possible non-invertible point is in fact achieved. Via re-scaling, we can WLOG assume $\big\Vert T\big\Vert =1$.  

Start off with the *special case* that $T$ is positive hence $I\geq I-T\geq \mathbf 0$.  Now since $W(T)=1$ there is a sequence of unit length $v_k$ such that   
$\big\langle Tv_k,v_k\big\rangle \to 1^{-}\implies \big\Vert \sqrt{(I- T)}v_k\big\Vert^2 = \big\langle (I- T)v_k,v_k\big\rangle \to 0^{+}$    

Thus it must be the case that either  
(a.) $\dim \ker \sqrt{(I- T)}\gt 0\implies v_k \in \ker (I-T)$ so $T$ has an eigenvalue of $1$ or  
(b.) $\dim \ker \sqrt{I-T}=0\implies \text{image}\sqrt{I-T}\subsetneq \overline{\text{image}\sqrt{I-T}}$  
since $\big\Vert \sqrt{(I- T)}v_k\big\Vert$ is not bounded below [ex 3.27 in MacCluer or ex 5.3.4 in Eidelman et al] hence  $\text{image}\big(I- T\big)=\text{image}\big(\sqrt{(I- T)}\sqrt{(I- T)}\big)\subseteq \text{image}\big(\sqrt{(I- T)}\big)\subsetneq \mathscr H$ i.e. $(I-T)$ is not surjective   

$\implies 1\in \sigma(S)$.

The more general case of self-adjoint $T$ follows per the preceding argument since  $T=\displaystyle \left[\begin{matrix}T_{\vert \ker S} & \mathbf 0\\ \mathbf 0& T_{\vert \text{image } S}\end{matrix}\right]$ and  $\big\Vert T_{\vert \ker S} \big\Vert=\big\Vert T\big\Vert$ and/or $T_{\vert \text{image } S}=\big\Vert T\big\Vert$ so we may restrict to the subspace on which $T$ is either positive or $-T$ is positive and run the argument there.  

*remark:*  
This ends up being a very important development since the spectrum [points where $T$ is not invertible] is a purely algebraic property yet we have it links with the numerical range [a generalization of quadratic forms] and the norm of a self-adjoint operator.
