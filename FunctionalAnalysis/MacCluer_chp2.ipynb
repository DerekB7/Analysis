**2.17**  
If $T$ is a bounded self adjoint operator on $\mathscr H$ and $T^2=T$, show that $T$ is the orthogonal projection onto its image  

Recall the (sort of) definition of orthogonal projection (Theorem 1.24 + page 15 commentary, except the co-domain is assumed to be equal to the domain, not the erroneous definition given by MacCluer), which has not been cleanly given by MacCluer, but it is really part a and b of Theorem 1.24, namely that the image and kernel of $P$ are orthogonal   

note: the comment at the bottom of page 36 is that by the statement of Lemma 1.30 we see orthogonal projections (whose images are closed subspaces) are self-adjoint operators, so this exercise is to show the other direction. **Thus after this exercise we have the criterion** $T\in B(\mathscr H)$ is an orthogonal projection $\iff T^2=T$ and $T=T^*$  

the algebraic result implied by idempotence of $T$, that (which holds for any Banach Space)  
$\mathscr H = \text{image }T \oplus \ker T$  
is given in Pryce 10S (ii) and (iii) [developed below as part of ex 2.21]  

To show that the projection is orthogonal, we need to show $\text{image }T\perp \ker T$ which follows from the linear algebra argument that eigenvectors with different eigenvalues are orthogonal for self-adjoint operators, i.e.  
if $v\in \text{image }T$ and $w\in \ker T$  

$\big\langle v,w\big\rangle=\big\langle Tv,w\big\rangle=\big\langle v,Tw\big\rangle=\big\langle v,\mathbf 0\big\rangle  =0 $  

Note that since these are orthogonal subspaces and $\mathscr H$ is a direct sum of them, then they are orthogonal complements, so $ \text{image }T=\overline  {\text{image }T}$ by ex 1.12 (alternatively consider $\ker (T-I)$).  

**Conversely**  
If we have orthogonal projection $T$, i.e. $T^2=T$ and $\big(\text{image }T\big)=\big(\ker T\big)^\perp$  then $T\in B(\mathscr H)$ and in fact $T=T^*$, i.e. $T$ is self-adjoint.  The fact that $\big\Vert T\big\Vert \leq 1$ is shown a few lines below here [at the end of this cell].  

Using ex 1.12 we have  
$\big(\text{image }T\big)^\perp=\big(\ker T\big)^{\perp\perp} = \ker T$  
where $\ker T=T^{-1}\big\{\mathbf 0\big\}$ is closed since $T$ is continuous and similarly $I-T$ is idempotent so $\text{image }T = \ker \big(I-T\big)$ is closed.    

$\implies \mathscr H = \ker T\oplus \big(\ker T\big)^\perp =\ker T \oplus  \big(\text{image }T\big)^{\perp\perp} =  \ker T \oplus  \overline{\text{image }T}= \ker T \oplus \text{image }T$  


So for $x,y\in \mathscr H$ we have $x=p_1 + q_1$ and $y=p_2 + q_2$ with $p_i\in \text{image }T$ and $q_i \in \ker T$ and  
$\big\langle Px,y \big\rangle$  
$=\big\langle p_1,p_2 + q_2 \big\rangle$  
$=\big\langle p_1,p_2 \big\rangle+\big\langle p_1,q_2 \big\rangle$  
$=\big\langle p_1,p_2 \big\rangle+0 $  
$= \big\langle p_1,P(p_2 + q_2) \big\rangle$  
$=\big\langle p_1,P(p_2 + q_2) \big\rangle+\big\langle q_1,P(p_2 + q_2) \big\rangle$  
$= \big\langle p_1+q_1,P(p_2 + q_2) \big\rangle$  
$=\big\langle x,Py \big\rangle$  
$\implies \big\langle x,\big(P-P^*\big)y \big\rangle=0$  
for all $x,y\in \mathscr H$, including $x:=\big(P-P^*\big)y$ for arbitrary $y\in\mathscr H$ $\implies P=P^*$, so $P$ is self-adjoint.  

note: the argument can be streamlined to verifying for arbitrary $x\in\mathscr H$  
$\big\langle Px,x \big\rangle=\big\langle p_1,p_1+q_1 \big\rangle=\big\langle p_1,p_1 \big\rangle\in \mathbb R\implies P=P^*$ when $\mathscr H$ is a complex Hilbert-space [ref ex 4.12]

**2.23**  
We have been working thus far in the norm topology aka the uniform operator topology ("UOT").  Given sequence $T_n\in B(X)$ where $X$ is Banach,  

we say $T_n\to T\in B(X)$ in the *strong operator topology* ("SOT") if for each $x\in X$ we have  
$\big\Vert T_n x-Tx\big\Vert \to 0$  

note: it is immediate that if $T\to T_n$ [convergence in norm topology] then $T_n\to T$ SOT since  
$\big\Vert T_n x-Tx\big\Vert=\big\Vert (T_n -T)x\big\Vert\leq  \big\Vert T_n -T\big\Vert\cdot \big\Vert x\big\Vert\to 0$  


we say $T_n\to T\in B(X)$ in the *weak  operator topology* ("WOT")  
if for each $x\in X$ we have  
$\big\vert f(T_n x)-f(Tx)\big\vert \to 0$  
for each fixed $x\in X$ and $f \in X^*$  
note: it is immediate that if $T\to T_n$ SOT then $T_n\to T$ WOT since    
$\big\vert f(T_n x) -f(Tx)\big\vert=\big\vert f\big(T_nx -Tx\big)\big\vert\leq \big\Vert f\big\Vert\cdot \big\Vert T_n x -Tx\big\Vert\to 0$  

**Conclude:** $\text{convergence in UOT }\implies \text{convergence in SOT }\implies\text{convergence in WOT }$  


also note: MacCluer specializes to $X:=\mathscr H$ and so writes WOT in terms of inner products which via Riesz-Frechet is an equivalent definition
